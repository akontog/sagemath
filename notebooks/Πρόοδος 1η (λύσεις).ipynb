1. Σχεδιάστε τα παραπάνω σημεία.

2. Σχεδιάστε το πολύγωνο (είναι ρόμβος) με κορυφές τα σημεία Α, Β, C, D

3. Σχεδιάστε τις διαγωνίους χρησιμοποιώντας διαφορετικά χρώματα.

3. Βρείτε και σχεδιάστε το σημείο τομής, Ο, των διαγωνίων. Δίνεται:
$$
O\left(x_O, y_O\right) = \left( \frac{x_1 + x_3}{2}, \frac{y_1 + y_3}{2} \right)
$$
όπου (x1,y1), (x3,y3) απέναντι σημεία.

1. Ορίστε την παραπάνω συνάρτηση.

2. Βρείτε την παράγωγο της $f(x)$ και την εξίσωση της εφαπτομένης στο σημείο $x = 2$. Η εξίσωση εφαπτομένης σε σημείο a δίνεται: $y = f(a) + f'(a) \cdot (x - a)$

3. Βρείτε τα σημεία που παρουσιάζει τοπικά ακρότατα και προσδιορίστε αν είναι μέγιστα ή ελάχιστα.

4. Υπολογίστε το ορισμένο ολοκλήρωμα της $f(x)$ στο διάστημα $[0, 3]$.

5. Υπολογίστε την εξίσωση της εφαπτομένης στο σημείο $x=3.5$ κάνοντας χρήση του ορισμού του ορίου $f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h}$. Χρησιμοποιώντας το όριο η εξίσωση δίνεται: $y = f(a) + \left( \lim_{h \to 0} \frac{f(a + h) - f(a)}{h} \right) \cdot (x - a)$

6. Σχεδιάστε την γραφική παράσταση της $f(x)$ στο [-1,4] με γραμμοσκιασμένο το χωρίο στο διάστημα $[0, 3]$. Στην παράσταση να φαίνονται οι εφαπτομένες στα $x = 2$ και $x=3.5$ τα αντίστοιχα σημεία καθώς και τα σημεία στα οποία παρουσιάζει ακρότατα.

```python
# 1. Ορισμός της συνάρτησης
f(x) = x^3 - 6*x^2 + 9*x + 1
```

```python
# 2. Παράγωγος και Εφαπτομένη
fprime = f.derivative(x)
a=2
slope = fprime(x=a)
tangent = f(a) + slope * (x - a)
show(fprime)
show(tangent)
```

```python
# 3. Εύρεση κρίσιμων σημείων (όπου f'(x) = 0)
criticalPoints = solve(fprime == 0, x)
p1 = (criticalPoints[0].rhs(),f(criticalPoints[0].rhs()))
p2 = (criticalPoints[1].rhs(),f(criticalPoints[1].rhs()))

# προσδιορισμός μεγίστων ελαχίστων
fdoubleprime = diff(fprime, x)
secondderivativep1 =fdoubleprime(x=p1[0])
secondderivativep2 =fdoubleprime(x=p2[0])
secondderivativep1, secondderivativep2
```

```python
# 4. Ολοκλήρωμα
# Αόριστο
show(integrate(f,x))
# Ορισμένο
integralValue = integrate(f, (x, 0, 3))
integralValue
```

```python
# 5. Όριο και εφαπτομένη με χρήση ορίου
h = var('h')
a=3.5
fprimelimit = limit((f(a + h) - f(a)) / h, h = 0)
tangentLimit = f(a) + fprimelimit * (x - a)
show(tangentLimit)
```

```python
# 6. Σχεδίαση της συνάρτησης και της εφαπτομένης
plot_f = plot(f, (x, -1, 4), color="blue", legend_label="f(x)")
tangent_plot2 = plot(tangent, (x, 1.5, 2.5), color="red", linestyle="--", legend_label="Εφαπτομένη στο x=2")
tangent_plot35 = plot(tangentLimit, (x, 3, 4), color="purple", linestyle="--", legend_label="Εφαπτομένη στο x=3.5")
# Γραμμοσκιασμένο χωρίο για το ολοκλήρωμα στο διάστημα [0, 3]
area = plot(f, (x, 0, 3), fill=True, fillcolor="lightblue", alpha=0.3)

# Σημεία τοπικών ακροτάτων
p1_plot = point(p1, color="green", size=30)
p2_plot = point(p2, color="green", size=30)

# Σημείο εφαπτομένης  x=2
tp2 = point((2,f(2)), color="purple", size=30)

# Σημείο εφαπτομένης  x=3.5
tp35 = point((3.5,f(3.5)), color="purple", size=30)

# Συνδυασμός όλων των σχεδίων
final_plot = plot_f + tangent_plot2 + tangent_plot35 + area + p1_plot + p2_plot + tp2 + tp35
final_plot.show()
```
