# Άσκηση 1η - Προσέγγιση π

Το άθροισμα Basel που προσεγγίζει τον αριθμό 1.644934 είναι:
$$
\sum_{n=1}^{\infty} \frac{1}{n^2} = 1 + \frac{1}{2^2} + \frac{1}{3^2} + \frac{1}{4^2} + \cdots + ...
$$

Ο Euler βρήκε ότι το άθροισμα έχει την τιμή $\frac{π^2}{6}$. Έτσι μπορούμε να προσεγγίσουμε την τιμή του π βρίσκοντας το άθροισμα και υπολογίζοντας:
$$\pi \approx \sqrt{6 \sum_{n=1}^{\infty} \frac{1}{n^2}}$$

# Άσκηση 3η - Δέντρο του Πυθαγόρα

Το Δέντρο του Πυθαγόρα, pythagoras tree, (https://en.wikipedia.org/wiki/Pythagoras_tree_(fractal)) αποτελεί φράκταλ βασισμένο στο Πυθαγόρειο θεώρημα. Ξεκινά από ένα τετράγωνο, και σε κάθε βήμα χτίζονται δύο νέα τετράγωνα σχηματίζοντας σταδιακά ένα είδος «δέντρου».

Θα πρέπει να υλοποιήσετε αναδρομική συνάρτηση Python, έστω pythagoras_tree η οποία δέχεται:
    
    - δύο σημεία ως tuples τα οποία αποτελούν τις κορυφές της κάτω πλευράς του τετραγώνου

    - το τρέχων βάθος και

    - το μέγιστο βάθος του δέντρου.
```
pythagoras_tree(p1, p2, depth, maxdepth)
```
Θα πρέπει να υπολογίσετε τις άλλες δύο κορυφές, έστω p3, p4 και να σχεδιάσετε το τετράγωνο. Στην συνέχεια, θα πρέπει να καλείται δύο φορές η  pythagoras_tree με depth+1 και βάσεις των δύο αντίστοιχων τετραγώνων:

- p4 και top

- top και p3

όπου top η κορυφή του ορθογώνιου και, έστω, ισοσκελές τριγώνου με υποτίνουσα την p3p4 όπως φαίνεται και στο επόμενο σχήμα 
```
           top
           / \
          /   \
       p4 ----- p3
        |       |
        |       |
       p1 ----- p2
```
Όταν το depth γίνει ίσο με το maxdepth η διαδικασία σταματά.

Μπορείτε να χρωματίζετε τα τετράγωνα με τυχαίο χρώμα για ομορφότερο αποτέλεσμα. Υπενθυμίζεται ότι η `RR.random_element(0,1)` επιστρέφει (ψευδο)τυχαίο αριθμό από 0 έως 1.

# Άσκηση 5η - Τρίγωνο του Pascal 

Στα μαθηματικά, συνδυασμός των n στοιχείων ενός συνόλου A ανά k ονομάζεται κάθε υποσύνολο του συνόλου A με k στοιχεία.
Το πλήθος των συνδυασμών δίνονται από τον διωνυμικό συντελεστή:
$$
\binom{n}{k} = \frac{n!}{k! \cdot (n-k)!}
$$
Στο SageMath η μέθοδος `binomial(n,k)` επιστρέφει τον διωνυμικό συντελεστή.

Ο υπολογισμός μπορεί να γίνει και με αναδρομικό τρόπο βασισμένο στον επόμενο ορισμό:
$$
\binom{n}{k} = 
\begin{cases} 
1, & \text{αν } k = 0 \text{ ή } k = n \\ 
\binom{n-1}{k-1} + \binom{n-1}{k}, & \text{αν } 0 < k < n 
\end{cases}
$$

- Εφόσον υπάρχει μόνο ένας τρόπος να επιλέξεις κανένα στοιχείο k=0 ή όλα τα στοιχεία k=n έχουμε: $\binom{n}{k} = 1$

- Ο αριθμός των τρόπων για να επιλέξουμε k στοιχεία από n μπορεί να υπολογιστεί ως το άθροισμα $\binom{n-1}{k-1} + \binom{n-1}{k}$ δηλαδή το άθροισμα των επιλογών όταν το πρώτο στοιχείο περιλαμβάνεται στην επιλογή και όταν αυτό δεν περιλαμβάνεται.

# Άσκηση 7η - Γενική προσέγγιση σειράς Taylor με γεννήτορα

Έστω συνάρτηση $f(x)$ και σημείο επέκτασης $x_0$. Το ανάπτυγμα Taylor (μιας απείρως παραγωγίσιμης συνάρτησης) στη γειτονιά του $x_0$ είναι:
$$
f(x) = f(x_0) 
       + f'(x_0)(x - x_0) 
       + \frac{f''(x_0)}{2!}(x - x_0)^2 
       + \frac{f^{(3)}(x_0)}{3!}(x - x_0)^3 
       + \cdots
$$
Συμπαγέστερα το ανάπτυγμα μπορεί να γραφτεί και:
$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(x_0)}{n!} (x - x_0)^n
$$

```
taylor_generator(f, x, x0=0)
```
