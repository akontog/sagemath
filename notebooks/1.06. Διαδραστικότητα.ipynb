# Εισαγωγή στο @interact

Το SageMath υποστηρίζει **interactive widgets** μέσα από τον διακοσμητή `@interact`. Με αυτόν μπορούμε να δημιουργούμε μικρές διεπαφές χρήστη, **user interfaces**, που μας επιτρέπουν να αλλάζουμε τιμές παραμέτρων και να βλέπουμε **δυναμικά** το αποτέλεσμα (π.χ. γραφήματα, πίνακες, υπολογισμούς). Οι παράμετροι αλλάζουν μέσω κατάλληλων widgets και τα αποτελέσματα ενημερώνονται αυτόματα.

Τα βήματα δημιουργίας ενός Interact είναι τα εξής:

1. Υλοποιούμε μια συνάρτηση Python.

2. Προσθέτουμε την ένδειξη `@interact` στην αρχή της.

3. Ορίζουμε μια ή περισσότερες μεταβλητές ως `slider`, `checkbox` κ.λπ.

**Παράδειγμα**

Θα ορίσουμε την συνάρτηση $f(x) = e^{-x^2}\cos(ax)$ και θα την κάνουμε διαδραστική αλλάζοντας την παράμετρο a με τη βοήθεια `slider`.

````python
f(x, a) = exp(-x^2) * cos(a*x)
@interact
def inter_f(a = slider(1, 10, 1, default=5)):
    show(plot(f(x,a), (x, -pi, pi),ymin=-2, ymax=2, \
              legend_label='$'+latex(f(x,a))+'$'))
````

Χρησιμοποιώντας το ίδιο slider θα προσαρμόζουμε, ταυτόχρονα με την συνάρτηση, και την παράγωγό της $f'(x) = \frac{d}{dx}\Big(e^{-x^2}\cos(ax)\Big) = -2x e^{-x^2}\cos(ax) - b e^{-x^2}\sin(ax)$.

````python
f(x, a) = exp(-x^2) * cos(a*x)
f_prime(x, a) = diff(f(x, a), x)

@interact
def inter_f(a = slider(1, 10, 1, default=3)):
    g = plot(f(x,a), (x, -pi, pi),ymin=-2, ymax=2, \
              legend_label='$'+latex(f(x,a))+'$')
    g += plot(f_prime(x, a), (x, -5, 5), color='red', \
              legend_label='$'+latex(f_prime(x,a))+'$')
    show(g)
````

Έχουμε την δυνατότητα η διαδραστική εφαρμογή μας να έχει παραπάνω από ένα slider. Η συνάρτησή μας, $f(x) = be^{-x^2}\cos(ax)$, θα έχει πλέον και παράμετρο $b$ η οποία προσαρμόζεται από ένα δεύτερο slider.

````python
var('a, b')
f(x, a, b) = b * exp(-x^2) * cos(a*x)

@interact
def inter_f(
    a=slider(1, 10, 1, default=5, label='Συχνότητα a'),
    b=slider(0.1, 2, 0.1, default=1, label='Πλάτος b')
):
    p = plot(f(x, a, b), (x, -pi, pi), ymin=-2, ymax=2, \
             color='darkorange', thickness=2,
             legend_label='$'+latex(f(x,a))+'$')
    show(p)
````

# Widgets

Τα πιο συνηθισμένα widgets που υποστηρίζονται είναι:

- `slider`: ρυθμίζει μια αριθμητική τιμή (π.χ. πολλαπλασιαστής σε μια συνάρτηση).

- `selector`: επιτρέπει την επιλογή από μια λίστα επιλογών (π.χ. ποια τριγωνομετρική συνάρτηση να σχεδιάσουμε: `sin`, `cos`, `tan`).

- `checkbox`: ενεργοποιεί ή απενεργοποιεί την εμφάνιση κάποιου στοιχείου (π.χ. εμφάνιση/απόκρυψη ενός γραφήματος).

- `input_box`: για εισαγωγή ελεύθερου κειμένου ή αριθμών.

## slider

Το `slider` είναι από τα πιο χρήσιμα widgets για διαδραστικότητα, αφού μας επιτρέπει να αλλάζουμε την τιμή μιας παραμέτρου. Η σύνταξη της έχει την μορφή `slider(start, stop, step, default, label=None)` όπου:

- `start`: Το κατώτερο όριο του `slider`.

- `stop`: Το ανώτερο όριο του `slider`.

- `step`: Το βήμα με το οποίο μετακινείται το slider, π.χ. `step=0.1` για δεκαδικά βήματα.

- `default`: Η αρχική τιμή που παίρνει το slider όταν φορτώσει το interact.

- `label`: Μια προαιρετική  ετικέτα που εμφανίζεται δίπλα στο `slider` για επεξηγήσεις.

**Παράδειγμα**

Θα δημιουργήσουμε μια δυναμική τριγωνομετρική συνάρτηση $f(x)=a⋅sin(bx+c)$ με τρεις sliders:

- Slider a: ρυθμίζει το πλάτος του ημιτόνου.

- Slider b: ρυθμίζει τη συχνότητα.

- Slider c: μετατοπίζει τη συνάρτηση κατά φάση στον άξονα x.

````python
f(x, a, b, c) = a * sin(b*x + c)
@interact
def plot_trig(a = slider(1, 5, 0.1, default=2,label="Πλάτος"),
              b = slider(1, 10, 0.1, default=3,label="Συχνότητα"),
              c = slider(0.0, n(2*pi), 0.1, default=0,label="Φάση")):
    a_fmt = f"{a:.2f}"
    b_fmt = f"{b:.2f}"
    c_fmt = f"{c:.2f}"

    label = f"$f(x) = {a_fmt} \\sin({b_fmt}x + {c_fmt})$"
    g = plot(f(x, a, b, c), (x, 0, 2*pi), color='blue', legend_label=label)
    show(g)
````

## range_slider

Ενώ η `slider` επιτρέπει την επιλογή μιας τιμής σε κάποιο διάστημα, η `range_slider` επιτρέπει την επιλογή κάτω και πάνω ορίου κάνοντάς την ιδανική για επιλογή διαστήματος. Σημειώνεται ότι η `range_slider`, αντίθετα από την `slider` που επιστρέφει τιμή, θα επιστρέψει `tuple` με το άνω και κάτω επιλεγμένο όριο.

**Παράδειγμα**

Στο επόμενο interact ο χρήστης μπορεί να επιλέξει το διάστημα στο οποίο θα εμφανίσει την συνάρτηση.

````python
@interact
def example_range_slider(interval = range_slider(-5, 5, 0.5, default=(-2, 2), label='Διάστημα')):
    x1, x2 = interval
    show(plot(sin(x), (x, x1, x2), color='green', thickness=2))
````

## selector

Το `selector` μας επιτρέπει να διαλέγουμε μια (ή περισσότερες) επιλογές από μια λίστα επιλογών. Είναι πολύ χρήσιμο όταν θέλουμε να αλλάζουμε μεταξύ συναρτήσεων, μεθόδων ή παραμέτρων. Η σύνταξη της έχει την μορφή `selector(options, default=None, label=None, buttons=False, nrows=None, ncols=None)` όπου:

- `options` : Η λίστα με τις διαθέσιμες επιλογές.

- `default` : Η αρχική τιμή (προεπιλογή).

- `label` : Προαιρετική ετικέτα δίπλα στο widget.

- `buttons` : False → εμφανίζει drop-down μενού, True → εμφανίζει κουμπιά για κάθε επιλογή.

- `nrows`, `ncols` : Αν έχουμε πολλά κουμπιά, ρυθμίζουν πόσες σειρές / στήλες θα εμφανιστούν.

**Παράδειγμα**

Στο επόμενο παράδειγμα θεωρούμε τις τρεις βασικές τριγωνομετρικές συναρτήσεις $\sin(x)$, $\cos(x)$, $\tan(x)$. Κατασκευάζουμε διαδραστική εφαρμογή που επιτρέπει στον χρήστη να επιλέγει μέσω `selector`:

- ποια τριγωνομετρική συνάρτηση θέλει να προβάλλεται,

- το χρώμα της γραφικής παράστασης.

````python
@interact
def g(func = selector(['sin', 'cos', 'tan'],default='cos', label='Τριγωνομετρική συνάρτηση'),
     color = selector(['blue', 'red', 'green', 'orange'], default='blue', label='Χρώμα γραφήματος')
    ):
    if func == 'sin':
        f(x) = sin(x)
    elif func == 'cos':
        f(x) = cos(x)
    else:
        f(x) = tan(x)

    label = '$'+latex(f)+'$'
    p = plot(f, (x, -pi, pi), ymin=-1,ymax=1,color=color, thickness=2, legend_label=label)
    show(p, figsize=5)
````

## checkbox

Το `checkbox` είναι το πιο απλό εργαλείο interactivity στο SageMath. Επιστρέφει λογική τιμή True ή False, ανάλογα με το αν είναι επιλεγμένο ή όχι. Χρησιμοποιείται όταν θέλουμε να ελέγχουμε την ενεργοποίηση ή απενεργοποίηση μιας λειτουργίας. Η σύνταξή της έχει την μορφή: `checkbox(default=True, label=None)` όπου:

- `default`: Αρχική τιμή του checkbox (True = τσεκαρισμένο).

- `label`:  Κείμενο που εμφανίζεται δίπλα στο κουτάκι.

**Παράδειγμα**

Στην συνέχεια με την βοήθεια δύο checkboxes ο χρήστης του interact μας έχει την δυνατότητα να επιλέξει αν θα φαίνεται η συνάρτηση ή/και τα διαδοχικά σημεία της.

````python
f(x) = sin(x)
@interact
def g(show_curve = checkbox(default=True, label="Δείξε καμπύλη"),
      show_points = checkbox(default=False, label="Δείξε σημεία")):
    
    p = Graphics()
    
    if show_curve:
        p += plot(f(x), (x, -pi, pi), color='blue', legend_label="$f(x)=sin(x)$")
    
    if show_points:
        pts = [(a, f(a)) for a in srange(-pi, pi, pi/8)]
        p += point(pts, color='red', size=30, legend_label="Σημεία")
    
    show(p)
````

## input_box

Το `input_box` επιτρέπει στον χρήστη να εισάγει ελεύθερο κείμενο ή αριθμούς κατά την εκτέλεση ενός interact. Συντάσσεται ως εξής:  `input_box(default=None, label=None, type=None)`

- `default`: η αρχική τιμή που εμφανίζεται.

- `label`: το κείμενο που φαίνεται αριστερά από το πεδίο.

- `type`: μετατρέπει την είσοδο σε συγκεκριμένο τύπο (`int`, `float`, `str` κ.λπ.).

````python
@interact
def f(name = input_box("Κόσμε", label="Όνομα:", type=str)):
    show("Γεια σου " + name + "!")
````

**Παράδειγμα**

Στο πολυώνυμο δευτέρου βαθμού, $f(x)=ax^2+bx+c$ έχουμε τρία ελεύθερα πεδία κειμένου, ένα για κάθε συντελεστή, που όμως επιτρέπουν μόνο πραγματικές τιμές για είσοδο.

````python
@interact
def poly2nd(a = input_box(1, label="a:", type=float),
              b = input_box(0, label="b:", type=float),
              c = input_box(0, label="c:", type=float)):
    
    f(x) = a*x^2 + b*x + c
    
    xv = -b/(2*a)
    yv = f(xv)
    
    # Γράφημα
    p = plot(f(x), (x, -10, 10), color='blue', legend_label=f"$f(x)={a}x^2+{b}x+{c}$$")
    p += point((xv, yv), color='red', size=40, legend_label="Κορυφή")
    
    show(p)
    print(f"Κορυφή: ({float(xv):.2f}, {float(yv):.2f})")
````

## Συνδυασμός widgets

Συχνά χρησιμοποιούμε περισσότερα από ένα widget μαζί. Με τον τρόπο αυτό φτιάχνουμε πιο ευέλικτα και διαδραστικά περιβάλλοντα.

**Παράδειγμα**

Θα υλοποιήσουμε ένα interact στο οποίο θα σχεδιάσουμε την συνάρτηση $f(x) = x^2$ καθώς και την εφαπτομένη της σε ένα σημείο a. Ένα slider θα αλλάζει το σημείο $a$ στο οποίο εξετάζουμε τη συνάρτηση. Επιπλέον, ένα checkbox μας επιτρέπει να δείξουμε ή να κρύψουμε την εφαπτομένη.

````python
f(x) = x^2

@interact
def tangent_plot(show_tangent = checkbox(default=True, label="Εμφάνιση εφαπτομένης"),
                 a = slider(-3, 3, 0.1, default=1, label="Σημείο x")):
    g = Graphics()
    g += plot(f(x), (x, -3, 3), color='blue', legend_label="$f(x)=x^2$")
    point_a = (a, f(a))
    g += point(point_a, color='red', size=30, legend_label="Σημείο")
    if show_tangent:
        f_prime(x) = diff(f(x), x)
        tangent = f_prime(a)*(x - a) + f(a)
        g += plot(tangent, (x, -3, 3), color='green', linestyle='--', legend_label="Εφαπτομένη")
    show(g)
````

Η μέθοδος `taylor` στο SageMath υπολογίζει το **ανάπτυγμα Taylor** μιας συνάρτησης γύρω από ένα σημείο $a$, μέχρι μια επιθυμητή τάξη $n$. Η γενική μορφή είναι `f.taylor(x, a, n)` όπου:

- f είναι η συνάρτηση,

- x η μεταβλητή,

- a το σημείο γύρω από το οποίο γίνεται η ανάπτυξη,

- n η τάξη του αναπτύγματος.

Έτσι, για την συνάρτηση %f(x)=e^x% το ανάπτυγμα taylor μπορεί να βρεθεί με τον τρόπο που ακολουθεί.

````python
f(x) = exp(x)
f.taylor(x, 0, 5)
````

**Παράδειγμα**

θα δημιουργηθεί ένα interact το οποίο δημιουργεί αναπτύγματα taylor για τις συναρτήσεις που θέτει ο χρήστης σε αντίστοιχο input_box. Επιπλέον, μέσω ενός `slider` τίθεται ο αριθμός όρων της σειράς. Στο γράφημα θα φαίνονται όλα τα αναπτύγματα με αριθμό όρων μέχρι και το επιλεγμένο από τον `slider`.

````python
var('x')
@interact
def taylor_expansion(
    f = input_box(default=sin(x), label="Συνάρτηση f(x):"),
    nn = slider(1, 30, 1, default=3, label="Αριθμός όρων:")
):
    xmin, xmax = 0, 6*pi
    g = Graphics()
    g += plot(f, (x, xmin, xmax), color='blue', thickness=2)

    for n in range(1, nn+1):
        T = f.taylor(x, 0, n)
        g += plot(T, (x, xmin, xmax),
                  linestyle='--')
    show(g, figsize=(6,4), ymin=-5, ymax=5, gridlines=True)
````

# Δημοσίευση

Μόλις έχει δημιουργηθεί ένα διαδραστικό παράδειγμα στο SageMath μπορεί να ενσωματωθεί σε ιστοσελίδα και να διαμοιραστεί. Η δυνατότητα αυτή είναι πολύ χρήσιμη, γιατί δεν απαιτείται η εγκατάσταση ή η γνώση του SageMath από τους χρήστες του παραδείγματος.

## HTML

Αρχικά, για να δημοσιευτεί ένα παράδειγμα σε κάποια ιστοσελίδα, θα πρέπει να δημιουργηθεί μια σελίδα HTML (HyperText Markup Language) που αποτελεί γλώσσα σήμανσης που χρησιμοποιείται για τη δημιουργία και τη δομή των ιστοσελίδων. Δεν είναι γλώσσα προγραμματισμού απλώς περιγράφει τη δομή της σελίδας.

Θα πρέπει να δημιουργηθεί ένα αρχείο με κατάληξη .html, έστω _interactive.html_, στον υπολογιστή σας με το περιεχόμενο να έχει παρόμοια με την επόμενη μορφή.
```
<!DOCTYPE html>
<html lang="el">
<head>
  <meta charset="UTF-8">
  <title>Παράδειγμα SageMath</title>
</head>
<body>
<h1>Διαδραστικό Παράδειγμα με SageMath</h1>
<p>Οδηγίες
<!-- Εδώ μπορούμε να προσθέσουμε διάφορες οδηγίες 
χρήσης για την εφαρμογή μας -->
</p>

<div class="sage-interact">
<!-- Εδώ θα "παίξει" το interact μας-->
</div>
</body>
</html>
```
Αναλυτικά:

- `<!DOCTYPE html>`: Δηλώνει ότι το αρχείο είναι HTML5 — βοηθά τον browser να το καταλάβει σωστά.

- `<html lang="el"> … </html>`: Η ιδιότητα lang="el" δηλώνει ότι το περιεχόμενο είναι στα ελληνικά. Η ετικέτα `<html> … </html>` καθορίζει την αρχή και το τέλος μιας σελίδας.

- `<head> … </head>`:Περιέχει πληροφορίες για τη σελίδα (όχι περιεχόμενο που βλέπει ο χρήστης), όπως:

    - το `charset="UTF-8"` για σωστή εμφάνιση ελληνικών χαρακτήρων,

    - τον τίτλο (`<title> … </title>`) που φαίνεται στην καρτέλα του browser,

    - συνδέσεις με CSS ή JavaScript αρχεία.

- `<body> … </body>`: ό,τι φαίνεται στην ιστοσελίδα: κείμενα, εικόνες, κουμπιά, διαδραστικά στοιχεία κ.λπ.

- `<h2> … </h2>`: Επικεφαλίδα (heading). Υπάρχουν από `<h1>` (μεγαλύτερη) έως `<h6>` (μικρότερη).

- `<p> … </p>`: Παράγραφος κειμένου.

- `<!-- … -->`: Σχόλια, δεν εμφανίζονται στη σελίδα, αλλά χρησιμεύουν για επεξηγήσεις μέσα στον κώδικα.

- `<div class="sage-interact"> ... </div>`: Περιοχή περιεχομένου. Η ιδιότητα `class="sage-interact"` προσδίδει στην ετικέτα έναν προσδιορισμό ώστε να μπορεί στη συνέχεια να τεθεί ως ετικέτα στην οποία θα περιέχεται το interact.

Μπορείτε να δείτε την σελίδα σας με τον αγαπημένο σας browser. Για να τροποποιήσετε το περιεχόμενο όμως θα πρέπει να ανοίξετε το αρχείο με κάποιο πρόγραμμα επεξεργασίας κειμένου (notepad, sublime κλπ.).

## Προσθήκη κώδικα SageMath

Για να προστεθεί ο κώδικας Sage στην σελίδα θα πρέπει να συνδεθεί με το Sage cell. Η σύνδεση γίνεται στην ετικέτα head. Εκεί θα πρέπει να προστεθεί ο κώδικας.
```
…
<head>
  <meta charset="UTF-8">
  <title>Παράδειγμα SageMath</title>
  <script src="https://sagecell.sagemath.org/static/jquery.min.js"></script>
  <script src="https://sagecell.sagemath.org/embedded_sagecell.js"></script>
  <script>
    $(function () {
      sagecell.makeSagecell({
        inputLocation: '.sage-interact',
        template: sagecell.templates.interact,
        evalButtonText: '',
        autoeval: true,
        hide: ['editor', 'evalButton']
      });
    });
  </script>
</head>
...
```
Οι παράμετροι που έχουν τεθεί για το κελί μας είναι:

- `inputLocation: '.sage-interact'`: Ποια στοιχεία της σελίδας θα μετατραπούν σε SageCell (εδώ: όσα έχουν class="sage-interact")

- `template: sagecell.templates.interact`: Ορίζει ότι θα χρησιμοποιηθεί το πρότυπο “interact” για διαδραστικά widgets.

- `evalButtonText`: Το κείμενο που θα εμφανίζεται στο κουμπί εκτέλεσης

- `autoeval`: Αν έχει τεθεί true ενεργοποιεί την αυτόματη εκτέλεση του κώδικα μόλις φορτωθεί η σελίδα, χωρίς να χρειαστεί ο χρήστης να πατήσει το κουμπί “Εκτέλεση”

- `hide`: Ορίζει τι θα κρυφτεί από την σελίδα. Εδώ κρύβεται ο editor και το κουμπί εκτέλεσης.

Στην συνέχεια και στο σημείο που έχει υποδειχθεί από την παράμετρο `inputLocation` θα τεθεί το interact.
```
…
<div class="sage-interact">
<script type="text/x-sage"> 
@interact
def my_function(a = slider(1, 10, 1, default=5)): 
  show(plot(sin(a*x), (x, -pi, pi))) 
</script>
</div>
...
```

## Εφαρμογή - Σειρές fourier

Στην επόμενη εφαρμογή, δημιουργούμε μια διαδραστική ιστοσελίδα στην οποία περιλαμβάνονται τρία κελιά Sage καθώς και επεξηγήσεις ενός φαινομένου. Αρχικά, παρουσιάζονται μόνο τα interacts και στην συνέχεια όλος ο κώδικας της σελίδας.

````python
@interact
def square_wave_approximation(
    N = slider(1, 15, 1, default=3, label="Αριθμός αρμονικών (N)")
):
    f = (4/pi)*sum(((1/(2*k-1))*sin((2*k-1)*2*pi*x) for k in range(1, N+1)))
    g = Graphics()
    g += plot(f, (x, 0, 1), color='blue', thickness=2,\
              legend_label=f'{N} αρμονικές')
    g += line([(0,0), (1,0)], color='black', linestyle='--')
    
    show(g, figsize=(6,4), gridlines=True) 
````

````python
@interact
def fourier_harmonics_breakdown(
        N = slider(1, 25, 1, default=5, \
        label="Αριθμός αρμονικών (N)") 
):
    g = Graphics()
    harmonics = []
    for k in range(1, N+1):
        harmonic = (4/pi) * (1/(2*k-1)) * sin((2*k-1)*2*pi*x)
        harmonics.append(harmonic)
        g += plot(harmonic, (x, 0, 1), thickness=1, 
                 legend_label=f'Αρμονική {k}')
    total_sum = sum(harmonics)

    g += plot(total_sum, (x, 0, 1), color='red', thickness=2, 
             legend_label=f'Άθροισμα {N} αρμονικών')
    show(g, figsize=(6,4), gridlines=True, ymin=-1.5, ymax=1.5, 
         title=f'Αποσύνθεση σε {N} αρμονικές')
````

````python
@interact
def fourier_square_comparison(
    N = slider(1, 25, 1, default=5, label="Αριθμός αρμονικών (N)")
):
    f = (4/pi)*sum(((1/(2*k-1))*sin((2*k-1)*2*pi*x) for k in range(1, N+1)))
    square = piecewise([((0, 0.5), 1), ((0.5, 1), -1)])
    g = Graphics()
    g += plot(f, (x, 0, 1), color='blue', thickness=2, \
              legend_label=f'{N} αρμονικές')
    g += plot(square, (x, 0, 1), color='red', thickness=2, \
            legend_label='Τετραγωνικό κύμα')
    
    show(g, figsize=(6,4), gridlines=True, ymin=-1.5, ymax=1.5)
````
