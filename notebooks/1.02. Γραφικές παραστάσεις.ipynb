## Η `plot`

Η συνάρτηση `plot` επιτρέπει την σχεδίαση της γραφικής παράστασης συνάρτησης f(x) για $-1.0 \le x \le 1.0$.

## Αναλογία Διαστάσεων

Είναι πιθανό το SageMath να σχεδιάσει μια συνάρτηση με διαφορετικές κλίμακες στους άξονες. Για παράδειγμα αν ζητήσουμε το πάνω ημικύκλιο, δηλαδή $y = \sqrt{1 - z^2}$, αυτό θα εμφανιστεί παραμορφωμένο. Υπενθυμίζεται ότι πέρα από την x, θα πρέπει όλες οι συμβολικές μεταβλητές να δηλώνονται πριν την χρήση τους όπως φαίνεται και στην επόμενη υλοποίηση.

```python
var('z')
y=sqrt(1-z^2)
plot(y)
```

Το `aspect_ratio` καθορίζει την αναλογία της μονάδας μήκους του άξονα Oy ως προς την αντίστοιχη του Ox. Έτσι, θέτοντας την παράμετρο `aspect_ratio` ίση με 1 διατηρείται η σωστή γεωμετρία του ημικυκλίου και αποδίδεται το σχήμα με τις σωστές αναλογίες.

```python
plot(y,aspect_ratio=1)
```

Φυσικά, η παράμετρος `aspect_ratio` μπορεί να πάρει και άλλες τιμές ανάλογα με το επιθυμητό αποτέλεσμα. Στο επόμενο παράδειγμα οι μονάδες διπλασιάζονται στον y σε σχέση με τον x άξονα.

```python
plot(y,aspect_ratio=2)
```

Όλα τα παραπάνω παραδείγματα μας αποδεικνύουν ότι μπορούμε να κάνουμε το γράφημα μιας οποιασδήποτε συμβολικής έκφρασης που ορίζεται με χρήση αριθμητικών μεταβλητών (που έχουν ορισθεί μέσω `var` ή του x). Με την `plot` μπορούμε να κάνουμε και το γράφημα μιας οποιασδήποτε συνάρτησης μιας μόνο μεταβλητής! Είναι καλή τακτική να χρησιμοποιούμε μέσα στην `plot` μόνο το όνομα της συνάρτησης που θέλουμε να σχεδιάσουμε (π.χ. της f) και όχι το συνδυασμό του ονόματος της συνάρτησης με την μεταβλητή της (π.χ. f(z)). Έτσι το `plot(f, -3, 3)` είναι πιο ασφαλές από το `plot(f(z), -3, 3)` σε κάποιες περιπτώσεις περίπλοκων συναρτήσεων f. Το f(z) δεν είναι σύμβολο συνάρτησης, αλλά μια συμβολική έκφραση που ίσως να δημιουργήσει πρόβλημα στην `plot.` Για παράδειγμα, το `plot(sin,-3,3)` είναι καλύτερο από το `plot(sin(x),-3,3)` ή το `plot(sin(x), (x, -3, 3))`.

## Χρώμα

Το χρώμα μπορεί να τεθεί και από τον χρήστη με την παράμετρο `color`. Η τιμή της παραμέτρου μπορεί να τεθεί με έναν από τους επόμενους τρόπους:

- ένα λεκτικό, με πεζούς λατινικούς χαρακτήρες `red`, `green`, `blue`, `black` κ.ο.κ.

- μια τριάδα των βασικών χρωμάτων (r, g, b) όπου κάθε χρώμα αντιστοιχεί σε ένα δεκαδικό αριθμό από 0 έως 1.

- ένας δεκαεξαδικός κωδικός 6 ψηφίων με κάθε δυάδα να αναφέρεται και πάλι στα τρία βασικά χρώματα.

Στην πραγματικότητα ο τρόπος αναπαράστασης χρωμάτων στον SageMath χρησιμοποιεί True Color δηλαδή 24 bits (binary digit 0 ή 1) πληροφορίας. Τα 24 bits χωρίζονται σε 3 ομάδες των 8 bits (ή 1 byte) για καθένα από τα τρία βασικά χρώματα: 

- κόκκινο (red)

- πράσινο (green)

- μπλε (blue).

Με αυτόν τον τρόπο, όμως, κάθε βασικό χρώμα μπορεί να πάρει τιμές από 0 (=00000000) έως 255 (=11111111) ή στο δεκαεξαδικό σύστημα 00 έως ff. Στο σύνολο μπορούμε να έχουμε $2^24$ ή 256x256x256 = 16.777.216 συνδυασμούς και άρα διαφορετικά χρώματα.

Παραδείγματα:

- Μαύρο: #000000 (απουσία χρώματος)

- Λευκό: #ffffff (και τα 3 βασικά χρώματα στο μέγιστό τους)

- Γκρι: #a0a0a0 (ίση ποσότητα από όλα τα χρώματα)

- Κόκκινο: #ff0000

- Πράσινο: #0000ff

- Κίτρινο: #ffff00 (συνδυασμός κόκκινου και πράσινου)

**Παράδειγμα**

Στο παρακάτω παράδειγμα βάζουμε μέσα σε μια λίστα τις συναρτήσεις sin(x), cos(x), sin(2x) και cos(2x), και στην συνέχεια τις σχεδιάζουμε μαζί με το ίδιο όμως χρώμα `color=(1, 0.5, 1)`.

```python
plot([sin(x),cos(x),sin(2*x),cos(2*x)], 0, 3*pi, color=(1,0.5,1))
```

Για να έχουμε την δυνατότητα να θέσουμε ξεχωριστά χρώματα σε κάθε συνάρτηση θα πρέπει:

-  είτε να χρησιμοποιήσουμε την color με μια λίστα από τα διαφορετικά χρώματα που θέλουμε να χρησιμοποιήσουμε για κάθε συνάρτηση (π.χ. ’#ff00af’ για την πρώτη, (0.3,0.2,0.6) για την δεύτερη, κοκ)

- είτε να προσθέσουμε τα plots των συναρτήσεων όπου η κάθε συνάρτηση έχει το δικό της χρώμα.

```python
plot([sin(x), cos(x), sin(2*x), cos(2*x)], (x, 0, 3*pi), \
     color=['#ff00af', (0.3,0.2,0.6), 'green', 'orange'])
```

```python
p1 = plot(sin(x), (x, 0, 3*pi), color='red')
p2 = plot(cos(x), (x, 0, 3*pi), color='blue')
p3 = plot(sin(2*x), (x, 0, 3*pi), color='green')
p4 = plot(cos(2*x), (x, 0, 3*pi), color='orange')
p = p1 + p2 + p3 + p4
p
```

Σημείωση: τα γραφήματα σχεδιάζονται από τα αριστερά στα δεξιά με το αριστερότερο να βρίσκεται κάτω από το επόμενό του κοκ.

Σε πολλές περιπτώσεις, μεταξύ των οποίων και η υπέρθεση γραφημάτων, υπάρχει περίπτωση μια εντολή να μην χωράει σε μία γραμμή. Στην περίπτωση αυτή ο τελεστής `\` επιτρέπει την συνέχιση της εντολής στην επόμενη γραμμή.

```python
plot((x-1)*(x-2)*(x-3), (x, 1, 5), color='red') + \
plot((x-1)*(x-2)*(x-4), (x, 1, 5), color='green') +\
plot((x-1)*(x-2)*(x-5), (x, 1, 5), color='blue') + \
plot((x-1)*(x-2)*(x-6), (x, 1, 5), color='pink') +\
plot((x-1)*(x-2)*(x-7), (x, 1, 5), color='purple')
```

## Υπόμνημα

Σε γραφήματα που περιέχουν περισσότερες από μια συναρτήσεις ή δεδομένα, το υπόμνημα (legend στα αγγλικά), βοηθά στην αντιστοίχιση και στην κατανόηση του γραφήματος. Η παράμετρος `legend_label` θέτει το υπόμνημα κάθε συνάρτησης.

```python
plot([sin(x), cos(x), sin(2*x), cos(2*x)], (x, 0, 3*pi), 
     color=['red', 'blue', 'green', 'orange'], legend_label=['sin(x)', 'cos(x)', 'sin(2x)', 'cos(2x)'])
```

**Παράδειγμα**

Θα σχεδιάσουμε τη γραφική παράσταση της παραβολής $f(x) = x^2 + 1$ στο διάστημα $[-2,2]$, καθώς και την εφαπτομένη στο $x_0=1$  (Είναι γνωστό ότι η ζητούμενη εφαπτόμενη στο σημείο $x_0=1$ της $f(x)$ έχει εξίσωση $y = 2x$).

Σημειώσεις:

-  Οι εφαπτόμενες ευθείες σε μια καμπύλη f(x) υπολογίζονται εύκολα με την παράγωγο της $f'(x)$. Σε επόμενη ενότητα θα δούμε πώς υπολογίζουμε τις παραγώγους με τη βοήθεια του Sage.

- Με την χρήση των δολαρίων `$` στο `legend_label` επιτρέπεται η προσθήκη μαθηματικού κειμένου (γινόμενα, εκθετικά, ρίζες, λογάριθμοι, εκθετικά, κλάσματα κοκ) μέσα σε απλό κείμενο. Μπορείτε να δοκιμάσετε να τρέξετε την παρακάτω εντολή με legend_label='x^2+1', ώστε να δείτε την διαφορά στην τελική μορφοποίηση του κειμένου.

```python
plot( x^2+1, -2, 2,color='red', legend_label='$x^2+1$') + \
plot( 2*x, -2, 2,color='blue', legend_label='$2x$')
```

Θέτοντας στη γραφική παράσταση της εφαπτομένης μικρότερο διάστημα, μπορεί να δημιουργηθεί το επόμενο εφέ. Το μήκος της εφαπτόμενης ευθείας είναι τώρα μικρότερο. Έτσι μπορούμε να προσθέσουμε στην ίδια καμπύλη και άλλες εφαπτόμενες ευθείες σε άλλα σημεία της παραβολής.

```python
plot( x^2+1, -2, 2,color='red', legend_label='$ x^2+1 $') +\
plot( 2*x, 0.5, 1.5,color='blue', legend_label='$ 2x $')
```

## Σημεία

Στο παραπάνω γράφημα, το σημείο $(1, 2)$ μας ενδιαφέρει να τονιστεί, διότι είναι το σημείο στο οποίο εφάπτεται η ευθεία $y = 2x$ στην $f(x) = x^2 + 1$ . Αυτό μπορεί να γίνει με την συνάρτηση `point`. Μπορεί να αλλαχθεί και το μέγεθος του σημείου κάνοντας χρήση της παραμέτρου `size`.

```python
p1 = plot( x^2+1, -2, 2,color='red', legend_label='$x^2+1$') 
p2 = plot( 2*x, 0.5, 1.5,color='blue', legend_label='$2x$')
p3 = point((1,2),size=20, color='green')
p1+p2+p3
```

Ένα σημείο, ή και μια λίστα σημείων, μπορούν να υπάρχουν μόνα τους ανεξάρτητα σε ένα γράφημα.

```python
point((1,3)) + point((2,4)) + point((3,5)) + point((4,6))
```

Σε περίπτωση που τα σημεία είναι αρκετά μπορεί να χρησιμοποιηθεί η δομή της λίστας  (μέσα σε `[` και `]`)  που θα συζητηθεί σε επόμενη ενότητα. Σε αυτή την περίπτωση χρησιμοποιούμε εναλλακτικά την `points` αντί των διαδοχικών `point`.

```python
f(x) = x^2
points([(0,f(0)), (1,f(1)), (2,f(2)), (3,f(3)), (4,f(4))])
```

Ένας κύκλος μπορεί και να είναι γεμάτος, θέτοντας στην `fill` την τιμή True αλλά και να έχει περίγραμμα διαφορετικού χρώματος με την `edgecolor`. ## Βέλη

Με την `arrow` σχεδιάζουμε βέλη στο γράφημά μας. Απαιτούνται οι συντεταγμένες της αρχής και του τέλους του κάθε βέλους. Με την `color` έχουμε την δυνατότητα να προσδιορίσουμε το χρώμα του βέλους.

```python
arrow((0,0), (1,1))+arrow((0.5,0.5), (0,1), color="red")
```

Στην περίπτωση που δεν επιθυμούμε οι άξονες να είναι ορατοί έχουμε την δυνατότητα να θέσουμε την παράμετρο `axes` ίση με `False`. Ακόμη και αν έχουμε διάφορα γραφικά στο γράφημά μας, αρκεί να δηλωθεί σε ένα από αυτά. Παρατηρούμε ακόμη στο επόμενο παράδειγμα ότι όπως και σε οποιαδήποτε γραφική παράσταση μπορεί να τεθεί το πάχος του βέλους με την `thickness`.

```python
g = arrow((0, 0), (1, 1), color="blue",  thickness=2, axes=False)
g += arrow((0.5, 0.5), (0, 1), color="red",  thickness=2.5)
g += arrow((0.2, 0.2), (1, 0.5), color="green",  thickness=3)
g += arrow((0.7, 0.3), (0.2, 0.8), color="purple",  thickness=3.5)
g += arrow((0.9, 0.1), (0.5, 0.9), color="orange",  thickness=4)
g.show()
```

Ο συνδυασμός του `+` με το ίσον δηλ. ο `+=` είναι ένας εύκολος τρόπος για τους προγραμματιστές να προσθέτουν ένα αντικείμενο (π.χ. το γραφικό `arrow((0.5, 0.5), (0, 1), color="red",  thickness=2.5))` σε ένα κάποιο άλλο προϋπάρχων αντικείμενο.

Παράδειγμα:
το `g += arrow((0.5, 0.5), (0, 1), color="red",  thickness=2.5)` σημαίνει με απλά ελληνικά: πρόσθεσε στο g επιπλέον το αντικείμενο που βρίσκεται δεξιά του `+=`. Οπότε το g θα αυξηθεί κατά `arrow((0.5, 0.5), (0, 1), color="red",  thickness=2.5)`.

Υπάρχουν και άλλοι συνδυασμοί όπως ο `*=` αλλά μόνο στις περιπτώσεις αντικειμένων που έχει νόημα ο τελεστής π.χ. αριθμού με αριθμό ή αριθμού με string κοκ. Οι τελεστές αυτοί λέγονται τελεστές **σύντομης ανάθεσης**.

## Κείμενο

Σε ένα γράφημα μπορούμε να συμπεριλάβουμε και ετικέτες μέσω της συνάρτησης `text`. Η πρώτη παράμετρος αφορά το κείμενο το οποίο θέλουμε να συμπεριλάβουμε και η επόμενη τις συντεταγμένες του σημείου στο οποίο αυτό θα εμφανιστεί. Πέρα από τις παραπάνω μπορούμε να προσαρμόσουμε την εμφάνισή του με μία σειρά από επιπλέον παραμέτρους.

- `fontsize`: Το μέγεθος του font του κειμένου. Η τιμή μπορεί να είναι αριθμός ή μία σειρά από λέξεις κλειδιά, `xx-small`, `x-small`, `medium`, `large`, `x-large`, `xx-large`.

- `fontstyle`: Το στυλ της γραμματοσειράς. Η τιμή του μπορεί να είναι `normal`, `italic`, `oblique`

- `fontweight`: Χρησιμοποιείται για να δίνουμε έμφαση στο κείμενο. Μπορούμε να δώσουμε αριθμητική τιμή στο  0-1000 ή μία από τις τιμές `ultralight`, `light`, `normal`, `regular`, `book`, `medium`, `roman`, `semibold`, `demibold`, `demi`, `bold`, `heavy`, `extra bold`, `black`.

- `vertical_alignment` και `horizontal_alignment`: Κατακόρυφη και οριζόντια στοίχιση αντίστοιχα.

- `rotation`: Περιστρέφει το κείμενο

- `alpha`: Καθορίζει το ποσοστό διαφάνειάς του.

Μία ακόμη τεχνική που παρατηρούμε στο επόμενο παράδειγμα είναι η κλήση, αρχικά, της `Graphics` η οποία δημιουργεί μία άδεια λίστα γραφικών αντικειμένων στην οποία μπορούμε να προσθέσουμε κάθε αντικείμενο. (Θα μπορούσαμε βέβαια αντί του κενού γραφήματος να δημιουργήσουμε το πρώτο μας γραφικό αντικείμενο, παράδειγμα ένα plot. Στην συνέχεια θα προσθέταμε σε αυτό και τα υπόλοιπα γραφήματα).

**Παράδειγμα**

Στο παρακάτω παράδειγμα δημιουργούμε το g και με συνεχόμενη χρήση του συνδυασμού `+=` προσθέτουμε στο g όλα τα υπόλοιπα. Με την g ή την `g.show()` βλέπουμε το συνολικό αθροιστικό αποτέλεσμα.

```python
# Δημιουργία κενής γραφικής παράστασης
g = Graphics()

g += line([(1, 0), (1, 2)], color='yellow')
g += line([(-1, 1), (2, 1)], color='magenta')

g += text('Left-aligned', (1, 1.5), fontsize=12, color='blue', \
          horizontal_alignment='left')
g += text('Center-aligned', (1, 1.7), fontsize=12, color='green', \
          horizontal_alignment='center')
g += text('Right-aligned', (1, 1.9), fontsize=12, color='red', \
          horizontal_alignment='right')

g += text('Top-aligned', (-0.5, 1), fontsize=12, color='blue', \
          vertical_alignment='top')
g += text('Center-aligned', (0, 1), fontsize=12, color='green', \
          vertical_alignment='center')
g += text('Bottom-aligned', (0.5, 1), fontsize=12, color='red', \
          vertical_alignment='bottom')

g += text('Rotated Text', (1, 1), fontsize='xx-large', color='blue', \
          rotation=45, alpha=0.5)

g += text('oblique text', (-0.5, 1.9), fontsize=12, color='blue', \
          fontstyle='oblique')
g += text('italic text', (-0.5, 1.7), fontsize=12, color='blue', \
          fontstyle='italic')

g += text('ultralight text', (-0.8, 1.5), fontsize=12, color='blue', \
          fontweight='ultralight')
g += text('medium text', (-0.8, 1.3), fontsize=12, color='blue', \
          fontweight='medium')
g += text('extra bold text', (-0.8, 1.1), fontsize=12, color='blue', \
          fontweight='extra bold')

g.show()
```

Η εντολή `Graphics()` δεν μπορεί να πάρει κάποιο όρισμα! Είναι λάθος δηλαδή να ορίσουμε ένα νέο γραφικό αντικείμενο ως g=Graphics('g'). Το σωστό είναι `g=Graphics()`. Το g είναι φυσικά ένα κενό γράφημα, οπότε αν ζητήσουμε `plot(g)` απλά θα πάρουμε τους δυο κάθετους άξονες. Επίσης το g δεν ανήκει στο Symbolic Ring ή στις συναρτήσεις. Η εντολή `parent(g)` μας αποκαλύπτει ότι ανήκει στην κλάση `Graphics`!

## Ανισότητες

Για να σχεδιαστεί μια ανισότητα στο επίπεδο μπορεί να τονιστεί, με κάποιο χρώμα, είτε τα σημεία που ικανοποιούν την ανισότητα είτε αυτά που δεν την ικανοποιούν.

**Παράδειγμα**

Έστω η ανισότητα $y \geq 1 - x^2$. Θα πρέπει πρώτα να σχεδιαστεί το γράφημα της συνάρτησης $y = 1 - x^2$ και στην συνέχεια να γίνει ο χρωματισμός. Η γραφική παράσταση της $y$ χωρίζει το επίπεδο σε δυο διαφορετικές περιοχές. Η παραπάνω ανισότητα ικανοποιείται ακριβώς από όλα εκείνα τα σημεία που βρίσκονται πάνω από την γραφική παράσταση.

```python
plot( 1-x^2, (x, -2, 2), ymin=-2, ymax=2, \
     gridlines='minor',fill=2, fillcolor='yellow')
```

Η ιδιότητα `fill` καθορίζει ότι το γέμισμα θα γίνει  από τον οριζόντιο άξονα  $y = 2$ και μέχρι την γραφική παράσταση. Μπορεί όμως για παράδειγμα να τεθεί `fill= -2` ή και κάτι μικρότερο ώστε αυτό να γίνει στο κάτω μέρος της γραφικής παράστασης. Τα σημεία που χρωματίζονται παρακάτω ικανοποιούν την (ανάποδη)  ανισότητα $y \leq 1 - x^2$.

```python
plot(1 - x^2, (x, -2, 2), ymin=-2, ymax=2, \
     gridlines='minor', fill=-2, fillcolor='yellow')
```

Αντί για την σταθερά -2 μπορούμε να θέσουμε στην ιδιότητα `fill` άλλη συνάρτηση ώστε να χρωματιστεί  το ενδιάμεσο χωρίο. Το χωρίο αυτό προσδιορίζει όλα τα σημεία $(x,y)$ του επιπέδου που βρίσκονται μέσα  στο ενδιάμεσο χωρίο, δηλαδή για τα οποία ισχύει $g(x)\leq y\leq f(x)$ ή $f(x)\leq y\leq g(x)$.

```python
f(x) = 1 - x^2
g(x) = x^2 - 1
plotf = plot(f(x), (x, -2, 2), ymin=-2, ymax=2, \
             color='blue', legend_label='$f(x)=1-x^2$')
plotg = plot(g(x), (x, -2, 2), color='red', \
             legend_label='$ g(x) = x^2 - 1 $')
fillplot = plot(f(x), (x, -1, 1), ymin=-2, ymax=2, \
                fill=g(x), fillcolor='yellow', \
                gridlines='minor')
plotf + plotg + fillplot
```

## Πολικές συντεταγμένες

Η θέση ενός σημείου στο επίπεδο, πέρα από την x και την y συντεταγμένη, μπορεί να οριστεί με τις πολικές συντεταγμένες του. Δηλαδή με το ζευγάρι (r, θ), όπου με r συμβολίζουμε την απόσταση του σημείου P από την αρχή των αξόνων, σημείο $O$, και με θ την γωνία που πρέπει να στρίψουμε (αριστερόστροφα) τον θετικό οριζόντιο άξονα γύρω από το $O$ για να συναντήσει το P.

![Πολικές συντεταγμένες](polar.png)

Το SageMath παρέχει την μέθοδο `polar_plot` για σχεδιασμό γραφήματος με χρήση πολικών συντεταγμένων. Η γενική μορφή της είναι `polar_plot(f(theta), range_of_theta)`, όπου με `f(theta)` εννοούμε αποκλειστικά μια συνάρτηση της γωνίας (και όχι της ακτίνας f(r) ή της ακτίνας και της γωνίας `f(r,theta))` και με range_of_theta το πεδίο ορισμού της γωνίας (που κανονικά δεν πρέπει να έχει εύρος μεγαλύτερο του 2*π).

**Παράδειγμα**

Στο παράδειγμα που ακολουθεί σχεδιάζεται ένας κύκλος με ακτίνα 1 με τη γωνία θ να είναι από 0 έως 2π. Έτσι σχεδιάζεται ένας ολόκληρος κύκλος. Επίσης, σχεδιάζεται ένα ημικύκλιο ακτίνας 0.5 με το θ από 0 έως π. Το τελευταίο ημικύκλιο, ακτίνας r=0.75, ξεκινά από τις 90 μοίρες, π/2, μέχρι και τις 270, 3π/2.

```python
var('theta')
p1 = polar_plot(1, (theta, 0, 2*pi),\
                color = "purple")
p2 = polar_plot(0.5, (theta, 0, pi),\
                color="red")
p3 = polar_plot(0.75, (theta, pi/2, 3*pi/2),\
                color="brown")
p1+p2+p3
```

Η γραφική παράσταση του ημιτόνου (ή του συνημιτόνου) σε πολικές συντεταγμένες δημιουργούν την λεγόμενη _rose curve_ με διπλάσια πέταλα από το συντελεστή της, θ.

```python
var("theta")
g = Graphics()
g+=polar_plot(cos(2*theta), (theta, 0, 2*pi))
g+=polar_plot(sin(4*theta), (theta, 0, 2*pi),\
              color="red")
g.show(figsize=(5,3))
```

## Πεπλεγμένες Συναρτήσεις

Οι τυπικές μορφές συναρτήσεων, των οποίων τα γραφήματα είδαμε παραπάνω, εκφράζονται ως $y = f(x)$, με την τιμή της y να εξαρτάται άμεσα από τις τιμές του x. Οι πεπλεγμένες συναρτήσεις εκφράζονται συνήθως με εξισώσεις που έχουν την μορφή $g(x,y) = 0$, όπου η τιμή $y$ εξαρτάται έμμεσα από το x.  Όταν έχουμε μια πεπλεγμένη έκφραση, η γραφική της παράσταση μπορεί να αποδοθεί με την `implicit_plot`.

**Παράδειγμα**

Η πεπλεγμένη έκφραση $x^2 + y^2 - 16 = 0$ παριστάνει την περιφέρεια ενός κύκλου με ακτίνα 4.

```python
var('x y')
g(x,y)=x^2+y^2-16
implicit_plot(g, (x,-5,5), (y,-5,5))
```

**Παράδειγμα**

Στο επόμενο παράδειγμα, με την βοήθεια της `implicit_plot` μπορεί κανείς να δει ότι η γραφική παράσταση της αντίστροφης συνάρτησης είναι η αντανάκλαση της γραφικής παράστασης της $𝑓^{-1}$ ως προς την ευθεία $y=x$. Η εκθετική $f(x)=e^x$ είναι γνησίως αύξουσα και συνεχής σε όλο το $\mathbb{R}$. Επομένως είναι 1-1 και αντιστρέψιμη. Η αντίστροφη συνάρτησή της είναι ο φυσικός λογάριθμος $f^{−1}(x)=ln(x)$.

Η επόμενη γραφική παράσταση δείχνει:

- την εκθετική $𝑒^x$

- την αντίστροφη (μέσω της εξίσωσης $𝑒^y=x$

- την ευθεία συμμετρίας $𝑦=𝑥$

- την εξίσωση της $ln(x)$ με διακεκομμένη ώστε να φαίνεται ότι συμπίπτει με την αντίστροφη της $𝑒^x$.

```python
x, y = SR.var('x, y')
f(x)=e^x
eq1 = f(y) == x
g = Graphics()
g += plot(f, (x,-2,5),aspect_ratio=1, \
          ymin = -2, ymax = 5,color='blue')
g += implicit_plot(eq1, (x,-2,5), (y,-2,5),\
                   color='green')
g += plot(log(x), (x,0.001,5), color='orange', \
          linestyle="--")
g += plot(x,(x,-2,5),color='red')
g.show(aspect_ratio=1,axes=True,gridlines="minor")
```

## Σφαιρικές, κυλινδρικές συντεταγμένες

Στο SageMath, οι συναρτήσεις `spherical_plot3d` και `cylindrical_plot3d` χρησιμοποιούνται για τη δημιουργία τρισδιάστατων γραφικών με χρήση σφαιρικών και κυλινδρικών συντεταγμένων αντίστοιχα. Αυτά τα εργαλεία είναι χρήσιμα για την οπτικοποίηση αντικειμένων που ορίζονται σε αυτές τις συντεταγμένες, οι οποίες είναι συχνά πιο κατάλληλες από τις καρτεσιανές συντεταγμένες για συγκεκριμένα σχήματα, όπως σφαίρες ή κυλίνδρους.

### Σφαιρικές συντεταγμένες

Σε κάθε σημείο  του χώρου αντιστοιχούν οι **σφαιρικές συντεταγμένες** (spherical coordinates) του:

- **r** ακτινική απόσταση  (η απόσταση του σημείου από την αρχή των αξόνων O)

- **θ** η πολική γωνία θ (η θετική γωνία που πρέπει να στρίψουμε τον θετικό άξονα $Ox$ έως ότου συναντήσει την προβολή του σημείου στο $Oxy$ επίπεδο)

- **φ** το αζιμούθιο φ (η θετική γωνία που πρέπει να στρίψουμε τον θετικό άξονα $Oz$ έως ότου συναντήσει το σημείο).

![Σφαιρικές συντεταγμένες](spherical.png)

Η γενική μορφή της είναι `spherical_plot3d(f(θ, φ), range_of_θ, range_of_φ)`, όπου με $f(θ, φ)$ εννοούμε μια συνάρτηση του $θ$ και του $φ$  που δίνει  την ακτινική απόσταση $r$. Το  range_of_θ είναι  το πεδίο ορισμού της μεταβλητής $θ$ (που κανονικά δεν πρέπει να έχει εύρος μεγαλύτερο του $2*π$) και το range_of_φ  το πεδίο ορισμού της μεταβλητής $φ$ (που κανονικά  έχει εύρος από $0$ έως $π$).

Το παράδειγμα που ακολουθεί σχεδιάσαμε μια επιφάνεια στις σφαιρικές συντεταγμένες, με  $r=1$. Θέσαμε δηλαδή $f(θ,φ)=1$. Προφανώς πρόκειται για την επιφάνεια μια σφαίρας ακτινας $r=1$.

```python
var('theta phi')
spherical_plot3d(1, (theta, 0, 2*pi), (phi, 0, pi))
```

### Κυλινδρικές συντεταγμένες

Οι **κυλινδρικές συντεταγμένες** (cylindrical coordinates) αποτελούν μια επέκταση των πολικών συντεταγμένων σε τρεις διαστάσεις. Χρησιμοποιούνται για να περιγράψουν σημεία στο τρισδιάστατο χώρο με τις εξής παραμέτρους:

- **r** η  (οριζόντια) απόσταση του σημείου από τον κάθετο άξονα $Oz$.

- **θ** η γωνία στο επίπεδο Oxy (η θετική γωνία που πρέπει να στρίψουμε τον θετικό άξονα Ox έως ότου συναντήσει την προβολή του σημείου στο Oxy επίπεδο).

- **z** η τρίτη συντεταγμένη του σημείου $(x,y,z)$ (το πόσο πάνω η κάτω απέχει από το Oxy επίπεδο).

![Κυλινδρικές συντεταγμένες](cylindrical.jpg)

Η γενική μορφή της είναι `cylindrical_plot3d(f(θ,z), range_of_theta, range_of_z)`, όπου με f(θ, z) εννοούμε μια συνάρτηση του θ και του z (ύψους) που δίνει την απόσταση r. Το range_of_theta είναι το πεδίο ορισμού της μεταβλητής θ (που κανονικά δεν πρέπει να έχει εύρος μεγαλύτερο του 2∗π). Το range_of_z καθορίζει τις επιθυμητές διαστάσεις του ύψους και του βάθους z της κυλινδρικής μας επιφάνειας με εξίσωση r = f(θ, z). 

**Παράδειγμα**

Στο παρακάτω παράδειγμα σχεδιάζουμε μια κυλινδρική επιφάνεια που απέχει σταθερή απόσταση r = 1 από τον Oz. Για το λόγο αυτό θέσαμε $f(θ, z) = 1$. Προκύπτει ένας κύλινδρος ακτίνας $r = 1$ και ύψους 2.

```python
var('z theta')
cylindrical_plot3d(1, (theta, 0, 2*pi),(z, 0, 2))
```

# Τρισδιάστατα γραφικά στο επίπεδο

Μια συνάρτηση δύο μεταβλητών μπορεί να οπτικοποιηθεί και μέσω γραφικών στις δύο διαστάσεις. Κάτι τετοιο μπορεί να γίνει: 

- Διάγραμμα περιοχής (**region plot**): Παρουσιάζει την περιοχή του επιπέδου όπου ισχύει κάποια ανισότητα που εμπλέκει τη συνάρτηση, π.χ. f(x,y)>0 ή f(x,y)<k. Είναι ιδιαίτερα χρήσιμο για τη μελέτη περιοχών ορισμού ή συνόλων λύσεων.

- Διάγραμμα ισοϋψών (**contour plot**): Εμφανίζει καμπύλες που συνδέουν σημεία του επιπέδου (x,y) για τα οποία η συνάρτηση παίρνει την ίδια τιμή f(x,y)=c. Κάθε καμπύλη αντιστοιχεί σε ένα σύνολο σταθερής τιμής (level set) και παριστάνει μια «τομή» της επιφάνειας z=f(x,y) με το επίπεδο z=c.


- Διάγραμμα πυκνότητας (**density plot**): Χρησιμοποιεί χρωματική διαβάθμιση για να αποδώσει την τιμή της συνάρτησης f(x,y) σε κάθε σημείο του επιπέδου. Οι περιοχές με παρόμοιες τιμές χρωματίζονται με παρόμοιο τόνο, επιτρέποντας μια διαισθητική απεικόνιση της μορφής της συνάρτησης.

## Region plots

Για να αποδώσουμε μια μη γραμμική ανισότητα με δυο μεταβλητές μπορούμε να χρησιμοποιήσουμε την `region_plot` όπως στο επόμενο παράδειγμα. Με μπλε χρώμα εμφανίζονται τα σημεία του επιπέδου που ισχύει η ανισότητα $cos(x^2 + y^2) \leq 0$ και με λευκό εκείνα που δεν ισχύει.

```python
var("y")
region_plot(cos(x^2+y^2) <= 0, (x, -2*pi, 2*pi), (y, -2*pi, 2*pi) )
```

Αν έχουμε πάνω από μια ανισώσεις ή εξισώσεις και θέλουμε να σχεδιάσουμε τα σημεία που αυτές ικανοποιούνται συγχρόνως, τότε μπορούμε να τις βάλουμε σε μια λίστα π.χ. [cos(x 2 +y 2 ) <= 0, x+y ≥ 2, abs(x−y) >= 1] και στην συνέχεια να την περάσουμε στην `region_plot`. Υπάρχει και η δυνατότητα να τις συνδυάσουμε με οποιοσδήποτε λογικούς συνδέσμους, όπως για παράδειγμα τους `not` (δεν ισχύει), `and` (και, συγχρόνως) και `or` (ισχύει μια τουλάχιστον απ’ αυτές). Όμως τότε απαιτείται η χρήση του τελεστή `lambda`. Θα μελετήσουμε όμως αυτές τις δυνατότητες σε επόμενες ενότητες. Παρακάτω στην ενότητα με την density plot θα δούμε ένα τέτοιο παράδειγμα με την `lambda`.

## Διάγραμμα ισοϋψών καμπυλών, contour plots

Ένα διάγραμμα ισοϋψών καμπυλών αποτελεί μια γραφική αναπαράσταση που δείχνει τα διαφορετικά  επίπεδα τιμών μιας συνάρτησης $f$  δύο μεταβλητών χρησιμοποιώντας κάποιες χρωματικές αποχρώσεις, συνήθως του γκρίζου χρώματος. Δημιουργείται με την `contour_plot`. Οι πιο μαύροι κύκλοι αντιστοιχούν σε μικρές τιμές της $f$ ενώ οι πιο φωτεινοί σε μεγαλύτερες.

```python
var('x y')
f(x,y) = x^2 + y^2
contour_plot(f, (x, -2, 2), (y, -2, 2))
```

Μπορούμε να εμφανίσουμε και τις τιμές της f χρησιμοποιώντας την παράμετρο `labels`.

```python
contour_plot(f, (x, -2, 2), (y, -2, 2),labels=True)
```

Έχουμε την δυνατότητα να θέσουμε τον αριθμό των ισοϋψών με την παράμετρο `contours`. Η παράμετρος `axes` εμφανίζει τους άξονες αν τεθεί `True`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), \
             contours=5, axes=True, labels=True)
```

**Παράδειγμα**

Στην συνέχεια, επιλέξαμε να σχεδιαστούν οι ισοϋψείς καμπύλες που αντιστοιχούν σε τρεις μόνο τιμές. Σε αυτές τις καμπύλες η συνάρτησή μας έχει πάρει τιμές -0.1, 0.1 και 0.5 αντίστοιχα. Για να το πετύχουμε αυτό θέσαμε `contours=(-0.1,0.1,0.5)`. Η `fill=False` μας αφαίρεσε τα γκρίζα, ώστε να αναδειχθούν οι ισοϋψείς.

```python
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), \
             contours=(-0.1,0.1,0.5), fill=False)
```

Ο χρωματικός χάρτης, `cmap`, στα παραπάνω παραδείγματα είναι ο εξορισμού `grey`. Μπορεί να αλλάξει σε διάφορες τιμές όπως `plasma`, `hot`, `jet`, `coolwarm` ή ακόμη και σε φιλική στους έχοντες αχρωματοψία `cividis`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), \
             contours=20, axes=True, cmap = "jet")
```

Χρήσιμο είναι να φαίνεται στο γράφημα τί τιμές αντιπροσωπεύει κάθε χρώμα. Αυτό επιτυγχάνεται θέτοντας `True` στην ιδιότητα `colorbar`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), \
             contours=20, axes=True, cmap = "jet", colorbar=True)
```

##  density plot

Η `density_plot` χρησιμοποιείται για να δείξει πώς οι τιμές μιας συνάρτησης δύο μεταβλητών μεταβάλλονται σε μια συγκεκριμένη περιοχή, με διαφορετικά χρώματα να αντιστοιχούν σε διαφορετικές τιμές. Μοιάζει με την `contour_plot`, χωρίς να έχει βέβαια contours ή colorbar. Επαναλαμβάνουμε το προηγούμενο παράδειγμα με την `density_plot` για να δούμε την διαφορά.

```python
var('x y')
f(x, y) = cos(x) + sin(y)
density_plot(f(x,y), (x,-pi,pi), (y,-pi,pi), \
             axes=True, cmap='jet')
```

Φυσικά, τα `density_plot` και `contour_plot` μπορούν να συνδυαστούν για να πετύχουμε περισσότερες οπτικές πληροφορίες. Όμοια μπορούμε να συνδυάσουμε τα `region_plot` και `contour_plot`.

```python
f(x,y) = cos(x) + sin(y)
p1=contour_plot(f(x,y), (x,-pi,pi), (y,-pi,pi), \
                axes=True, fill=false)
p2=density_plot(f(x,y), (x,-pi,pi), (y,-pi,pi), \
                axes=True, cmap='jet')
p1+p2
```

Στο παρακάτω παράδειγμα σχεδιάζουμε (με μπλε χρώμα) τα σημεία (x, y) του επιπέδου που ικανοποιούν τις σχέσεις $-1 \leq f(x,y) \leq 0$. Στην συνέχεια προσθέτουμε την `region_plot` πάνω από την `contour_plot`.

```python
p3=region_plot(lambda x,y: f(x,y) <= 0 and -1<=f(x,y), \
               (x, -pi, pi), (y,-pi, pi),plot_points=200)
p4=contour_plot(f(x,y), (x,-pi,pi), (y,-pi,pi), axes=True)
# Η p3 σχεδιάζεται πάνω από την p4.
# Το αντίθετο p3+p4 δεν βοηθάει στην κατανόηση των ανισοτήτων.
p4+p3
```
