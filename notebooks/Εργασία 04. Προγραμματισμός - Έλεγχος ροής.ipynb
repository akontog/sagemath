# Άσκηση, Αναδρομή

Μια αλγοριθμική μέθοδο για τον υπολογισμό της δύναμης ενός αριθμού σε λογαριθμικό χρόνο αποτελεί η ύψωση σε δύναμη δια του τετραγωνισμού. Είναι αναδρομική διαδικασία η οποία έχει ως εξής:
$$
a^n =
\begin{cases}
1 & \text{αν } n = 0 \\
\left(a^{n/2}\right)^2 & \text{αν } n \text{ είναι άρτιος} \\
a \times a^{n-1} & \text{αν } n \text{ είναι περιττός}
\end{cases}
$$

Υλοποιήστε συνάρτηση, έστω fastExponentiation, που θα εφαρμόζει τον παραπάνω αναδρομικό τύπο.
