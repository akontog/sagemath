## Ταυτολογίες και αντιφάσεις

Δείξτε ποιες από τις επόμενες προτάσεις αποτελούν ταυτολογίες, ποιες αντιφάσεις και ποιες απλά ικανοποιούνται.

1. $p \lor \neg p$

2. $p \land \neg p$

3. $\neg (p \land q) \leftrightarrow (\neg p \lor \neg q)$

4. $(p \lor q) \land \neg p \land \neg q$

5. $p \land q) \rightarrow p$

6. $q \rightarrow (p \lor q)$

7. $(p \land q) \lor (\neg p \land q)$

8. $(p \lor q) \leftrightarrow (q \lor p)$

````python
statements = [
    '(p | ~p)',                 # 1. p ∨ ¬p
    '(p & ~p)',                 # 2. p ∧ ¬p
    '(~(p & q)<->(~p | ~q))', # 3. ¬(p ∧ q) ↔ (¬p ∨ ¬q)
    '((p | q) & ~p & ~q)',      # 4. (p ∨ q) ∧ ¬p ∧ ¬q
    '((p & q) -> p)',           # 5. (p ∧ q) → p
    '(q -> (p | q))',           # 6. q → (p ∨ q)
    '((p & q) | (~p & q))',     # 7. (p ∧ q) ∨ (¬p ∧ q)
    '((p | q) <-> (q | p))'     # 8. (p ∨ q) ↔ (q ∨ p)
]
for i, stmt in enumerate(statements, 1):
    print('Πρόταση'+str(i)+': '+stmt)
    expr = propcalc.formula(stmt)
    print("Ταυτολογία:", bool(expr.is_tautology()))
    print("Αντίφαση:", bool(expr.is_contradiction()))
    print("Ικανοποιείται:", bool(expr.is_satisfiable()))
    print("-" * 30)
````

Δίνονται οι επόμενες λογικές προτάσεις. Βρείτε ποιες είναι λογικά ισοδύναμες μεταξύ τους.

1. $(p \land q) \lor (\neg p \land q)$

2. $(p \land r) \lor q$

3. $q \rightarrow p$

4. $q \lor \neg p$

5. $p \rightarrow q$

6. $p \lor (r \lor q)$

````python
A = Set([1, 2, 3, 4])
B = Set([5, 6, 7])
C = Set([4, 5, 6, 7])
# Υπολογισμός τομής
intrsct = A.intersection(C)
# Έλεγχος αν είναι ξένα σύνολα
disjoint = intrsct.is_empty()
if disjoint:
    print("Τα σύνολα είναι ξένα μεταξύ τους.")
else:
    print("Τα σύνολα δεν είναι ξένα. Η τομή τους είναι:", intrsct)
````

4. Για να έχει γίνει flush θα πρέπει το μήκος του set με το χρώμα του χεριού, {suit for (_, suit) in hand},  να είναι 1.

````python
Suits   = Set(["Hearts", "Diamonds", "Spades", "Clubs"])
Values  = Set([2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King", "Ace"])
Cards   = cartesian_product([Values, Suits])
Hands   = Subsets(Cards, 5)
# Θεωρητική πιθανότητα flush
total_hands = binomial(52, 5)
flush_count = 4 * binomial(13, 5)
theoretical_prob = n(flush_count / total_hands)
@interact
def simulate_flushes(k=slider(100, 5000, default=1000, label="Αριθμός μοιρασμάτων (k)")):
    k = int(k)
    cumulative_flushes = 0
    experimental_probs = []
    for i in range(1, k+1):
        hand = Hands.random_element()
        if len({suit for (_, suit) in hand}) == 1:
            cumulative_flushes += 1
        experimental_probs.append(n(cumulative_flushes / i))
    g = Graphics()
    # Πειραματικές πιθανότητες
    
    g += line([(i+1, experimental_probs[i]) 
               for i in range(len(experimental_probs))], 
              color='blue', thickness=1.5, 
              legend_label='Πειραματική πιθανότητα')
    # Θεωρητική πιθανότητα
    g += line([(1, theoretical_prob), (k, theoretical_prob)], 
              color='red', linestyle='--', thickness=2, 
              legend_label=f'Θεωρητική: {theoretical_prob:.6f}')
    g.show(figsize=(4,3))
````
