# Άσκηση 4

Έστω η συνάρτηση
$$
f(x)=x^2-4x+3
$$
Σχεδιάστε:

- την καμπύλη της συνάρτησης στο (-1,5)

- το σημείο (1,f(1))

- κείμενο που να δείχνει το σημείο.

- την εφαπτομένη της στο σημείο.

Χρησιμοποιήστε gridlines minor και διαφορετικά χρώματα.


Σημειώσεις:
Η εφαπτομένη της στο (1,f(1)) δίνεται από τον τύπο
$$
g(x)=-2(x-1)
$$

## Λύση

```python
# Ορισμός της συνάρτησης
f(x) = x^2 - 4*x + 3

# Ορισμός του σημείου
p = (1, f(1))

# Υπολογισμός της εφαπτομένης
# slope = h.diff(x)(1)  # Παράγωγος στο x=1
# g(x) = slope*(x - 1) + h(1)  # Εφαπτομένη
g(x) = -2*(x-1)
# Δημιουργία κειμένου για το σημείο
pt = text('(1, ' + str(f(1)) + ')', (1.2, f(1)-0.5), fontsize=12)

# Σχεδίαση της καμπύλης της συνάρτησης
fp = plot(f, (x, -1, 5), gridlines='minor', color='blue')

# Σχεδίαση του σημείου
pp = point(p, size=20, color='green')

# Σχεδίαση της εφαπτομένης
tp = plot(g, (x, 0.5, 1.5), color='red')

# Συνδυασμός όλων των στοιχείων
pp + fp + tp + pt
```

# Άσκηση 5

Σχεδιάστε ένα γεωμετρικό σχήμα που να περιέχει:

- Έναν κύκλο με ακτίνα 4 και κέντρο την αρχή των αξόνων.

- Κείμενο στα σημεία A = (0, 0), B = (4, 0), C = (0, 4), D = (-4, 0).
- Ένα ορθογώνιο με κορυφές: (4,0), (4,3), (0,3), (0,0).
- Έναν γεμισμένο δίσκο με κέντρο την αρχή των αξόνων και ακτίνα 0.6.
- Δύο βέλη που δείχνουν από το κέντρο του κύκλου προς το σημείο (4,0) και προς το (2,3).
- Πρόσθεσε υπόμνημα που να δείχνει κατάλληλο κείμενο για κάθε σχήμα, όπως Ορθογώνιο, Βέλος 1, Βέλος 2.
- Οι άξονες να μην φαίνονται.

## Λύση

```python
# Ορισμός των σημείων
A = (0, 0)
B = (4, 0)
C = (0, 4)
D = (-4, 0)

# Σχεδίαση κύκλου με ακτίνα 4
circle_plot = circle(A, 4, color='blue', legend_label='Circle, radius=4', axes=False)

# Σχεδίαση γεμισμένου δίσκου με ακτίνα 0.6
disc_plot = disk((0, 0), 0.6,(pi, 3*pi/2), color='orange', legend_label='Disc radius=0.6')

# Σχεδίαση ορθογωνίου
rectangle_plot = polygon([(4, 0), (4, 3), (0, 3), (0, 0)], color='lightgreen', legend_label='Rectangle')

# Προσθήκη βελών
arrow_1 = arrow((0, 0), (4, 0), color='red', legend_label='Βέλος 1')
arrow_2 = arrow((0, 0), (2, 3), color='red', legend_label='Βέλος 2')

# Προσθήκη ετικετών
label_A = text('A (0, 0)', A, fontsize=12)
label_B = text('B (4, 0)', B, fontsize=12)
label_C = text('C (0, 4)', C, fontsize=12)
label_D = text('D (-4, 0)', D, fontsize=12)

# Συνδυασμός όλων των στοιχείων
circle_plot + disc_plot + rectangle_plot + arrow_1 + arrow_2 + label_A + label_B + label_C + label_D
```

# Άσκηση 6

Συνδυάστε κατάλληλα τις density_plot και contour_plot για να αποδώσετε τους "κυκλικούς παλμούς" που προκύπτουν από την συνάρτηση:
$$
f(x, y) = sin(x^2 + y^2)
$$

## Λύση

```python
# Ορισμός μεταβλητών
var('x y')

# Ορισμός συνάρτησης
f(x, y) = sin(x^2 + y^2)

# Δημιουργία density plot
density = density_plot(f(x, y), (x, -3, 3), (y, -3, 3), cmap='plasma')

# Δημιουργία contour plot
contour = contour_plot(f(x, y), (x, -3, 3), (y, -3, 3), contours=15, fill = False)

# Συνδυασμός density και contour plot
density + contour
```
