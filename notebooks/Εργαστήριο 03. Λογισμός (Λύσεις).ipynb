# Άσκηση, δίκλαδη συνάρτηση

Σχεδιάστε την γραφική παράσταση της επόμενης συνάρτησης ορίζοντάς την ως δίκλαδη με την `piecewise`.
$$
f(x) =
\begin{cases} 
x^2 + x, & \text{αν } x < 1 \\

\end{cases}
$$

# Άσκηση, Κοινά σημεία συναρτήσεων

$$
f(x) = x^2 - 4
$$
και η συνάρτηση
$$
g(x) = -x^2 + 2*x + 2

# Άσκηση Μέση Τιμή

Ποια η μέση τιμή της επόμενης συνάρτησης στο [-2,2];
$$
f(x) = \sqrt{4 - x^2}
$$
Δίνεται ότι αν f ολοκληρώσιμη στο [α, β] η μέση τιμή της στο διάστημα αυτό δίνεται από τον τύπο:
$$
\frac{1}{\beta - \alpha} \int_{\alpha}^{\beta} f(x) \, dx
$$
Σχεδιάστε γραφική παράσταση στην οποία θα φαίνεται και η μέση τιμή.

## Λύση

```python
f(x) = sqrt(4-x^2)
p = plot(f, (x, -2, 2), fill=True, fillcolor='lightblue',  ymin=0, aspect_ratio=1,figsize=(4,3))
p
```

```python
a = -2
b = 2
i = integral(f, x, a, b)
print('Ολοκλήρωμα:',i)
meanvalue = (1 / (b - a)) * i
print('Μέση τιμή:',meanvalue)
meanline = line([(a, meanvalue), (b, meanvalue)], color='red')
p+meanline
```

# Άσκηση, Εμβαδόν μεταξύ καμπύλης και άξονα

Δίνεται η επόμενη καμπύλη.
$$
f(x)=x^3-x^2-2x
$$
Βρείτε το εμβαδόν της καμπύλης που περικλείεται μεταξύ της καμπύλης και του άξονα xx' στο διάστημα [-1.5, 2.5]. Θα πρέπει να:

1. βρείτε τα σημεία στα οποία f(x)=0

2. σχεδιάσετε την γραφική παράσταση και δείτε σε ποια διαστήματα η συνάρτηση είναι θετική ή αρνητική

4. υπολογίστε τα επιμέρους ολοκληρώματα ώστε να βρείτε το συνολικό εμβαδόν.

## Λύση

```python
# 1. σημεία στα οποία f(x)=0
f(x) = x^3 - x^2 - 2*x
roots = solve(f(x) == 0, x)
roots
```

```python
x0 = roots[0].rhs()
x1 = roots[1].rhs()
x2 = roots[2].rhs()
fpoints = [(x0,f(x0)),(x1,f(x1)),(x2,f(x2))]
fpoints
```

```python
 # 2. γραφική παράσταση 
fp = plot(f, (x, -2, 3),ymin=-3,ymax=3, fill=0, fillcolor="lightblue", legend_label='$ f(x) = x^3 - x^2 - 2x $', aspect_ratio=1)
pp = points(fpoints, size=30,color='red')
fp+pp
```

```python
# Ολοκλήρωμα στο διάστημα [-1.5, -1]
integral1 = abs(integral(f, x, -1.5, -1)) 
# Ολοκλήρωμα στο διάστημα [-1, 0]
integral2 = integral(f, x, -1, 0)
# Ολοκλήρωμα στο διάστημα [0, 2]
integral3 = abs(integral(f, x, 0, 2))
# Ολοκλήρωμα στο διάστημα [2, 2.5]
integral4 = abs(integral(f, x, 2, 2.5))
print('Συνολικό εμβαδόν:',integral1+integral2+integral3+integral4)
```

# Άσκηση, Εμβαδόν μεταξύ καμπυλών

Θέλουμε να καθορίσουμε την περιοχή που βρίσκεται μεταξύ δύο καμπυλών
$$
f(x) = \frac{x^2}{2} - x + 1, g(x) = -\frac{x^2}{2} + 3x - 2
$$

1. Θα σχεδιάσουμε την γραφική τους παράσταση

2. Θα βρούμε τα κοινά τους σημεία (μπορούμε να σχεδιάσουμε και πάλι την παράσταση με τα σημεία)

3. Θα βρούμε την απόλυτη τιμή του ορισμένου ολοκληρώματος της διαφοράς τους ανάμεσα στα κοινά τους σημεία.

## Λύση

```python
# 1. γραφική παράσταση
f(x) = x^2/2-x+1
g(x) = -x^2/2+3*x-2
fgp = plot([f,g],(x,-1,4),legend_label=[r'$ \frac{x^2}{2} - x + 1 $',r'$ -\frac{x^2}{2} + 3x - 2 $'],figsize=(4,3))
fgp
```

```python
# 2. Κοινά σημεία
common = solve(f == g, x);
print(common)
```

```python
# γραφική παράσταση (με σημεία και σκιασμένο χωρίο)
x1 = common[0].rhs()
x0 = common[1].rhs()
p0 = (x0,f(x0))
p1 = (x1,f(x1))

gp = plot(g,(x,-1,4),legend_label=r'$ g(x) = -\frac{x^2}{2} + 3x - 2 $', color='green',figsize=(4,3))
fp = plot(f,(x, -1, 4), legend_label=r'$ f(x) = \frac{x^2}{2} - x + 1 $')
fillp = plot(g,(x,x0,x1),fill = f, color='yellow')
p = points([p0,p1],size=30,color='red')
fp +gp + p + fillp
```

```python
# 3. εύρεση περιοχής
integral(g(x) - f(x), x, x0, x1)
```

# Άσκηση, Διαφορικές Εξισώσεις

Έστω η επόμενη διαφορική εξίσωση:
$$
\frac{dy}{dx} + 2y = e^{3x}
$$

1. Λύστε την εξίσωση χωρίς αρχικές συνθήκες.

2. Λύστε για αρχική συνθήκη y(0) == 1

## Λύση

```python
# Ορισμός των μεταβλητών
x = var('x')
y = function('y')(x)
# Ορισμός της διαφορικής εξίσωσης
de = y.diff(x) + 2*y - exp(3*x)
```

```python
# 1. χωρίς αρχικές συνθήκες
# Λύση της διαφορικής εξίσωσης
solution = desolve(de, y)
# Εμφάνιση της γενικής λύσης
show(solution)
```

```python
# 2. αρχική συνθήκη y(0) == 1
solution = desolve(de, y, ics=[0,1])
show(solution)
```
