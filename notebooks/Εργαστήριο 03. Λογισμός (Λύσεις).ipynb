# Άσκηση, Εμβαδόν μεταξύ καμπυλών

Θέλουμε να καθορίσουμε την περιοχή που βρίσκεται μεταξύ δύο καμπυλών
$$
f(x) = \frac{x^2}{2} - x + 1, g(x) = -\frac{x^2}{2} + 3x - 2
$$

1. Θα σχεδιάσουμε την γραφική τους παράσταση

2. Θα βρούμε τα κοινά τους σημεία (μπορούμε να σχεδιάσουμε και πάλι την παράσταση με τα σημεία)

3. Θα βρούμε την απόλυτη τιμή του ορισμένου ολοκληρώματος της διαφοράς τους ανάμεσα στα κοινά τους σημεία.

```python
# 1. γραφική παράσταση
f(x) = x^2/2-x+1
g(x) = -x^2/2+3*x-2
fgp = plot([f,g],(x,-1,4),legend_label=[r'$ \frac{x^2}{2} - x + 1 $',r'$ -\frac{x^2}{2} + 3x - 2 $'])
fgp
```

```python
# 2. Κοινά σημεία
common = solve(f == g, x);
common
x1 = common[0].rhs()
x2 = common[1].rhs()
p1 = (x1,f(x1))
p2 = (x2,f(x2))

gp = plot(g,(x,-1,4),legend_label=r'$ -\frac{x^2}{2} + 3x - 2 $')
fp = plot(f,(x, x1, x2),fill=g,legend_label=r'$ \frac{x^2}{2} - x + 1 $')
p = points([p1,p2],size=30,color='red')
finalp = fp +gp + p
show(finalp)
```

```python
fgp = plot(f, (x, x1-1, x2+1), color='blue', legend_label='$f(x)$', fill=g, fillcolor='lightgray') + \
      plot(g, (x, x1-1, x2+1), color='green', legend_label='$g(x)$')
fgp
```
