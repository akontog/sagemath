# Αναδρομή

_(Μονάδες 2)_ Στα μαθηματικά, συνδυασμός των n στοιχείων ενός συνόλου A ανά k ονομάζεται κάθε υποσύνολο του συνόλου A με k στοιχεία.
Το πλήθος των συνδυασμών δίνονται από τον διωνυμικό συντελεστή:
$$
\binom{n}{k} = \frac{n!}{k! \cdot (n-k)!}
$$
Στο SageMath η μέθοδος `binomial(n,k)` επιστρέφει τον διωνυμικό συντελεστή.

Ο υπολογισμός μπορεί να γίνει και με αναδρομικό τρόπο βασισμένο στον επόμενο ορισμό:
$$
\binom{n}{k} = 
\begin{cases} 
1, & \text{αν } k = 0 \text{ ή } k = n \\ 
\binom{n-1}{k-1} + \binom{n-1}{k}, & \text{αν } 0 < k < n 
\end{cases}
$$

- Εφόσον υπάρχει μόνο ένας τρόπος να επιλέξεις κανένα στοιχείο k=0 ή όλα τα στοιχεία k=n έχουμε: $\binom{n}{k} = 1$

- Ο αριθμός των τρόπων για να επιλέξουμε k στοιχεία από n μπορεί να υπολογιστεί ως το άθροισμα $\binom{n-1}{k-1} + \binom{n-1}{k}$ δηλαδή το άθροισμα των επιλογών όταν το πρώτο στοιχείο περιλαμβάνεται στην επιλογή και όταν αυτό δεν περιλαμβάνεται.

α) Υλοποιήστε αναδρομική συνάρτηση Python, _combinations(n,k)_, που να υπολογίζει τον διωνυμικό συντελεστή αναδρομικά και καλέστε κατάλληλα την δική σας υλοποίηση και την `binomial(n,k)` του SageMath για να επαληθεύσετε το αποτέλεσμά σας.

β) Κάνοντας χρήση των διωνυμικών συντελεστών μπορούμε να κατασκευάσουμε το τρίγωνο του Pascal. Καλέστε κατάλληλα την _combinations(n,k)_ για k από 0 έως n όπου το n θα αλλάζει από 0 έως ένα ορισμένο ύψος για κατασκευάστε ένα τέτοιο τρίγωνο. Δίνεται παράδειγμα για ύψος τριγώνου 6.
