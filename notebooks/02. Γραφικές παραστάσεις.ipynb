# Δισδιάστατα γραφικά

## plot

Η συνάρτηση `plot` επιτρέπει την σχεδίαση της γραφικής παράστασης συνάρτησης.

```python
plot(x^2)
```

## Εμφάνιση

Διάφορες επιλογές εμφάνισης μπορούν να τεθούν στις γραφικές μας παραστάσεις. Η `alpha`, από 0 έως την εξορισμού τιμή 1 θέτει την διαφάνεια από εντελώς διάφανη έως αδιαφανή. Η `thickness` με εξορισμού τιμή το 1 θέτει το πάχος της γραμμής. 

```python
plot(x *sin(1/x), thickness=4, alpha=0.5)
```

Η `linestyle` θέτει το στυλ της γραμμής στο εξορισμού συμπαγές `-`, σε διακεκομμένο `--`, διάστικτο `:`, παύλα και τελεία `-.`.

```python
plot(sin(pi*x),linestyle=':')
```

## Διάστημα

Το διάστημα εξ'ορισμού για τις τιμές του οριζόντιου άξονα που θα συμπεριληφθούν στην γραφική παράσταση είναι το [-1,1],  αλλά μπορεί να τεθεί και διαφορετικό ως παράμετρος στην `plot`. Στο επόμενο γράφημα περιλαμβάνουμε όλες τις τιμές από -pi (περίπου -3.141) έως pi.

```python
plot(sin(2*x), -pi, pi)
```

Υπάρχει και η δυνατότητα αλλαγής διαστήματος στον `y` άξονα μέσω των παραμέτρων `ymin`, `ymax`.

```python
plot( x^3-x, -3, 3, ymin = -6, ymax = 6)
```

## Αναλογία Διαστάσεων

Είναι πιθανό το SageMath να σχεδιάσει μία συνάρτηση με διαφορετικές κλίμακες στους άξονες. Για παράδειγμα αν ζητήσουμε το πάνω ημικύκλιο, δηλαδή
$$
y = \sqrt{1 - z^2}
$$
αυτό θα εμφανιστεί παραμορφωμένο. Υπενθυμίζεται ότι πέρα από την x, θα πρέπει όλες οι συμβολικές μεταβλητές να δηλώνονται πριν την χρήση τους όπως φαίνεται και στην επόμενη υλοποίηση.

```python
var('z')
y=sqrt(1-z^2)
plot(y)
```

Το `aspect_ratio` καθορίζει την αναλογία της μονάδας μήκους του άξονα Oy ώς προς την αντίστοιχη του Ox. Έτσι, θέτοντας την παράμετρο `aspect_ratio` ίση με 1 διατηρείται η σωστή γεωμετρία του ημικυκλίου και αποδίδεται το σχήμα με τις σωστές αναλογίες. 

```python
plot(y,aspect_ratio=1)
```

Φυσικά, η παράμετρος `aspect_ratio` μπορεί να πάρει και άλλες τιμές ανάλογα με το επιθυμητό αποτέλεσμα. Στο επόμενο παράδειγμα οι μονάδες διπλασιάζονται στον y σε σχέση με τον x άξονα.

```python
plot(y,aspect_ratio=2)
```

## Υπέρθεση γραφημάτων

Δύο συναρτήσεις μπορούν να απεικονιστούν στην ίδια γραφική παράσταση κάνοντας χρήση των `[ ]` και παραθέτοντας τες χωρισμένες με `,` όπως βλέπουμε στο επόμενο παράδειγμα.

```python
plot([sin(x),cos(x)], 0, 3*pi) 
```

Μπορούμε να έχουμε και παραπάνω από δύο συναρτήσεις, στις οποίες δίνεται εξορισμού κάποιο ξεχωριστό χρώμα.

```python
plot([sin(x),cos(x),sin(2*x),cos(2*x)], 0, 3*pi)
```

## Χρώμα

Το χρώμα μπορεί να τεθεί και από τον χρήστη με την παράμετρο `color`. Η τιμή της παραμέτρου μπορεί να τεθεί με έναν από τους επόμενους τρόπους:

- ένα λεκτικό, `red`, `green`, `blue`, `black` κ.ο.κ.

- μία τριάδα των βασικών χρωμάτων (r, g, b) από 0 έως 1. 

- ένας δεκαεξαδικός κωδικός 6 ψηφίων με κάθε δυάδα να αναφέρεται και πάλι στα τρία βασικά χρώματα.

Στην πραγματικότητα ο τρόπος αναπαράστασης χρωμάτων στον SageMath χρησιμοποιεί True Color δηλαδή 24 bits (binary digit 0 ή 1) πληροφορίας. Τα 24 bits χωρίζονται σε 3 ομάδες των 8 bits (ή 1 byte) για καθένα από τα τρία βασικά χρώματα: 

- κόκκινο (red)

- πράσινο (green)

- μπλε (blue).

Με αυτόν τον τρόπο, όμως, κάθε βασικό χρώμα μπορεί να πάρει τιμές από 0 (=00000000) έως 255 (=11111111) ή στο δεκαεξαδικό σύστημα 00 έως ff. Στο σύνολο μπορούμε να έχουμε 2^24 ή 256x256x256 = 16.777.216 συνδιασμούς και άρα διαφορετικά χρώματα.

Παραδείγματα:
- Μαύρο: #000000 (απουσία χρώματος)

- Λευκό: #ffffff (και τα 3 βασικά χρώματα στο μέγιστό τους)

- Γκρι: #a0a0a0 (ίση ποσότητα από όλα τα χρώματα)

- Κόκκινο: #ff0000

- Πράσινο: #0000ff

- Κίτρινο: #ffff00 (συνδυασμός κόκκινου και πράσινου)

```python
plot([sin(x),cos(x),sin(2*x),cos(2*x)], 0, 3*pi, color=(1,0.5,1))
```

Για να έχουμε όμως την δυνατότητα να θέσουμε ξεχωριστά χρώματα σε κάθε συνάρτηση θα πρέπει:

- να χρησιμοποιήσουμε λίστα και για τα χρώματα

- να προσθέσουμε τα plots κάθε συνάρτησης.

```python
plot([sin(x), cos(x), sin(2*x), cos(2*x)], (x, 0, 3*pi),  color=['#ff00af', (0.3,0.2,0.6), 'green', 'orange'])
```

```python
p1 = plot(sin(x), (x, 0, 3*pi), color='red')
p2 = plot(cos(x), (x, 0, 3*pi), color='blue')
p3 = plot(sin(2*x), (x, 0, 3*pi), color='green')
p4 = plot(cos(2*x), (x, 0, 3*pi), color='orange')
p = p1 + p2 + p3 + p4
p
```

Σημείωση: τα γραφήματα σχεδιάζονται από τα αριστερά στα δεξιά με το αριστερότερο να βρίσκεται κάτω από το επόμενό του κ.ο.κ.

Σε πολλές περιπτώσεις, μεταξύ των οποίων και η υπέρθεση γραφημάτων υπάρχει περίπτωση μία εντολή να μην χωράει σε μία γραμμή. Στην περίπτωση αυτή ο τελεστής `\` επιτρέπει την συνέχιση της εντολής στην επόμενη γραμμή.

```python
plot((x-1)*(x-2)*(x-3), (x, 1, 5), color='red') + plot((x-1)*(x-2)*(x-4), (x, 1, 5), color='green') +\
plot((x-1)*(x-2)*(x-5), (x, 1, 5), color='blue') + plot((x-1)*(x-2)*(x-6), (x, 1, 5), color='pink') +\
plot((x-1)*(x-2)*(x-7), (x, 1, 5), color='purple')
```

## Σημεία

Στο παραπάνω γράφημα, το σημείο $(1,2)$ μας ενδιαφέρει να τονιστεί. Αυτό μπορεί να γίνει με την συνάρτηση `point`. Μπορεί να τεθεί και το μέγεθος του σημείου κάνοντας χρήση της παραμέτρου `size`.

```python
p1 = plot( x^2+1, -2, 2,color='red', legend_label='$ x^2+1 $') 
p2 = plot( 2*x, 0.5, 1.5,color='blue', legend_label='$ 2x $')
p3 = point((1,2),size=20, color='green')
p1+p2+p3
```

Ένα σημείο, ή και μία λίστα σημειων, μπορούν να υπάρχουν μόνα τους ανεξάρτητα σε ένα γράφημα.

```python
point((1,3)) + point((2,4)) + point((3,5)) + point((4,6))
```

Σε περίπτωση που τα σημεία είναι αρκετά μπορεί να χρησιμοποιηθεί η δομή της λίστας που θα συζητηθεί σε επόμενη ενότητα. Σε αυτή την περίπτωση χρησιμοποιούμε εναλλακτικά την points αντί της point.

```python
f(x) = x^2
points([(0,f(0)), (1,f(1)), (2,f(2)), (3,f(3)), (4,f(4))])
```

## Κύκλοι, δίσκοι

Οι κύκλοι σχεδιάζονται με την `circle`, και οι δίσκοι με την `disk`. Και οι δύο συναρτήσεις δέχονται παράμετρο το κέντρο τους και την ακτίνα τους. Και στις δύο περιπτώσεις μπορούμε να θέσουμε κάποιο χρώμα, `color`, με το οποίο, στην περίπτωση του κύκλου θα σχεδιαστεί το περίγραμμα ενώ στην περίπτωση του δίσκου και το εσωτερικό του. Επιπλέον στον δίσκο θα πρέπει να θέσουμε το τμήμα του δίσκου που θα είναι ορατός. Ο κύκλος μπορεί και να είναι γεμάτος, θέτοντας την `fill` `True` αλλά και να έχει περίγραμμα διαφορετικού χρώματος με την `edgecolor` θέτοντας το εσωτερικό με την `facecolor`. 

```python
c = circle((0,0),1,color='pink')
d1 = disk((0,0), 0.8, (pi, 3*pi/2), color='yellow')
d2 = disk((0,0), 0.8, (0, pi/2), color='red')
d3 = disk((-0.5,0.5),0.2, (0, 2*pi),color='blue')
d4 = circle((0.5,-0.5),0.2,fill=True, edgecolor='black', facecolor='cyan')
c+d1+d2+d3+d4
```

Για παράδειγμα η επόμενη εντολή δημιουργεί μία έλλειψη με μήκος 2 και ύψος 1. Προφανώς αν αυτές είναι ίσες σχεδιάζεται ένας κύκλος. Οι παράμετροι  `edgecolor` για περίγραμμα και `facecolor` για γέμισμα ισχύουν και στις ελλείψεις. Για το τελευαίο θα πρέπει να έχει τεθεί `fill = True`.

```python
ellipse((0.5, 0), 1, 0.5)
```

## Βέλη

Με την `arrow` σχεδιάζουμε βέλη στο γράφημά μας. Απαιτούνται οι συντεταγμένες της αρχής και του τέλους του κάθε βέλους. Με την `color` έχουμε την δυνατότητα να προσδιορίσουμε το χρώμα του βέλους. 

```python
arrow((0,0), (1,1))+arrow((0.5,0.5), (0,1), color="red")
```

Στην περίπτωση που δεν επιθυμούμε οι άξονες να είναι ορατοί έχουμε την δυνατότητα να θέσουμε την παράμετρο `axes` ίση με `False`. Ακόμη και αν έχουμε διάφορα γραφικά στο γράφημά μας, αρκεί να δηλωθεί σε ένα από αυτά. Παρατηρούμε ακόμη στο επόμενο παράδειγμα ότι οπως και σε οποιαδήποτε γραφική παράσταση μπορεί να τεθεί το πάχος του βέλους με την `thickness`.

```python
g = arrow((0, 0), (1, 1), color="blue",  thickness=2, axes=False)
g += arrow((0.5, 0.5), (0, 1), color="red",  thickness=2.5)
g += arrow((0.2, 0.2), (1, 0.5), color="green",  thickness=3)
g += arrow((0.7, 0.3), (0.2, 0.8), color="purple",  thickness=3.5)
g += arrow((0.9, 0.1), (0.5, 0.9), color="orange",  thickness=4)
g.show()
```

## Κείμενο

Σε ένα γράφημα μπορούμε να συμπεριλάβουμε και ετικέτες μέσω της συνάρτησης `text`. Η πρώτη παράμετρος αφορά το κείμενο το οποίο θέλουμε να συμπεριλάβουμε και η επόμενη τις συντεταγμένες του σημείου στο οποίο αυτό θα εμφανιστεί. Πέρα από τις παραπάνω μπορούμε να προσαρμόσουμε την εμφάνισή του με μία σειρά από επιπλέον παραμέτρους.

- `fontsize`: Το μέγεθος του font του κειμένου. Η τιμή μπορεί να είναι αριθμός ή μία σειρά από λέξεις κλειδιά, `xx-small`, `x-small`, `medium`, `large`, `x-large`, `xx-large`.

- `fontstyle`: Το στυλ της γραμματοσειράς. Η τιμή του μπορεί να είναι `normal`, `italic`, `oblique`

- `fontweight`: Χρησιμοποιείται για να δίνουμε έμφαση στο κείμενο. Μπορούμε να δώσουμε αριθμητική τιμή στο  0-1000 ή μία από τις τιμές `ultralight`, `light`, `normal`, `regular`, `book`, `medium`, `roman`, `semibold`, `demibold`, `demi`, `bold`, `heavy`, `extra bold`, `black`.

- `vertical_alignment` και `horizontal_alignment`: Κατακόρυφη και οριζόντια στοίχιση αντίστοιχα.

- `rotation`: Περιστρέφει το κείμενο

- `alpha`: Καθορίζει το ποσοστό διαφάνειάς του.

Μία ακόμη τεχνική που παρατηρούμε στο επόμενο παράδειγμα είναι η κλήση, αρχικά, της `Graphics` η οποία δημιουργεί μία άδεια λίστα γραφικών αντικειμένων στην οποία μπορούμε να προσθέσουμε κάθε αντικείμενο. Θα μπορούσαμε να δημιουργήσουμε το πρώτο αντικείμενο, παράδειγμα ένα plot, και να προσθέσουμε σε αυτό στην συνέχεια και άλλα γραφικά αντικείμενα.  

Στο παρακάτω παράδειγμα δημιουργούμε το g και με συνεχόμενη χρήση του συνδυσμού += προσθέτουμε στο g όλα τα υπόλοιπα. Με την g ή την g.show() βλέπουμε το συνολικό αθροιστικό αποτέλεσμα

```python
# Δημιουργία κενής γραφικής παράστασης
g = Graphics()

g += line([(1, 0), (1, 2)], color='yellow')
g += line([(-1, 1), (2, 1)], color='magenta')

g += text('Left-aligned', (1, 1.5), fontsize=12, color='blue', horizontal_alignment='left')
g += text('Center-aligned', (1, 1.7), fontsize=12, color='green', horizontal_alignment='center')
g += text('Right-aligned', (1, 1.9), fontsize=12, color='red', horizontal_alignment='right')

g += text('Top-aligned', (-0.5, 1), fontsize=12, color='blue', vertical_alignment='top')
g += text('Center-aligned', (0, 1), fontsize=12, color='green', vertical_alignment='center')
g += text('Bottom-aligned', (0.5, 1), fontsize=12, color='red', vertical_alignment='bottom')

g += text('Rotated Text', (1, 1), fontsize='xx-large', color='blue', rotation=45, alpha=0.5)

g += text('oblique text', (-0.5, 1.9), fontsize=12, color='blue', fontstyle='oblique')
g += text('italic text', (-0.5, 1.7), fontsize=12, color='blue', fontstyle='italic')

g += text('ultralight text', (-0.8, 1.5), fontsize=12, color='blue', fontweight='ultralight')
g += text('medium text', (-0.8, 1.3), fontsize=12, color='blue', fontweight='medium')
g += text('extra bold text', (-0.8, 1.1), fontsize=12, color='blue', fontweight='extra bold')

g.show()
```

## LaTeX

Το κείμενο, όπως και το υπόμνημα, μπορεί να περιέχει μαθηματικά σύμβολα με LaTeX ανάμεσα στα σύμβολα `$`. Αποτελεί ισχυρό σύστημα στοιχειοθεσίας που χρησιμοποιείται για τη δημιουργία επιστημονικών και τεχνικών εγγράφων. Ένα από τα μεγαλύτερα πλεονεκτήματα του LaTeX, που μπορούμε να εκμεταλευτούμε στο Sage, είναι η υποστήριξή του για τη δημιουργία μαθηματικών συμβόλων και εξισώσεων υψηλής ποιότητας.

Αξίζει να σημειωθεί ότι latex μπορούμε να έμφανίσουμε μέσω της οδηγίας `%display latex` και σε κελιά στο notebook ή ακόμη και στο https://sagecell.sagemath.org/.

```python
%display latex
var('z')
z^12
```

Σημειώνεται ότι το σύμβολο `\` αποτελεί ειδικό σύμβολο στις συμβολοσειρές της Python και εφόσον θέλουμε να το χρησιμοποιήσουμε, έστω για κάποιο latex σύμβολο, θα πρέπει είτε να χρησιμοποιηθεί διπλό `\\` είτε να γίνει χρήση του ειδικού συμβόλου r στα αριστερά του κειμένου (που βρίσκεται μέσα στα ' ή στα"). Το r  αναγκάζει την Python να βλέπει το \ ως έχει δηλ. ως \ και όχι ως ξεκίνημα κάποιας 'ειδικής' συμβολοσειράς. Για παράδειγμα, μπορείτε να δοκιμάσετε την διαφορά των print('Hello,\n World') και print(r'Hello,\n World'). Στην πρώτη περίπτωση λόγω της εμφάνισης του ειδικού συνδυασμού \n (= αλλαγή γραμμής) η λέξη World εμφανίζεται στην επόμενη γραμμή. Η χρήση όμως του r αποτρέπει την Python (και άρα και την Sage) να βλέπει το \ ως κάποιο ειδικό χαρακτήρα, οπότε εκτυπώνει το 'Hello,\n World' ως έχει.

```python
# Αλλαγή γραμμής
print('Hello,\n World')
# Εμφάνιση συμβόλου \ με χρήση \\
print('Hello,\\n World')
# χρήση r, εμφάνιση αυτούσιας συμβολοσειρά
print(r'Hello,\n World')
```

```python
text(r'$\int_0^2 x^2\, dx$', (0.5,2))+plot(x^2,(x,0,2),fill=True)
```

## Ανισότητες

Για να σχεδιαστεί μία ανισότητα στο επίπεδο μπορεί να τονιστεί, με κάποιο χρώμα, είτε τα σημεία που ικανοποιούν την ανισότητα είτε αυτά που δεν την ικανοποιούν.

**Παράδειγμα**

Έστω η ανισότητα 

$$
y \geq 1 - x^2
$$

Θα πρέπει πρώτα να σχεδιαστεί το γράφημα της συνάρτησης 

$$
y = 1 - x^2
$$

και στην συνέχεια να γίνει ο χρωματισμός. Η γραφική παράσταση της $y$ χωρίζει το επίπεδο σε δυο διαφορετικές περιοχές.Η παραπάνω ανισότητα ικανοποιείται ακριβώς από όλα εκείνα τα σημεία που βρίσκονται πάνω από την γραφική παράσταση. 

```python
plot( 1-x^2, (x, -2, 2), ymin=-2, ymax=2, gridlines='minor',fill=2, fillcolor='yellow')
```

Η ιδιότητα `fill` καθορίζει ότι το γέμισμα θα γίνει  από τον οριζόντιο άξονα  y = 2 και μέχρι την γραφική παράσταση. Μπορεί όμως για παράδειγμα να τεθεί fill= -2 ή και κάτι μικρότερο ώστε αυτό να γίνει στο κάτω μέρος της γραφικής παράστασης. Τα σημεία που χρωματίζονται παρακάτω ικανοποιούν την (ανάποδη)  ανισότητα $
y \leq 1 - x^2
$

```python
plot(1 - x^2, (x, -2, 2), ymin=-2, ymax=2, gridlines='minor', fill=-2, fillcolor='yellow')
```

Αντί για την σταθερά -2 μπορούμε να θέσουμε στο fill άλλη συνάρτηση ώστε να χρωματιστεί  το ενδιάμεσο χωρίο. Το χωρίο αυτό προσδιορίζει όλα τα σημεία $(x,y)$ του επιπέδου που βρίσκονται μέσα  στο ενδιάμεσο χωρίο, δηλαδή για τα οποία ισχύει $g(x)\leq y\leq f(x)$ ή $f(x)\leq y\leq g(x)$.

```python
f(x) = 1 - x^2
g(x) = x^2 - 1
plotf = plot(f(x), (x, -2, 2), ymin=-2, ymax=2, color='blue', legend_label='$ f(x) = 1 - x^2 $')
plotg = plot(g(x), (x, -2, 2), color='red', legend_label='$ g(x) = x^2 - 1 $')
fillplot = plot(f(x), (x, -1, 1), ymin=-2, ymax=2, fill=g(x), fillcolor='yellow', gridlines='minor')
plotf + plotg + fillplot
```

## Πολικές συντεταγμένες

Η μέθοδος `polar_plot` σχεδιάζει γράφημα με πολικές συντεταγμένες. Στη γενική μορφή της, `polar_plot(f(theta), range_of_theta)`, όπου με `f(theta)` εννοούμε αποκλειστικά μια συνάρτηση της γωνίας (και όχι της ακτίνας f(r) ή της ακτίνας και της γωνίας `f(r,theta))` και με range_of_theta το πεδίο ορισμού της γωνίας (που κανονικά δεν πρέπει να έχει εύρος μεγαλύτερο του 2*π) . Οπότε αν ο χρήστης θέλει να σχεδιάσει την εξίσωση theta =r^2 σε πολικές, δεν μπορεί να χρησιμοποιήσει την `polar_plot(r^2, (r,1,2))` για τον σκοπό αυτό.

Στο παράδειγμα που ακολουθεί σχεδιάζεται ένας κύκλος με ακτίνα 1 με τη γωνία θ να είναι από 0 έως 2π. Έτσι σχεδιάζεται ένας ολόκληρος κύκλος. Επίσης, σχεδιάζεται ένα ημικύκλιο ακτίνας 0.5 με το θ από 0 έως π. Το τελευταίο ημικύκλιο, ακτίνας r=0.75, ξεκινά από τις 90 μοίρες, π/2, μέχρι και τις 270, 3π/2.

```python
var('theta')
p1 = polar_plot(1, (theta, 0, 2*pi),color = "purple")
p2 = polar_plot(0.5, (theta, 0, pi),color="red")
p3 = polar_plot(0.75, (theta, pi/2, 3*pi/2),color="brown")
p1+p2+p3
```

Η γραφική παράσταση του ημιτόνου (ή του συνημιτόνου) σε πολικές συντεταγένες δημιουργούν την λεγόμενη _rose curve_ με τα διπλάσια πέταλα από το συντελεστή της θ.

```python
var("theta")
polar_plot(cos(2*theta), (theta, 0, 2*pi))+polar_plot(sin(4*theta), (theta, 0, 2*pi),color="red")
```

## Σφαιρικές, κυλινδρικές συντεταγμένες

Στο SageMath, οι συναρτήσεις `spherical_plot3d` και `cylindrical_plot3d` χρησιμοποιούνται για τη δημιουργία τρισδιάστατων γραφικών με χρήση σφαιρικών και κυλινδρικών συντεταγμένων αντίστοιχα. Αυτά τα εργαλεία είναι χρήσιμα για την οπτικοποίηση αντικειμένων που ορίζονται σε αυτές τις συντεταγμένες, οι οποίες είναι συχνά πιο κατάλληλες από τις καρτεσιανές συντεταγμένες για συγκεκριμένα σχήματα, όπως σφαίρες ή κυλίνδρους. Σε κάθε σημείο  του χώρου αντιστοιχούν  οι:

- **r** ακτινική απόσταση (απόσταση από το $O$

- **θ** η πολική γωνία θ (η θετική γωνία που πρέπει να στρίψουμε τον θετικό άξονα $Ox$ έως ότου συναντήσει την προβολή του σημείου στο $Oxy$ επίπεδο)

- **φ** το αζιμούθιο φ (η θετική γωνία που πρέπει να στρίψουμε τον θετικό άξονα $Oz$ έως ότου συναντήσει το σημείο).




Η γενική μορφή της είναι spherical_plot3d(f(θ,φ), range_of_θ, range_of_φ), όπου με $f(θ,φ)$ εννοούμε μια συνάρτηση του $θ$ και του $φ$  που δίνει  την ακτινική απόσταση $r$. Το  range_of_θ είναι  το πεδίο ορισμού της μεταβλητής $θ$ (που κανονικά δεν πρέπει να έχει εύρος μεγαλύτερο του $2*π$) και το range_of_φ  το πεδίο ορισμού της μεταβλητής $φ$ (που κανονικά  έχει εύρος από $0$ έως $π$).

Το παράδειγμα που ακολουθεί σχεδιάσαμε μια επιφάνεια στις σφαιρικές συντεταγμένες, με  $r=1$. Θέσαμε δηλαδή $f(θ,φ)=1$. Προφανώς πρόκειται για την επιφάνεια μια σφαίρας ακτινας $r=1$.

```python
var('theta phi')
spherical_plot3d(1, (theta, 0, 2*pi), (phi, 0, pi))
```

Οι κυλινδρικές συντεταγμένες (cylindrical coordinates) αποτελούν μια επέκταση των πολικών συντεταγμένων σε τρεις διαστάσεις. Χρησιμοποιούνται για να περιγράψουν σημεία στο τρισδιάστατο χώρο με τις εξής παραμέτρους 

- **r** Η απόσταση του σημείου από τον κάθετο άξονα $Oz$.

- **θ** Η γωνία στο επίπεδο xy  (η θετική γωνία που πρέπει να στρίψουμε τον θετικό άξονα 𝑂𝑥 έως ότου συναντήσει την προβολή του σημείου στο 𝑂𝑥𝑦 επίπεδο).

- **z** Η τρίτη συντεταγμένη του σημείου $(x,y,z)$.

Η γενική μορφή της είναι cylindrical_plot3d(f(θ,z), range_of_theta, range_of_z), όπου με $f(θ,z)$ εννοούμε μια συνάρτηση του $θ$ και του $z$ (ύψους) που  δίνει την απόσταση $r$. Το  range_of_theta είναι  το πεδίο ορισμού της μεταβλητής $θ$ (που κανονικά δεν πρέπει να έχει εύρος μεγαλύτερο του $2*π$) και το  range_of_z  καθορίζει τις επιθυμητές διαστάσεις του ύψους $z$ της κυλινδρικής μας επιφάνειας με εξίσωση $r=f(θ,z)$. Στο παρακάτω παράδειγμα σχεδιάζουμε μια κυλινδρική επιφάνεια που απέχει σταθερή απόσταση $r=1$ από τον $Oz$. Για το λόγο αυτό θέσαμε $f(θ,z)=1$.  Προκύπτει ένας κύλινδρος ακτίνας $r=1$ και ύψους 2.

```python
var('z theta')
cylindrical_plot3d(1, (theta, 0, 2*pi),(z, 0, 2))
```

## Διάγραμμα ισοϋψών καμπυλών, contour plots

Ένα διάγραμμα ισοϋψών καμπυλών αποτελεί μια γραφική αναπαράσταση που δείχνει τα διαφορετικά  επίπεδα τιμών μιας συνάρτησης $f$  δύο μεταβλητών. Δημιουργείται με την `contour_plot`. Οι πιο μαύροι κύκλοι αντιστοιχούν σε μικρές τιμές της f ενώ οι πιο φωτεινοί σε μεγαλύτερες.

```python
var('x y')
f(x,y) = x^2 + y^2
contour_plot(f, (x, -2, 2), (y, -2, 2))
```

Έχουμε την δυνατότητα να θέσουμε τον αριθμό των ισοϋψών με την παράμετρο `contours`. Η παράμετρος `axes` εμφανίζει τους άξονες αν τεθεί `True`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), contours=20, axes=True)
```

Στο παράδειγμα που ακολουθεί επιλέξαμε να σχεδιαστούν  οι ισοϋψείς καμπύλες που αντιστοιχούν σε τρεις μόνο τιμές. Σε αυτές τις καμπύλες η συνάρτησή μας έχει πάρει τιμές -0.1, 0.1 και 0.5 αντίστοιχα. Για να το πετύχουμε αυτό θέσαμε contours=(-0.1,0.1,0.5). Η fill=False μας αφαίρεσε τα γκρίζα, ώστε να αναδειχθούν οι ισοϋψείς. 

```python
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), contours=(-0.1,0.1,0.5), fill=False)
```

Ο χρωματικός χάρτης, `cmap`, στα παραπάνω παραδείγματα είναι ο εξορισμού `grey`. Μπορεί να αλλάξει σε διάφορες τιμές όπως `plasma`, `hot`, `jet`, `coolwarm` ή ακόμη και σε φιλική στους έχοντες αχρωματοψία `cividis`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), contours=20, axes=True, cmap = "jet")
```

Χρήσιμο είναι να φαίνεται στο γράφημα τί τιμές αντιπροσωπεύει κάθε χρώμα. Αυτό πετυχαίνεται θέτοντας `True` στην ιδιότητα `colorbar`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), contours=20, axes=True, cmap = "jet", colorbar=True)
```
