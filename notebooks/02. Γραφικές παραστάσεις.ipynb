# Δισδιάστατα γραφικά

## plot

Η συνάρτηση `plot` επιτρέπει την σχεδίαση της γραφικής παράστασης συνάρτησης.

```python
plot(x^2)
```

## Εμφάνιση

Διάφορες επιλογές εμφάνισης μπορούν να τεθούν στις γραφικές μας παραστάσεις. Η `alpha`, από 0 έως την εξορισμού τιμή 1 θέτει την διαφάνεια από εντελώς διάφανη έως αδιαφανή. Η `thickness` με εξορισμού τιμή το 1 θέτει το πάχος της γραμμής. 

```python
plot(x *sin(1/x), thickness=4, alpha=0.5)
```

Η `linestyle` θέτει το στυλ της γραμμής στο εξορισμού συμπαγές `-`, σε διακεκομμένο `--`, διάστικτο `:`, παύλα και τελεία `-.`.

```python
plot(sin(pi*x),linestyle=':')
```

## Διάστημα

Το διάστημα εξορισμού είναι το [-1,1] αλλά μπορεί να τεθεί διαφορετικό ως παράμετρος στην `plot` 

```python
plot(sin(2*x), -pi, pi)
```

Υπάρχει και η δυνατότητα αλλαγής διαστήματος στον `y` άξονα μέσω των παραμέτρων `ymin`, `ymax`.

```python
plot( x^3-x, -3, 3, ymin = -6, ymax = 6)
```

## Αναλογία Διαστάσεων

Είναι πιθανό το SageMath να σχεδιάσει μία συνάρτηση με διαφορετικές κλίμακες στους άξονες. Για παράδειγμα αν ζητήσουμε το πάνω ημικύκλιο, δηλαδή
$$
y = \sqrt{1 - z^2}
$$
αυτό θα εμφανιστεί παραμορφωμένο. Υπενθυμίζεται ότι πέρα από την x, θα πρέπει όλες οι συμβολικές μεταβλητές να δηλώνονται πριν την χρήση τους όπως φαίνεται και στην επόμενη υλοποίηση.

```python
var('z')
y=sqrt(1-z^2)
plot(y)
```

Το `aspect_ratio` καθορίζει την αναλογία της μονάδας μήκους του άξονα Oy ώς προς την αντίστοιχη του Ox. Έτσι, θέτοντας την παράμετρο `aspect_ratio` ίση με 1 διατηρείται η σωστή γεωμετρία του ημικυκλίου και αποδίδεται το σχήμα με τις σωστές αναλογίες. 

```python
plot(y,aspect_ratio=1)
```

Φυσικά, η παράμετρος `aspect_ratio` μπορεί να πάρει και άλλες τιμές ανάλογα με το επιθυμητό αποτέλεσμα. Στο επόμενο παράδειγμα οι μονάδες διπλασιάζονται στον y σε σχέση με τον x άξονα.

```python
plot(y,aspect_ratio=2)
```

## Υπέρθεση γραφημάτων

Δύο συναρτήσεις μπορούν να απεικονιστούν στην ίδια γραφική παράσταση κάνοντας χρήση των `[ ]` και παραθέτοντας τες χωρισμένες με `,` όπως βλέπουμε στο επόμενο παράδειγμα.

```python
plot([sin(x),cos(x)], 0, 3*pi) 
```

Μπορούμε να έχουμε και παραπάνω από δύο συναρτήσεις, στις οποίες δίνεται εξορισμού κάποιο ξεχωριστό χρώμα.

```python
plot([sin(x),cos(x),sin(2*x),cos(2*x)], 0, 3*pi)
```

## Χρώμα

Το χρώμα μπορεί να τεθεί και από τον χρήστη με την παράμετρο `color`. Η τιμή της παραμέτρου μπορεί να τεθεί με έναν από τους επόμενους τρόπους:

- ένα λεκτικό, `red`, `green`, `blue`, `black` κ.ο.κ.

- μία τριάδα των βασικών χρωμάτων (r, g, b) από 0 έως 1. 

- ένας δεκαεξαδικός κωδικός 6 ψηφίων με κάθε δυάδα να αναφέρεται και πάλι στα τρία βασικά χρώματα.

Στην πραγματικότητα ο τρόπος αναπαράστασης χρωμάτων στον SageMath χρησιμοποιεί True Color δηλαδή 24 bits (binary digit 0 ή 1) πληροφορίας. Τα 24 bits χωρίζονται σε 3 ομάδες των 8 bits (ή 1 byte) για καθένα από τα τρία βασικά χρώματα: 

- κόκκινο (red)

- πράσινο (green)

- μπλε (blue).

Με αυτόν τον τρόπο, όμως, κάθε βασικό χρώμα μπορεί να πάρει τιμές από 0 (=00000000) έως 255 (=11111111) ή στο δεκαεξαδικό σύστημα 00 έως ff. Στο σύνολο μπορούμε να έχουμε 2^24 ή 256x256x256 = 16.777.216 συνδιασμούς και άρα διαφορετικά χρώματα.

Παραδείγματα:
- Μαύρο: #000000 (απουσία χρώματος)

- Λευκό: #ffffff (και τα 3 βασικά χρώματα στο μέγιστό τους)

- Γκρι: #a0a0a0 (ίση ποσότητα από όλα τα χρώματα)

- Κόκκινο: #ff0000

- Πράσινο: #0000ff

- Κίτρινο: #ffff00 (συνδυασμός κόκκινου και πράσινου)

```python
plot([sin(x),cos(x),sin(2*x),cos(2*x)], 0, 3*pi, color=(1,0.5,1))
```

Για να έχουμε όμως την δυνατότητα να θέσουμε ξεχωριστά χρώματα σε κάθε συνάρτηση θα πρέπει:

- να χρησιμοποιήσουμε λίστα και για τα χρώματα

- να προσθέσουμε τα plots κάθε συνάρτησης.

```python
plot([sin(x), cos(x), sin(2*x), cos(2*x)], (x, 0, 3*pi),  color=['#ff00af', (0.3,0.2,0.6), 'green', 'orange'])
```

```python
p1 = plot(sin(x), (x, 0, 3*pi), color='red')
p2 = plot(cos(x), (x, 0, 3*pi), color='blue')
p3 = plot(sin(2*x), (x, 0, 3*pi), color='green')
p4 = plot(cos(2*x), (x, 0, 3*pi), color='orange')
p = p1 + p2 + p3 + p4
p
```

Σημείωση: τα γραφήματα σχεδιάζονται από τα αριστερά στα δεξιά με το αριστερότερο να βρίσκεται κάτω από το επόμενό του κ.ο.κ.

Σε πολλές περιπτώσεις, μεταξύ των οποίων και η υπέρθεση γραφημάτων υπάρχει περίπτωση μία εντολή να μην χωράει σε μία γραμμή. Στην περίπτωση αυτή ο τελεστής `\` επιτρέπει την συνέχιση της εντολής στην επόμενη γραμμή.

```python
plot((x-1)*(x-2)*(x-3), (x, 1, 5), color='red') + plot((x-1)*(x-2)*(x-4), (x, 1, 5), color='green') +\
plot((x-1)*(x-2)*(x-5), (x, 1, 5), color='blue') + plot((x-1)*(x-2)*(x-6), (x, 1, 5), color='pink') +\
plot((x-1)*(x-2)*(x-7), (x, 1, 5), color='purple')
```

## Σημεία

```python
p1 = plot( x^2+1, -2, 2,color='red', legend_label='$ x^2+1 $') 
p2 = plot( 2*x, 0.5, 1.5,color='blue', legend_label='$ 2x $')
p3 = point((1,2),size=20, color='green')
p1+p2+p3
```

## Κύκλοι, δίσκοι

Οι κύκλοι σχεδιάζονται με την `circle`, και οι δίσκοι με την `disk`. Και οι δύο συναρτήσεις δέχονται παράμετρο το κέντρο τους και την ακτίνα τους. Και στις δύο περιπτώσεις μπορούμε να θέσουμε κάποιο χρώμα, `color`, με το οποίο, στην περίπτωση του κύκλου θα σχεδιαστεί το περίγραμμα ενώ στην περίπτωση του δίσκου και το εσωτερικό του. Επιπλέον στον δίσκο θα πρέπει να θέσουμε το τμήμα του δίσκου που θα είναι ορατός. Ο κύκλος μπορεί και να είναι γεμάτος, θέτοντας την `fill` `True` αλλά και να έχει περίγραμμα διαφορετικού χρώματος με την `edgecolor` θέτοντας το εσωτερικό με την `facecolor`. 

```python
c = circle((0,0),1,color='pink')
d1 = disk((0,0), 0.8, (pi, 3*pi/2), color='yellow')
d2 = disk((0,0), 0.8, (0, pi/2), color='red')
d3 = disk((-0.5,0.5),0.2, (0, 2*pi),color='blue')
d4 = circle((0.5,-0.5),0.2,fill=True, edgecolor='black', facecolor='cyan')
c+d1+d2+d3+d4
```

## Κείμενο

Σε ένα γράφημα μπορούμε να συμπεριλάβουμε και ετικέτες μέσω της συνάρτησης `text`. Η πρώτη παράμετρος αφορά το κείμενο το οποίο θέλουμε να συμπεριλάβουμε και η επόμενη τις συντεταγμένες του σημείου στο οποίο αυτό θα εμφανιστεί. Πέρα από τις παραπάνω μπορούμε να προσαρμόσουμε την εμφάνισή του με μία σειρά από επιπλέον παραμέτρους.

- `fontsize`: Το μέγεθος του font του κειμένου. Η τιμή μπορεί να είναι αριθμός ή μία σειρά από λέξεις κλειδιά, `xx-small`, `x-small`, `medium`, `large`, `x-large`, `xx-large`.

- `fontstyle`: Το στυλ της γραμματοσειράς. Η τιμή του μπορεί να είναι `normal`, `italic`, `oblique`

- `fontweight`: Χρησιμοποιείται για να δίνουμε έμφαση στο κείμενο. Μπορούμε να δώσουμε αριθμητική τιμή στο  0-1000 ή μία από τις τιμές `ultralight`, `light`, `normal`, `regular`, `book`, `medium`, `roman`, `semibold`, `demibold`, `demi`, `bold`, `heavy`, `extra bold`, `black`.

- `vertical_alignment` και `horizontal_alignment`: Κατακόρυφη και οριζόντια στοίχιση αντίστοιχα.

- `rotation`: Περιστρέφει το κείμενο

- `alpha`: Καθορίζει το ποσοστό διαφάνειάς του.

Μία ακόμη τεχνική που παρατηρούμε στο επόμενο παράδειγμα είναι η κλήση, αρχικά, της `Graphics` η οποία δημιουργεί μία άδεια λίστα γραφικών αντικειμένων στην οποία μπορούμε να προσθέσουμε κάθε αντικείμενο. Θα μπορούσαμε να δημιουργήσουμε το πρώτο αντικείμενο, παράδειγμα ένα plot και να προσθέστουμε τα υπόλοιπα σε αυτό.

```python
# Δημιουργία κενής γραφικής παράστασης
g = Graphics()

g += line([(1, 0), (1, 2)], color='yellow')
g += line([(-1, 1), (2, 1)], color='magenta')

g += text('Left-aligned', (1, 1.5), fontsize=12, color='blue', horizontal_alignment='left')
g += text('Center-aligned', (1, 1.7), fontsize=12, color='green', horizontal_alignment='center')
g += text('Right-aligned', (1, 1.9), fontsize=12, color='red', horizontal_alignment='right')

g += text('Top-aligned', (-0.5, 1), fontsize=12, color='blue', vertical_alignment='top')
g += text('Center-aligned', (0, 1), fontsize=12, color='green', vertical_alignment='center')
g += text('Bottom-aligned', (0.5, 1), fontsize=12, color='red', vertical_alignment='bottom')

g += text('Rotated Text', (1, 1), fontsize='xx-large', color='blue', rotation=45, alpha=0.5)

g += text('oblique text', (-0.5, 1.9), fontsize=12, color='blue', fontstyle='oblique')
g += text('italic text', (-0.5, 1.7), fontsize=12, color='blue', fontstyle='italic')

g += text('ultralight text', (-0.8, 1.5), fontsize=12, color='blue', fontweight='ultralight')
g += text('medium text', (-0.8, 1.3), fontsize=12, color='blue', fontweight='medium')
g += text('extra bold text', (-0.8, 1.1), fontsize=12, color='blue', fontweight='extra bold')

g.show()
```

## Ανισότητες

Για να σχεδιαστεί μία ανισότητα στο επίπεδο μπορεί να τονιστεί, με κάποιο χρώμα, είτε τα σημεία που ικανοποιούν την ανισότητα είτε αυτά που δεν την ικανοποιούν.

---

**Παράδειγμα**

Έστω η ανισότητα 

$$
y \geq 1 - x^2
$$

Θα πρέπει πρώτα να σχεδιαστεί το γράφημα της συνάρτησης 

$$
y = 1 - x^2
$$

και στην συνέχεια να γίνει ο χρωματισμός.

```python
plot( 1-x^2, (x, -2, 2), ymin=-2, ymax=2, gridlines='minor',fill=2, fillcolor='yellow')
```

Η ιδιότητα `fill` καθορίζει ότι το γέμισμα θα γίνει από το y = 2 και μέχρι την γραφική παράσταση. Μπορεί για παράδειγμα να τεθεί στο -2 ή και κάτι μικρότερο ώστε αυτό να γίνει στο κάτω μέρος της παράστασης.

```python
plot(1 - x^2, (x, -2, 2), ymin=-2, ymax=2, gridlines='minor', fill=-5, fillcolor='yellow')
```

Αντί για την σταθερά -5 μπορούμε να θέσουμε άλλη συνάρτηση ώστε να χρωματιστεί το ενδιάμεσο χωρίο.

```python
f(x) = 1 - x^2
g(x) = x^2 - 1
plotf = plot(f(x), (x, -2, 2), ymin=-2, ymax=2, color='blue', legend_label='$ f(x) = 1 - x^2 $')
plotg = plot(g(x), (x, -2, 2), color='red', legend_label='$ g(x) = x^2 - 1 $')
fillplot = plot(f(x), (x, -1, 1), ymin=-2, ymax=2, fill=g(x), fillcolor='yellow', gridlines='minor')
plotf + plotg + fillplot
```

## Πολικές συντεταγμένες

Η μέθοδος `polar_plot` σχεδιάζει γράφημα με πολικές συντεταγμένες. Στη γενική μορφή της, `polar_plot(f(theta), rangeoftheta)`, όπου με `f(theta)` εννοούμε αποκλειστικά μια συνάρτηση της γωνίας (και όχι της ακτίνας f(r) ή της ακτίνας και της γωνίας `f(r,theta))` και με rangetheta το πεδίο ορισμού της γωνίας (που κανονικά δεν πρέπει να έχει εύρος μεγαλύτερο του 2*π) . Οπότε αν ο χρήστης θέλει να σχεδιάσει την εξίσωση theta =r^2 σε πολικές, δεν μπορεί να χρησιμοποιήσει την `polar_plot(r^2, (r,1,2))` για τον σκοπό αυτό.

Στο παράδειγμα που ακολουθεί σχεδιάζεται ένας κύκλος με ακτίνα 1 με τη γωνία θ να είναι από 0 έως 2π σχεδιάζοντας ολόκληρο τον κύκλο. Επίσης, ένα ημικύκλιο ακτίνας 0.5 με το θ από 0 έως π. Το τελευταίο ημικύκλιο ξεκινά από τις 90 μοίρες, π/2, μέχρι και τις 270, 3π/2.

```python
var('theta')
p1 = polar_plot(1, (theta, 0, 2*pi),color = "purple")
p2 = polar_plot(0.5, (theta, 0, pi),color="red")
p3 = polar_plot(0.75, (theta, pi/2, 3*pi/2),color="brown")
p1+p2+p3
```

Η γραφική παράσταση του ημιτόνου (ή του συνημιτόνου) σε πολικές συντεταγένες δημιουργούν την λεγόμενη _rose curve_ με τα διπλάσια πέταλα από το συντελεστή της θ.

```python
var("theta")
polar_plot(cos(2*theta), (theta, 0, 2*pi))+polar_plot(sin(4*theta), (theta, 0, 2*pi),color="red")
```

Οι τυπικές μορφές συναρτήσεων, των οποίων γραφήματα είδαμε παραπάνω, εκφράζονται ως y = f (x), με την τιμή της y να εξαρτάται από τις τιμές του x. Οι πεπλεγμένες συναρτήσεις εκφράζονται με εξισώσεις που έχουν την μορφή f(x,y) = 0. Όταν έχουμε μία πεπλεγμένα εκφρασμένη συνάρτηση, η γραφική της παράσταση μπορεί να αποδοθεί με την `implicit_plot`.

$$
x^2 + y^2 - 16 = 0
$$

```python
var('x y')
g(x,y)=x^2+y^2-16
implicit_plot(g, (x,-5,5), (y,-5,5))
```

# Τρισδιάστατα γραφικά

Μία γραφική παράσταση συνάρτησης δύο μεταβλητών σχεδιάζεται με τη συνάρτηση `plot3d`.

Υπενθυμίζεται ότι, πέρα από τη μεταβλητή `x` η οποία εξορισμού θεωρείται συμβολική οι υπόλοιπες, εδώ το `y` θα πρέπει να δηλωθεί. Θα πρέπει να καθοριστούν τα διαστήματα και για τις δύο μεταβλητές.
Ακολουθεί η γραφική παράσταση της συνάρτησης 

$$ \sin(x^2 + y^2) $$

```python
y = var('y')
f = sin(x^2+y^2)
plot3d(f,(x,-5,5), (y,-5,5))
```

Τα τρισδιάστατα γραφήματα παρέχουν διάφορες λειτουργίες για βολικότερη εμφάνιση:

- Με τη βοήθεια του ποντικιού μπορεί κανείς να δει διάφορες όψεις του παραπάνω γραφήματος. Κρατώντας απλά το αριστερό click πατημένο το γράφημα περιστρέφεται. 

- Με αριστερό click και ctrl στο πληκτρολόγιο το γράφημα μετακινείται. Το ίδιο και με το δεξί click.

- Με την ροδέλα του ποντικιού γίνεται zoom.

- Στο μικρό κουμπί, κάτω δεξιά στο γράφημα, υπάρχει μία σειρά από επιλογές αποθήκευσης του γραφήματος.

## Δείγματα

Μέσω της παραμέτρου `plot_points` δίνεται η δυνατότητα αλλαγής στον αριθμό των δειγμάτων για την δημιουργία της γραφικής παράστασης. Τα πολλά σημεία, όμως, καθυστερούν την απόδοση του γραφήματος και επιπλέον μεγαλώνουν το μέγεθος του αρχείου.

```python
plot3d(f,(x,-5,5),(y,-5,5),plot_points=200)
```

## Υπέρθεση γραφημάτων

Ακριβώς όπως και στα δυσδιάστατα γραφήματα, υπάρχει η δυνατότητα υπέρθεσης 2 ή παραπάνω γραφημάτων.

Στο παράδειγμα που ακολουθεί πέρα από την συνάρτηση 

$$ \sin(x^2 + y^2) $$ 

εμφανίζεται και το xy-επίπεδο 

$$ f(x,y)=0 $$

```python
a=-2
b = 2
p1=plot3d(sin(x^2-y^2),(x,a,b),(y,a,b))
p2=plot3d(0, (x,a,b), (y,a,b), color='red')
p1+p2
```

## Διαφάνεια

Με την παράμετρο `opacity` μπορεί να τεθεί διαφάνεια σε κάποια παράσταση συνάρτησης ώστε να μην "κρύβει" κάποια άλλη συνάρτηση. Όταν η τιμή της παραμέτρου είναι `0.0`, η γραφική παράσταση δεν είναι ορατή, όταν είναι `1.0` είναι αδιαφανή. Όλες οι ενδιάμεσες τιμές εμφανίζουν την συνάρτηση περισσότερο ή λιγότερο διάφανη. 

```python
a=-2
b = 2
p2=plot3d(0, (x,a,b), (y,a,b), color='red', opacity=0.5)
p1+p2
```

## Διάγραμμα ισοϋψών καμπυλών, contour plots

Ένα διάγραμμα ισοϋψών καμπυλών αποτελεί μια γραφική αναπαράσταση που δείχνει τα επίπεδα τιμών μιας συνάρτησης δύο μεταβλητών. Δημιουργείται με την `contour_plot`.

```python
var('x y')
f(x,y) = x^2 + y^2
contour_plot(f, (x, -2, 2), (y, -2, 2))
```

Έχουμε την δυνατότητα να θέσουμε τον αριθμό των ισοϋψών με την παράμετρο `contours`. Η παράμετρος `axes` εμφανίζει τους άξονες αν τεθεί `True`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), contours=20, axes=True)
```

Ο χρωματικός χάρτης, `cmap`, παραπάνω είναι ο εξορισμού `grey`. Μπορεί να αλλάξει σε διάφορες τιμές όπως `plasma`, `hot`, `jet`, `coolwarm` ή ακόμη και η φιλική στους έχοντες αχρωματοψία `cividis`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), contours=20, axes=True, cmap = "jet")
```

Χρήσιμο είναι να φαίνεται στο γράφημα τί τιμές αντιπροσωπεύει κάθε χρώμα, κάτι που τίθεται θέτοντας `True` στην ιδιότητα `colorbar`.

```python
f(x,y) = cos(x) + sin(y)
contour_plot( f(x,y), (x,-pi,pi), (y,-pi,pi), contours=20, axes=True, cmap = "jet", colorbar=True)
```

##  density plot

Η `density_plot` χρησιμοποιείται για να δείξει πώς οι τιμές μιας συνάρτησης δύο μεταβλητών μεταβάλλονται σε μια συγκεκριμένη περιοχή, με διαφορετικά χρώματα να αντιστοιχούν σε διαφορετικές τιμές.

```python
var('x y')
f(x, y) = cos(x) + sin(y)
density_plot(f(x,y), (x,-pi,pi), (y,-pi,pi), axes=True, cmap='jet')
```

Φυσικά, τα `density_plot` και `contour_plot` μπορούν να συνδυαστούν. 

```python
f(x,y) = cos(x) + sin(y)
p1=contour_plot(f(x,y), (x,-pi,pi), (y,-pi,pi), axes=True, fill=false)
p2=density_plot(f(x,y), (x,-pi,pi), (y,-pi,pi), axes=True, cmap='jet')
p1+p2
```
