# Εξισώσεις και ανισότητες

Η `solve` αποτελεί την βασική συνάρτηση επίλυσης εξισώσεων

```python
solve( cot(x)==0, x )
```

## Σύστημα Εξισώσεων

Σε περίπτωση που έχουμε σύστημα εξισώσεων απαιτούνται παράμετροι για κάθε άγνωστο.

```python
var('y')
solve ([2* x - y ==3 , x +4* y == -2] , x , y)
```

## Πολλαπλές λύσεις

```python
#εύρεση των λύσεων
solutions=solve( log( x^2 ) == 5/3, x )

#το πλήθος των λύσεων το βρίσκω με len(λύσεις)

p1=plot( log( x^2 ) ,-3,3)
p2=plot(5/3,-3, 3, color = 'red')
p3= point( [solutions[0].rhs(), 5/3],size=50 )
(p1+p2+p3).show()
(len(solutions), solutions)
```

# Δακτύλιοι

Όταν ορίζουμε διανύσματα, πίνακες ή και πολυώνυμα, είναι χρήσιμο να καθορίζουμε το «δακτύλιο», ring, πάνω στον οποίο είναι ορισμένα. Ένας δακτύλιος είναι μια μαθηματική κατασκευή στην οποία υπάρχουν καλά ορισμένες έννοιες της πρόσθεσης και του πολλαπλασιασμού. Οι τέσσερεις συνηθέστεροι δακτύλιοι είναι οι:
- **ακέραιοι**, `ZZ` στο Sage.
- **ρητοί**  – δηλαδή, κλάσματα ή λόγοι ακεραίων – `QQ` στο Sage.
- **πραγματικοί** , `RR` στο Sage.
- **μιγαδικοί** , `CC` στο Sage.

Για την δημιουργία διανυσμάτων χρησιμοποιούμε την συνάρτηση `vector()` με είσοδο μία λίστα στοιχείων.

```python
v[0],v[-1]
```

Η προσπέλαση στοιχείου εκτός διανύσματος οδηγεί σε λάθος.

```python
v[4]
```

## Πίνακες

Οι πίνακες δημιουργούνται με τη μέθοδο `matrix`.

```python
matrix([[1,2],[3,4]])
```

Επιπλέον, δημιουργούμε πίνακες θέτοντας αρχικά τις διαστάσεις τους, γραμμές και στήλες αντίστοιχα. Η επόμενη για παράδειγμα εντολή δημιουργεί πίνακα 3 γραμμών και 2 στηλών.

```python
matrix(3,2, [1,2,3,4,5,6])
```

Σε περίπτωση τετραγωνικού πίνακα μπορούμε να παραλέιψουμε τον αριθμό των στηλών.

```python
m = matrix(3, [1,2,3,4,5,6,7,8,9]);m
```

Για την προσπέλαση στοιχείων πίνακα απαιτούνται δύο δείκτες, η γραμμή και η στήλη. 

```python
m[2,1] # στοιχείο 3ης γραμμής, 2ης στήλης. Αρίθμηση και πάλι από το 0.
```

## Αριθμητική πινάκων

Οι τελεστές `+`, `-`, `*`, `^` μπορούν να χρησιμοποιηθούν για πρόσθεση, αφαίρεση, πολλαπλασιασμό και δυνάμεις.

```python
A=matrix(2,[1,2,0,1])
B=matrix(2,[1,0,3,1])
A,B
```

```python
A+B
```

```python
A-B
```

```python
A*B
```

```python
A^3,A*A*A
```

Η ύψωση στην $ -1 $ θα οδηγήσει στον υπολογισμό του αντίστροφου του πίνακα.

```python
A = matrix([[1, 2], [3, 4]])
A^-1
```

Αν ο πίνακας είναι μη αντιστρέψιμος, θα προκύψει ZeroDivisionError.

```python
B = Matrix([[1, 2], [2, 4]])
show(B.det())
B^-1
```
