# Δομές επανάληψης

Οι υπολογιστές είναι πολύ καλοί στο να εκτελούν επαναλαμβανόμενες, ίδιες ή παρόμοιες, εργασίες γλιτώνοντας τον άνθρωπο από κουραστικές, βαρετές εργασίες. Οι εντολές που καθοδηγούν τον υπολογιστή να εκτελέσει κάτι επαναλαμβανόμενο είναι η for και η while.

## for

Η επόμενη δομή εμφανίζει τα τέλεια τετράγωνα από το 1 έως το n-11. Παρατηρούμε ότι η range ξεκινά από το 0 και φτάνει μέχρι 1 πριν την n. Με αυτόν τον τρόπο η range(n) θα δίνει n αντικείμενα.

````python
n=11
for i in range(11): print(i^2,end=" ")
````

Η γενική μορφή της `range` είναι: `range(αρχή, όριο, βήμα)` όπου το `όριο` είναι η *επόμενη* τιμή του τελευταίου στοιχείου της ακολουθίας. Οι παράμετροι `αρχή` και `βήμα` είναι προαιρετικοί.

Στην παραπάνω `for` η εντολή `print`, που εκτελείται επαναληπτικά, βρίσκεται δίπλα από την `for`. Θεωρείται καλή πρακτική, ειδικά όταν πολλές εντολές θέλουμε να εκτελεστούν μέσα στην επανάληψη, οι εντολές να γράφονται σε ξεχωριστές γραμμές και κάτω από την `for`. Θα πρέπε όμως να υπάρχει μία εσοχή, συνήθως ένα tab, το οποίο και διατηρείται στη διάρκεια της επανάληψης. Στο επόμεο παράδειγμα  η 1η `print` εκτελείται επαναληπτικά ενώ η 2η μετά το τέλος της επανάληψης.

````python
for i in range(5,105,5): 
    print(i, end= " ")
print('Φτου και βγαίνω')
````

Τί θα συμβεί αν και η 2η `print` έχει την ίδια εσοχή με την 1η;

````python
for x in range(-5,6):
    print(x^2-x, end= " ")
````

Στην πραγματικότητα, ο παραπάνω είναι ο τρόπος με τον οποίο σχεδιάζονται οι γραφικές παραστάσεις. Αποτιμώνται οι συναρτήσεις σε αρκετά σημεία τα οποία και σχεδιάζονται.

Υπάρχει περίπτωση να επιθυμούμε την αποτίμηση της παραπάνω συνάρτησης σε συγκεκριμμένες τιμές και όχι αναγκαστικά στις -5, -4, ...  που είδαμε παραπάνω. Τότε η for πρέπει να συνδυαστεί με μία λίστα τιμών. Οι λίστες θα αναλυθούν στην επόμενη ενότητα.

````python
for x in [-1, 0, 3/5, 1, 12/5]:
    print(x^2-x, end= " ")
````

Θα υπολογίσουμε το όριο:
$$
\lim_{x \to 1} \frac{2x^2 + x + 4}{x + 1}
$$
Αυτό θα γίνει με την limit αλλά και με επανάληψη κατά την οποία το x τείνει στο 1 από αριστερά και δεξιά. Αρχικά θα πάρουμε μία εικόνα της συνάρτησης από την γραφική της παράσταση.

````python
#TOTO plot
f(x) = (2*x^2 + x + 4) / (x + 1)
limit(f(x), x=1)
for i in range(10):
    v = 0.9 + i*0.01
    print ('f(',n(v,digits=2),')=',f(v), sep='')
for i in range(10):
    v = 1.1 - i*0.01
    print ('f(',n(v,digits=2),')=',f(v), sep='')
````

# Συναρτήσεις

Οι συναρτήσεις αποτελούν ένα μέρος προγράμματος, με ομαδοποιημένες εντολές που τους έχει δοθεί ένα όνομα και μπορούν να χρησιμοποιηθούν όσες φορές χρειαστεί.
Ορίζονται ως εξής:
```python
def f (λίστα_παραμέτρων) :
    εντολές
    return τιμή
```
Η λίστα παραμέτρων μπορεί να είναι κενή και η επιστρεφόμενη τιμή δεν είναι υποχρεωτική.

Ακολουθεί ο ορισμός μία συνάρτησης. Η εντολή `print` εμφανίζει στην οθόνη και θα αναλυθεί στην συνέχεια.

````python
def hello():
    print('Hello World')
````

Εκτελώντας τον προηγούμενο κώδικα παρατηρούμε ότι δεν εμφανίζεται κάτι. Για να εκτελεστούν οι εντολές που περιλαμβάνει μία συνάρτηση θα πρέπει αυτή να κληθεί. Ακολουθεί η κλήση τη συνάρτησης.

````python
hello()
````

Η παραπάνω συνάρτηση δεν μπορεί να κάνει και πολλά γιατί δεν επικοινωνεί με κάποιον τρόπο με το υπόλοιπο πρόγραμμα. Αυτό μπορεί να γίνει μέσω λίστας παραμέτρων, παραπάνω είναι κενή, και επιστρεφόμενης τιμής. Ακολουθεί η συνάρτηση με την προσθήκη μίας παραμέτρου.

````python
def hello(name):
    print('Hello',name)
hello('Isaac')
hello('Carl')
````
