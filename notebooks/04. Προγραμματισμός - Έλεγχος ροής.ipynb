# Δομές επανάληψης

Οι υπολογιστές είναι πολύ καλοί στο να εκτελούν επαναλαμβανόμενες, ίδιες ή παρόμοιες, εργασίες γλιτώνοντας τον άνθρωπο από κουραστικές, βαρετές εργασίες. Οι εντολές που καθοδηγούν τον υπολογιστή να εκτελέσει κάτι επαναλαμβανόμενο είναι η `for` και η `while`.

## for

Η επόμενη δομή εμφανίζει τα τέλεια τετράγωνα από το 1 έως το n-11. Παρατηρούμε ότι η `range` ξεκινά από το 0 και φτάνει μέχρι 1 πριν την n. Με αυτόν τον τρόπο η `range(n)` θα δίνει n αντικείμενα.

````python
n=11
for i in range(11): print(i^2,end=" ")
````

Η γενική μορφή της `range` είναι: `range(αρχή, όριο, βήμα)` όπου το `όριο` είναι η *επόμενη* τιμή του τελευταίου στοιχείου της ακολουθίας. Οι παράμετροι `αρχή` και `βήμα` είναι προαιρετικοί.

Στην παραπάνω `for` η εντολή `print`, που εκτελείται επαναληπτικά, βρίσκεται δίπλα από την `for`. Θεωρείται καλή πρακτική, ειδικά όταν πολλές εντολές θέλουμε να εκτελεστούν μέσα στην επανάληψη, οι εντολές να γράφονται σε ξεχωριστές γραμμές και κάτω από την `for`. Θα πρέπε όμως να υπάρχει μία εσοχή, συνήθως ένα tab, το οποίο και διατηρείται στη διάρκεια της επανάληψης. Στο επόμεο παράδειγμα  η 1η `print` εκτελείται επαναληπτικά ενώ η 2η μετά το τέλος της επανάληψης.

````python
for i in range(5,105,5): 
    print(i, end= " ")
print('Φτου και βγαίνω')
````

Τί θα συμβεί αν και η 2η `print` έχει την ίδια εσοχή με την 1η;

````python
for x in range(-5,6):
    print(x^2-x, end= " ")
````

Στην πραγματικότητα, ο παραπάνω είναι ο τρόπος με τον οποίο σχεδιάζονται οι γραφικές παραστάσεις. Αποτιμώνται οι συναρτήσεις σε αρκετά σημεία τα οποία και σχεδιάζονται.

Υπάρχει περίπτωση να επιθυμούμε την αποτίμηση της παραπάνω συνάρτησης σε συγκεκριμμένες τιμές και όχι αναγκαστικά στις -5, -4, ...  που είδαμε παραπάνω. Τότε η for πρέπει να συνδυαστεί με μία λίστα τιμών. Οι λίστες θα αναλυθούν στην επόμενη ενότητα.

````python
for x in [-1, 0, 3/5, 1, 12/5]:
    print(x^2-x, end= " ")
````

**Παράδειγμα**

Δημιουργήστε διαδοχικές συναρτήσεις της python οι οποίες δέχονται ακτίνα και ύψος και υπολογίζουν εμβαδόν και όγκο ενός κώνου.
Ο όγκος κώνου δίνεται από τον τύπο:
$$
V = \frac{1}{3} \pi r^2 h
$$
Το συνολικό εμβαδόν ττης επιφάνειας του κώνου δίνεται:
$$
A = \pi r s + \pi r^2
$$

Όπου:

- r: Η ακτίνα του κώνου, δηλαδή η ακτίνα της βάσης του κώνου

- h: Το ύψος του κώνου

- s: Η γενέτειρα του κώνου, δηλαδή η πλευρά από την κορυφή του κώνου προς το πείγραμμα της βάσης
$$
s = \sqrt{r^2 + h^2}
$$


````python
# TODO Να φτιάξω Plot;;
# Ακτίνα και ύψος του κώνου
radius = 3
height = 5

# Σχεδιάζουμε τον κώνο
var('x,y,z')
cone_plot = implicit_plot3d(x^2+y^2-z^2==0, (x, -3, 3), (y, -3,3), (z, 0,3))

# Σημεία K, O, A, B
K = (0, 0, height)   # Κορυφή του κώνου
O = (0, 0, 0)        # Κέντρο της βάσης
A = (radius, 0, 0)   # Σημείο πάνω στην ακτίνα της βάσης
B = (-radius, 0, 0)  # Αντίθετο σημείο πάνω στην ακτίνα της βάσης

# Προσθέτουμε τα σημεία στο διάγραμμα
points_plot = point3d([K, O, A, B], size=30, color='red')

# Προσθέτουμε ετικέτες για τα σημεία
text_labels = text3d('K', K, color='red', fontsize=12) + \
              text3d('O', O, color='red', fontsize=12) + \
              text3d('A', A, color='red', fontsize=12) + \
              text3d('B', B, color='red', fontsize=12)

# Εμφανίζουμε το τελικό διάγραμμα με τον κώνο, τα σημεία και τις ετικέτες
final_plot = cone_plot + points_plot + text_labels
show(final_plot)
````

````python
# Συνάρτηση που υπολογίζει εμβαδόν
def coneArea(radius, height):
    # Υπολογισμός γενέτειρας
    slant_height = sqrt(radius^2 + height^2)
    
    # Υπολογισμός εμβαδού επιφάνειας κώνου (πλάγια + βάση)
    lateral_area = pi * radius * slant_height
    base_area = pi * radius^2
    surface_area = lateral_area + base_area
    
    return surface_area
def coneVolume(radius, height):
    volume = (1/3) * pi * radius^2 * height
    return volume
# Παράδειγμα κλήσης της συνάρτησης
coneArea(3, 5),coneVolume(3,5)
````

````python
f(x) = x^3 - 2*x - 5
def bolzano_method(f, a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("Η συνάρτηση πρέπει να αλλάζει πρόσημο στο διάστημα [a, b]")
    
    for i in range(max_iter):
        # Υπολογισμός της μέσης τιμής
        c = (a + b) / 2
        # Έλεγχος αν βρέθηκε η ρίζα
        if abs(f(c)) < tol or (b - a) / 2 < tol:
            return c
        # Ενημέρωση του διαστήματος [a, b]
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    raise ValueError("Δεν βρέθηκε ρίζα μετά από το μέγιστο αριθμό επαναλήψεων")

# Εύρεση της ρίζας της συνάρτησης στο διάστημα [2, 3]
root = bolzano_method(f, a=2.0, b=3.0)
root
````
