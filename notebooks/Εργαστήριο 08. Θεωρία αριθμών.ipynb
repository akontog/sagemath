# Sophie Germain πρώτοι

Ένας πρώτος αριθμός **Sophie Germain** είναι ένας πρώτος αριθμός _p_ για τον οποίο ο αριθμός  $2p+1$ είναι επίσης πρώτος. Για παράδειγμα, αν p=11, τότε 2×11+1=23, και επειδή τόσο το 11 όσο και το 23 είναι πρώτοι, το 11 θεωρείται πρώτος αριθμός **Sophie Germain**. Η κατηγορία αυτή ονομάστηκε έτσι προς τιμήν της Γαλλίδας μαθηματικού η οποία συνέβαλε σημαντικά στη θεωρία αριθμών και ιδιαίτερα στο Θεώρημα του Fermat.
