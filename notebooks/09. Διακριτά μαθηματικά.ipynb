# Προτασιακός Λογισμός

## Λογικές πράξεις

Οι λογικές πράξεις είναι βασικές λειτουργίες στη μαθηματική λογική και τη θεωρία συνόλων. Χρησιμοποιούνται για να συνδυάζονται λογικές προτάσεις ή συνθήκες με στόχο τη λήψη συμπερασμάτων.

Ακολουθεί πίνακας με τις βασικές λογικές πράξεις.

| Λογική Πράξη                    | Σύμβολο     | Περιγραφή                                                      |
|----------------------------------|-------------|----------------------------------------------------------------|
| Άρνηση                          | `~`        | Το αντίθετο της πρότασης                                  |
| Σύζευξη (ΚΑΙ)                   | `&`         | Όλες οι προτάσεις είναι αληθείς             |
| Διάζευξη (Ή)                    | `\|`        | Τουλάχιστον μία από τις προτάσεις είναι αληθής      |
| Αποκλειστικό Ή (XOR)            | `^`         | Περιττός αριθμός προτάσεων είναι αληθής |
| Συνεπαγωγή (Αν ... τότε)         | `->`        | Αν ισχύει η πρώτη πρόταση, τότε πρέπει να ισχύει η δεύτερη                                               |
| Ισοδυναμία (Αν και Μόνο Αν)     | `<->`       | Οι προτάσεις είναι ταυτόσημες |


Για να εφαρμόσουμε τις παραπάνω λογικές πράξεις κάνουμε χρήση της μεθόδου `formula` του `propcalc`. Στη συνέχεια αποτιμάμε την έκφραση θέτοντας τις τιμές στις λογικές μεταβλητές με την μέθοδο `evaluate`. Οι τιμές δίνονται με τη βοήθεια λεξικού, `dict`.

### Άρνηση

Αν Ρ είναι ένας ισχυρισμός, ο ισχυρισμός **όχι Ρ** ονομάζεται άρνηση του Ρ, συμβολίζεται $\neg P$ και στο SageMath `~P`. Το αποτέλεσμα θα είναι: 

- Αληθής αν Ρ είναι Ψευδής

- Ψευδής αν Ρ Αληθής

```python
expr = propcalc.formula("~P")
expr.evaluate({'P': True}),expr.evaluate({'P': False})
```

### Σύζευξη

Aν $P$ και $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ και Q**, που συμβολίζεται με $P \land Q$ αληθεύει μόνο στην περίπτωση που και οι δύο ισχυρισμοί αληθεύουν. Στο SageMath η έκφραση δημιουργείται με το `&` και το αποτέλεσμα είναι:

- Αληθής αν και το $P$ και το $Q$ είναι Αληθή

- Σε όλες τις υπόλοιπες περιπτώσεις Ψευδής.

Αν ο λογικός τελεστής της σύζευξης εφαρμοστεί σε παραπάνω από δύο τελεστέους, για να προκύψει Αληθές αποτέλεσμα, θα πρέπει να είναι όλοι Αληθείς.

```python
expr = propcalc.formula("a & b")
expr.evaluate({'a': True, 'b': False}),expr.evaluate({'a': True, 'b': True})
```

```python
expr = propcalc.formula("a & b & c & d")
expr.evaluate({'a': True, 'b': True, 'c': True, 'd': False})
```

### Διάζευξη

Aν $P$ ή $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ ή Q**, που συμβολίζεται με $P \lor Q$ και στο SageMath με `|` θα είναι:

- Ψευδής αν και το $P$ και το $Q$ είναι Ψευδή

- Σε όλες τις υπόλοιπες περιπτώσεις Αληθής.

Αν ο λογικός τελεστής της διάζευξης, $\lor$, εφαρμοστεί σε παραπάνω από δύο τελεστέους, για να προκύψει Ψευδές αποτέλεσμα, θα πρέπει να είναι όλοι Ψευδείς.

```python
expr = propcalc.formula("a | b")
expr.evaluate({'a': True, 'b': False}),expr.evaluate({'a': True, 'b': True}),expr.evaluate({'a': False, 'b': False})
```

```python
expr = propcalc.formula("a | b | c | d")
expr.evaluate({'a': False, 'b': False, 'c': False, 'd': True })
```

### Αποκλειστικό ή

Aν $P$ ή $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ αποκλειστικό-ή Q**, που συμβολίζεται με $P \oplus Q$ και στο SageMath με τον τελεστή `^` είναι αληθής όταν μόνο ο ένας από τους δύο ισχυρισμούς είναι αληθής.

- Αληθής αν το $P$ είναι Αληθές και το $Q$ Ψευδές και αντιστρόφως.

- Ψευδής αν και τα δύο είναι Αληθή ή και τα δύο ψευδή.

Αν ο λογικός τελεστής του αποκλειστικού ή εφαρμοστεί σε παραπάνω από δύο τελεστέους, για να προκύψει Αληθές αποτέλεσμα, θα πρέπει περιττός αριθμός τελεστέων να είναι Αληθής.

```python
expr = propcalc.formula("a ^ b")
expr.evaluate({'a': True, 'b': False}),expr.evaluate({'a': False, 'b': False}),expr.evaluate({'a': True, 'b': True})
```

```python
expr = propcalc.formula("a ^ b ^ c")
expr.evaluate({'a': True, 'b': False, 'c': True})
```

### Συνεπαγωγή

Aν $P$ ή $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ συνεπάγεται Q**, $P \implies Q$ που εφαρμόζεται με τον τελεστή `->` είναι Αληθής όταν

- $P$ είναι αληθής και $Q$ αληθής

- $P$ Ψευδής, ανεξάρτητα της τιμής του $Q$

```python
expr = propcalc.formula("a -> b")
expr.evaluate({'a': True, 'b': False}),expr.evaluate({'a': False, 'b': False}),expr.evaluate({'a': True, 'b': True})
```

### Ισοδυναμία

Aν $P$, $Q$ είναι δύο ισχυρισμοί, η σχέση ισοδυναμίας τους, $Ρ \iff Q$, θα είναι αληθής αν και μόνο αν τα $P$ και $Q$ είναι ίσα.

- $P$ αληθές και $Q$ αληθές

- $P$ ψευδές και $Q$ ψευδές

```python
expr = propcalc.formula("a <-> b")
expr.evaluate({'a': True, 'b': False}),expr.evaluate({'a': False, 'b': False}),expr.evaluate({'a': True, 'b': True})
```

### Προτεραιότητα

Όπως και οι αριθμητικές, οι λογικές πράξεις μπορεί να είναι πιο πολύπλοκες. Υπάρχει και για αυτές προτεραιότητα και αν θέλουμε να την παρακάμψουμε χρησιμοποιούμε παρενθέσεις. Στον παραπάνω πίνακα οι πράξεις παρατίθενται με τη σειρά προτεραιότητάς τους.

Στην επόμενη έκφραση η πράξη `&` προηγείται της `|`. Έτσι θα αποτιμηθεί πρώτα η `b & c` και στη συνέχεια η διάζευξη του αποτελέσματος με την _a_.

```python
expr = propcalc.formula("a | b & c")
expr.evaluate({'a': True, 'b': False, 'c': False})
```

Αν θέλουμε να αποτιμηθεί πρώτα η διάζευξη και στην συνέχεια η σύζευξη θα χρησιμοποιήσουμε παρενθέσεις με, όπως αναμένεται, διαφορετικό αποτέλεσμα.

```python
expr = propcalc.formula("(a | b) & c")
expr.evaluate({'a': True, 'b': False, 'c': False})
```

## Πίνακες Αλήθειας

Ο πίνακας αληθείας χρησιμοποιείται στη λογική για να καταγράφει όλες τις δυνατές τιμές μίας έκφρασης για ένα σύνολο λογικών μεταβλητών. Δείχνει το αποτέλεσμα μιας λογικής πράξης για αυτές τις μεταβλητές. Ο πίνακας οργανώνεται με τις γραμμές να αντιστοιχούν σε όλους τους δυνατούς συνδυασμούς τιμών των μεταβλητών.

Στην συνέχεια, πίνακες αλήθειας δύο και τριών λογικών μεταβλητών $A$, $B$ και $C$ για την λογική πράξη της σύζευξης.
$$
\begin{array}{|c|c|c|}
\hline
A & B & A \land B \\
\hline
\text{False} & \text{False} & \text{False} \\
\text{False} & \text{True} & \text{False} \\
\text{True} & \text{False} & \text{False} \\
\text{True} & \text{True} & \text{True} \\
\hline
\end{array}
$$

$$
\begin{array}{|c|c|c|c|}
\hline
A & B & C & A \land B \land C \\
\hline
\text{False} & \text{False} & \text{False} & \text{False} \\
\text{False} & \text{False} & \text{True} & \text{False} \\
\text{False} & \text{True} & \text{False} & \text{False} \\
\text{False} & \text{True} & \text{True} & \text{False} \\
\text{True} & \text{False} & \text{False} & \text{False} \\
\text{True} & \text{False} & \text{True} & \text{False} \\
\text{True} & \text{True} & \text{False} & \text{False} \\
\text{True} & \text{True} & \text{True} & \text{True} \\
\hline
\end{array}
$$

Ο απαιτούμενος αριθμός γραμμών σε έναν πίνακα n μεταβλητών είναι $2^n$ και συνηθίζεται να γράφονται οι τιμές των μεταβλητών σαν να μετράμε στο δυαδικό σύστημα 000 (=0), 001 (=1), 010 (=2) κ.ο.κ. Σημειώνεται ότι όπου 0 εννοείται το False και όπου 1 το True.

Το SageMath δημιουργεί πίνακες αλήθειας μέσω της μεθόδου `truthtable` 

```python
s = propcalc.formula("a&b")
s.truthtable()
```

```python
s = propcalc.formula("~(a & b) | c")
s.truthtable()
```

Στην περίπτωση που δεν θέλουμε να εμφανίσουμε όλες τις γραμμές, όλους δηλαδή τους συνδυασμούς σε έναν πίνακα αλήθειας, οι παράμετροι `start` και `end` καθορίζουν ποιες γραμμές θα εμφανιστούν. Η γραμμή της παραμέτρου `end` δεν θα εμφανιστεί. Επιπλέον, στο επόμενο παράδειγμα η `truthtable` καλείται άμεσα με την δημιουργία της `formula` χωρίς να είναι ανάγκη να επιστραφεί σε μεταβλητή.

```python
propcalc.formula("~(a & b) | c").truthtable(start=3,end=6)
```

## Ικανοποιησιμότητα, ταυτολογία και αντίφαση

Μία πρόταση είναι ικανοποιήσιμη εάν υπάρχει κάποια αποτίμηση, που την ικανοποιεί, που κάνει δηλαδή την πρόταση Αληθή. Η μέθοδος του SageMath που ελέγχει την ικανοποιησιμότητα της έκφρασης είναι η `is_satisfiable`.

Έτσι, η πρόταση $a \land b$ είναι ικανοποιήσιμη. Αν μάλιστα δούμε τον πίνακα αληθείας θα δούμε ότι υπάρχει a και b για τα οποία η πρόταση ισχύει.

```python
f = propcalc.formula("a&b")
f.is_satisfiable()
```

```python
f.truthtable()
```

Αντίθετα, η πρόταση $a \land \lnot a$ είναι μη ικανοποιήσιμη, κάτι που φαίνεται και από τον πίνακα αλήθειας.

```python
f = propcalc.formula("a&~a")
f.is_satisfiable()
```

```python
f.truthtable()
```

Μια πρόταση είναι ταυτολογία αν είναι αληθής για όλες τις δυνατές αναθέσεις τιμών στις μεταβλητές της. Η ταυτολογία ελέγχεται με τη μέθοδο `is_tautology`.

Ελέγχοντας την λογική έκφραση $a \lor b$ βλέπουμε ότι δεν πρόκειται για ταυτολογία. Ελέγχοντας και τον σχετικό πίνακα αλήθειας βλέπουμε ότι σην περίπτωση που και οι δύο μεταβλητές έχουν την τιμή False η πρόταση είναι και αυτή False.

```python
f = propcalc.formula("a|b")
f.is_tautology()
```

```python
f.truthtable()
```

Η λογική πρόταση $a \lor \lnot a$ έχει πάντα την τιμή True ανεξάρτητα της τιμής της a και ως εκ τούτου αποτελεί ταυτολογία.

```python
f = propcalc.formula("a|~a")
f.is_tautology()
```

```python
f.truthtable()
```

Μια λογική έκφραση αποτελεί αντίφαση αν δεν υπάρχει καμία ανάθεση τιμών που να την καθιστά αληθή. Ουσιαστικά όταν η πρόταση δεν είναι ικανοποιήσιμη, είναι αντίφαση. Με την μέθοδο `is_contradiction` ελέγχουμε για αντιφάσεις.

```python
f = propcalc.formula("a&~a")
f.is_contradiction(),f.is_satisfiable()
```

Η μέθοδος `A.is_consequence(B)`, ή εναλλακτικά `is_consequence(A, B)` εξετάζει αν το A είναι μια λογική συνέπεια των υποθέσεων B. Εξετάζει το σύνολο των δυνατών καταστάσεων αληθείας για τις λογικές μεταβλητές και ελέγχει αν κάθε κατάσταση στην οποία η B είναι αληθής, η A είναι επίσης αληθής.

```python
A = propcalc.formula("(P | Q)")
B = propcalc.formula("P")
A.is_consequence(B)
```

```python
A = propcalc.formula("(P & Q)")
B = propcalc.formula("P")
A.is_consequence(B)
```

Οι υποθέσεις, μπορεί να είναι και περισσότερες από μία.

```python
propcalc.formula("P").is_consequence(propcalc.formula("Q"), propcalc.formula("Q -> P"))
```

```python
propcalc.formula("P").is_consequence(propcalc.formula("Q"), propcalc.formula("P -> Q"))
```

## Ισοδυναμία

Δύο εκφράσεις είναι ισοδύναμες αν έχουν ακριβώς τα ίδια αποτελέσματα, True ή False, για όλες τις πιθανές τιμές των μεταβλητών που περιέχουν. Το σύμβολο $\equiv$ δείχνει ότι δύο προτάσεις είναι ισοδύναμες και η μέθοδος `equivalent` ελέγχει την ισοδυναμία στο SageMath.

**Παράδειγμα**

Θα δείξουμε ότι οι προτάσεις $(p \land q) \lor (\neg p \land q)$ και $q$ είναι ισοδύναμες με χρήση της `equivalent` αλλά και πίνακα αληθείας.
$$(p \land q) \lor (\neg p \land q) \equiv q$$.

```python
lhs = propcalc.formula("(p & q) | (~p & q)")
rhs = propcalc.formula("q")

lhs.equivalent(rhs)
```

Παρατηρούμε στον επόμενο πίνακα αληθείας ότι οι στήλες αποτίμησης της $(p \land q) \lor (\neg p \land q)$ και η στήλη q ταυτίζονται.

```python
lhs.truthtable()
```

## Ιδιότητες

1. Προσεταιριστική ιδιότητα:
    - $(A \lor B) \lor C \equiv A \lor (B \lor C)$
    
    - $(A \land B) \land C \equiv A \land (B \land C)$

```python
lhs = propcalc.formula("(A | (B | C))")
rhs = propcalc.formula("((A | B) | C)")
lhs.equivalent(rhs)
```

```python
lhs = propcalc.formula("(A & (B & C))")
rhs = propcalc.formula("((A & B) & C)")
lhs.equivalent(rhs)
```

2. Μεταθετική ιδιότητα:

    - $A \lor B \equiv B \lor A$
    
    - $A \land B \equiv B \land A$

```python
lhs = propcalc.formula("(A | B)")
rhs = propcalc.formula("(B | A)")
lhs.equivalent(rhs)
```

```python
lhs = propcalc.formula("(A & B)")
rhs = propcalc.formula("(B & A)")
lhs.equivalent(rhs)
```

3. Επιμεριστική ιδιότητα:

    - $A \land (B \lor C) \equiv (A \land B) \lor (A \land C)$

    - $A \lor (B \land C) \equiv (A \lor B) \land (A \lor C)$

```python
lhs = propcalc.formula("(A & (B | C))")
rhs = propcalc.formula("((A & B) | (A & C))")
lhs.equivalent(rhs)
```

```python
lhs = propcalc.formula("(A | (B & C))")
rhs = propcalc.formula("((A | B) & (A | C))")
lhs.equivalent(rhs)
```

4. Αυτοπαθής ιδιότητα:

    - $A \lor A \equiv A$ 
 
    - $A \land A \equiv A$ 

```python
lhs = propcalc.formula("A | A")
rhs = propcalc.formula("A")
lhs.equivalent(rhs)
```

5. Ταυτοτική ιδιότητα:

    - $A \lor \text{False} \equiv A$

    - $A \land \text{True} \equiv A$
    
Σημείωση: Τα False, True θεωρούνται μεταβλητές και δεν αναγνωρίζονται ως σταθερές. Έτσι η `equivalent` δεν λειτουεργεί όπως αναμενόταν επιστρέφοντας False αντί για True.

```python
lhs = propcalc.formula("(A | False)")
rhs = propcalc.formula("A")
lhs.equivalent(rhs)
```

```python
lhs = propcalc.formula("(A & True)")
rhs = propcalc.formula("A")
lhs.equivalent(rhs)
```

6. Νόμος του De Morgan (De Morgan's Law): περιγράφουν την σχέση μεταξύ των λογικών πράξεων άρνησης, διάζευξης, και σύζευξης.

    - $\neg (A \lor B) \equiv (\neg A) \land (\neg B)$

    - $\neg (A \land B) \equiv (\neg A) \lor (\neg B)$

```python
lhs = propcalc.formula("~(A | B)")
rhs = propcalc.formula("(~A & ~B)")
lhs.equivalent(rhs)
```

```python
lhs = propcalc.formula("~(A & B)")
rhs = propcalc.formula("(~A | ~B)")
lhs.equivalent(rhs)
```

Όπως γίνεται αντιληπτό οι πίνακες αλήθειας για τις στήλες δύο οποιωνδήποτε ισοδυναμιών θα ταυτίζονται.

```python
lhs = propcalc.formula("~(A & B)")
lhs.truthtable()
```

```python
rhs = propcalc.formula("(~A | ~B)")
rhs.truthtable()
```

## Σχέσεις

| **Σχέση Συνόλων**       | **Περιγραφή**                                 | **SageMath Συνάρτηση** |
|-------------------------|-----------------------------------------------|------------------------|
| Υποσύνολο, Subset       | Έλεγχος αν το A είναι υποσύνολο του B         | `A.issubset(B)`, `A<=B`       |
| Γνήσιο υποσύνολο, Proper subset       | Έλεγχος αν το A είναι γνήσιο υποσύνολο του B         | `A<B`       |
| Υπερσύνολο, Superset    | Έλεγχος αν το A είναι υπερσύνολο του B        | `A.issuperset(B)`, `A>=B`   |
| Ισότητα Συνόλων          | Έλεγχος αν τα δύο σύνολα είναι ίσα            | `A == B`               |
| Μη ισότητα Συνόλων       | Έλεγχος αν τα δύο σύνολα δεν είναι ίσα        | `A != B`               |

Ένα σύνολο καθορίζεται πλήρως από τα στοιχεία του, **αξίωμα της επέκτασης**, με την σειρά με την οποία γράφονται να μην έχει σημασία. Δύο σύνολα A και B είναι ίσα αν και μόνο αν κάθε στοιχείου του Α είναι στο B και κάθε στοιχείο του B είναι στο A.

**Παράδειγμα**

Θα υποθέσουμε ότι οι Πέτρος, Δημήτρης, Χρόνης είναι φοιτητές που παρακολουθούν το μάθημα του Μαθηματικού Λογισμικού. Με όποια σειρά και να θέσουμε σε ένα σύνολο τα παραπάνω ονόματα όλα συμβολίζουν το ίδιο σύνολο.

```python
students1 = {'Πέτρος','Δημήτρης','Χρόνης'}
students2 = {'Δημήτρης','Πέτρος','Χρόνης'}
students3 = {'Δημήτρης','Χρόνης','Πέτρος'}
students1==students2,students1==students3,students2==students2
```

Ένα σύνολο $A$ ονομάζεται υποσύνολο ενός συνόλου $B$, $A \subseteq B$, αν όλα τα στοιχεία του $A$ είναι επίσης στοιχεία του $B$.

```python
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}
C = {1, 2, 3}

A.issubset(B),A.issubset(C)
```

Το σύνολο $A$ αποτελεί υπερσύνολο του $B$ αν κάθε στοιχείο του $Β$ ανήκει στο $Α$.

```python
A.issuperset(B),A>=B
```

Γνήσιο υποσύνολο του συνόλου $Β$ θα ονομάζεται το $Α$ , $A \subset B$, όταν κάθε στοιχείο του $A$ είναι και στοιχείο του $B$, αλλά το $A$ δεν είναι ίσο με το $B$.

Κάτι τέτοιο μπορεί να υπολογιστεί και με χρήση λογικής έκφρασης.

```python
A.issubset(B) and A != B,A.issubset(C) and A != C
```

Οι συγκριτικοί τελεστές `<` και `<=` έχουν υπερφορτωθεί για να ελέγχουν σύνολα και γνήσια υποσύνολα αντίστοιχα.

```python
A<C,A<=C
```

## Πράξεις

Όλες οι γνωστές πράξεις συνόλων υλοποιούνται είτε με αντίστοιχες μεθόδους είτε με υπερφορτωμένους τελεστές.

| **Πράξη Συνόλου**       | **Περιγραφή**                                 | **SageMath Συνάρτηση** |
|-------------------------|-----------------------------------------------|------------------------|
| Μέλος, Membership       | Έλεγχος αν ένα στοιχείο ανήκει σε σύνολο      | x `in` A               |
| Μη μέλος, Non-membership| Έλεγχος αν ένα στοιχείο δεν ανήκει σε σύνολο  | x `not in` A           |
| Ένωση, Union            | Ενώνει όλα τα στοιχεία από δύο σύνολα         | `A.union(B)`           |
| Τομή, Intersection      | Κοινά στοιχεία μεταξύ δύο συνόλων             | `A.intersection(B)`    |
| Διαφορά, Difference     | Στοιχεία στο A που δεν ανήκουν στο B          | `A.difference(B)`      |
| Συμμετρική διαφορά, Symmetric Difference | Στοιχεία σε ένα και μόνο ένα από τα δύο σύνολα | `A.symmetric_difference(B)` |

Οι τελεστές `in`, `not in` μας επιτρέπουν να ελέγξουμε αν ένα στοιχείο είναι μέλος του συνόλου.

```python
students = {'Πέτρος','Δημήτρης','Χρόνης'}
'Χριστίνα' in students, 'Δημήτρης' in students,'Χριστίνα' not in students,
```

```python
A = Set([1, 2, 3, 4])
B = Set([3, 4, 5, 6])
A, B
```

Η ένωση των συνόλων A και B $A \cup B = \{x \mid x \in A \text{ ή } x \in B\}$ μπορεί να γίνει με τη μέθοδο `union` ή τον τελεστή `|` που έχει υπερφορτωθεί, **overload** για σύνολα.

```python
A.union(B);A|B
```

Η τομή των συνόλων A και B $A \cap B = \{x \mid x \in A \text{ και } x \in B\}$ υλοποιείται με την `intersection` ή τον τελεστή `&`.

```python
A.intersection(B);A&B
```

Η διαφορά δύο συνόλων $A \setminus B = \{x \mid x \in A \text{ και } x \notin B\}$ με την `difference` ή τον τελεστή `-`.

```python
A.difference(B),A-B
```

Η συμμετρική διαφορά $A \triangle B = (A \cup B) \setminus (A \cap B)$, με την `symmetric_difference` και τον τελεστή `^^`.

```python
A.symmetric_difference(B),B.symmetric_difference(A),A^^B,B^^A
```

```python
(s1|s2)==s1,s2<=s1
```

## Διωνυμικός συντελεστής

Ο **διωνυμικός συντελεστής**, $\binom{n}{k}$, δίνει τον αριθμό των τρόπων επιλογής _k_ αντικειμένων από ένα σύνολο _n_ αντικειμένων, χωρίς να λαμβάνεται υπόψη η σειρά):

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

Για παράδειγμα, για να επιλέξουμε 3 αντικείμενα από 5;
$$
\binom{5}{3} = \frac{5!}{3!(5-3)!} = \frac{5 \cdot 4 \cdot 3!}{3! \cdot 2!} = \frac{5 \cdot 4}{2 \cdot 1} = 10
$$


Ο υπολογισμός του $\binom{n}{k}$ στο SageMath γίνεται με την `binomial`.

```python
binomial(5, 2)
```

Το παραπάνω μπορεί να βρεθεί και με πράξεις συνόλων.

```python
S = Subsets([1,2,3,4,5], 2); S.cardinality()
```

```python
# Δημιουργία γραφήματος
G = graphs.CompleteGraph(6)  # Πλήρης γράφος με 6 κορυφές

# Ορισμός χρωμάτων ακμών
edge_colors = {
    "red": [(0, 1), (2, 3)],
    "blue": [(1, 2), (4, 5)],
    "green": [(3, 4), (0, 5)]
}

# Σχεδίαση γραφήματος
G.show(edge_colors=edge_colors)
```
