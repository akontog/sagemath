# Τελεστές Operators

Το Sage μπορεί να χρησιμοποιηθεί σαν μία αριθμομηχανή με τις εκφράσεις να μπορούν να γίνουν όσο πολύπλοκες θέλουμε.

```python
(7+5)*(9-6)+8-2
```

Παρατηρούμε ότι η πράξη του πολλαπλασιασμού τίθεται με το σύμβολο `*`, της διαίρεησης με το `/` και σε περίπτωση που θέλουμε διαφορετική προτεραιότητα από την εξορισμού χρησιμοποιούμε αντίστοιχες παρενθέσεις. Η προτεραιότητα είναι αυτή που περιμένουμε από τα μαθηματικά και βλέπουμε στον επόμενο πίνακα.

| Τελεστής  | Προσεταιριστικότητα | Πράξη |
|:---------:|:------------:|:-------:|
| `**` ή `^` | ← | ύψωση σε δύναμη (`2**3 = 8`) |
| `*`, `/`, `//`, `%`  | → | Πολλαπλασιασμός, Διαίρεση, Ακέραια Διαίρεση, Υπόλοιπο (`2*3 = 6`, `6/2 = 3`, `5//2 = 2`, `5%2 = 1`) |
| `+`, `-` | → | Πρόσθεση, Αφαίρεση (`2 + 3 = 5`, `5 - 2 = 3`) |

Η προσεταιριστικότητα (ή αλλιώς associativity) αναφέρεται στο πώς ομαδοποιούνται οι πράξεις όταν υπάρχουν πολλαπλοί τελεστές του ίδιου τύπου. Ειδικά για τον τελεστή της δύναμης (`**` ή `^`), είναι αριστερή (←), δηλαδή, οι πράξεις ομαδοποιούνται από τα δεξιά προς τα αριστερά. 
 
Στο επόμενο παράδειγμα, πρώτα θα υπολογιστεί η δεξιά δύναμη, `3^2 = 9` και έπειτα η `2^9=512`. 

```python
2^3^2
```

Προφανώς αν επιθμούμε διαφορετική σειρά εκτέλεσης πράξεων κάνουμε χρήση παρενθέσεων.

```python
(2^3)^2
```

# Αριθμητική προσέγγιση n()

Εάν επιθυμούμε την αριθμητική τιμή, στις περιπτώσεις που οι τελεστέοι είναι ακέραιοι, θα πρέπει να κάνουμε χρήση της `N()` (ή `n()` ή `numerical_approx()`) που χρησιμοποιείται για να προσεγγίσει αριθμητικά μία έκφραση.

```python
N((5 - 2)^4 / (4 + (5 + 3)^3))
```

Η προαιρετική παράμετρος `digits` θέτει τον αριθμό των σημαντικών ψηφίων που θα εμφανιστούν.

```python
N((5 - 2)^4 / (4 + (5 + 3)^3),digits=5)
```

## Απλές μεταβλητές

## Συμβολικές μεταβλητές

Μια συμβολική έκφραση στο SageMath είναι κάθε αλγεβρική ή μαθηματική έκφραση που περιλαμβάνει μεταβλητές, αριθμούς και λειτουργίες που δεν έχουν αναγκαστικά συγκεκριμένες τιμές από την αρχή, σε αντίθεση με τις απλές μεταβλητές. Οι συμβολικές μεταβλητές θα πρέπει να δηλωθούν με τον επόμενο τρόπο πριν χρησιμοποιηθούν (SR -> Symbolic Ring).

Στην συνέχεια ορίζεται το `w` ώς συμβολική μεταβλητή. Πρέπει να σημειωθεί ότι η `x`, καθώς και σε πολλά περιβάλλοντα οι `y`, `z`, είναι εξορισμού συμβολικές χωρίς να απαιτείται να δηλωθεί από εμάς. 

```python
w = SR.var('w')
show(w**2 + 5)
```

```python
f = (w^2 - 1) / (w - 1);show(f)
```

```python
for i in range(10): print(i,end=" ")
```

```python
for i in range(5,105,5): print(i,end=" + ")
```
