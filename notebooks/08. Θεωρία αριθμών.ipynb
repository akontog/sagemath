Στην συνέχεια θα επαληθεύσουμε τα παραπάνω αποτελέσματα χρησιμοποιώντας:

- Τις ενσωματωμένες μεθόδους του SageMath `lcm` και `gcd`

- με χρήση της `factor` και της παραπάνω επαναληπτικής διαδικασίας.

```python
a = 108
b = 90
afactors = {p:e for p,e in factor(a)}
bfactors = {p:e for p,e in factor(b)}
mygcd = 1
for p in afactors:
    if p in bfactors:
        min_exp = min(afactors[p], bfactors[p])
        mygcd *= p ** min_exp


mylcm = 1
all_factors = set(afactors.keys()).union(bfactors.keys())
for p in all_factors:
    max_exp = max(afactors.get(p, 0), bfactors.get(p, 0))
    mylcm *= p ** max_exp
lcm(a, b),mylcm, gcd(a, b),mygcd
```

## Υπολογισμός μεγάλου εκθέτη

Η έκφραση $x^y mod n$ είναι πολύ χρήσιμη στην κρυπτογραφία και μεθόδους όπως ο RSA. Η προφανής λύση υπολογισμού της δύναμης $x^y$ και στη συνέχεια ο υπολογισμός του υπολοίπου δεν είναι αποδοτικός για μεγάλους αριθμούς. Η `power_mod(a, n, m)` επιστρέφει την n-οστη δύναμη του a modulo n 

```python
power_mod(2,5,6),2^5%6
```

Ο αλγόριθμος που ακολουθεί η παραπάνω μέθοδος είναι η ύψωση σε δύναμη δια του τετραγωνισμού, **exponentiation by squaring** που έχουμε μελετήσει σε άσκηση βρίσκοντάς την αναδρομικά. 

```python
def my_power_mod(b, e, m):
    r = 1  # Υπόλοιπο
    while e > 0:
        # Αν είναι περιττός
        if e % 2 == 1:  
            r = (r * b) % m
        b = (b * b) % m
        # Ακέραια διαίρεση με το 2
        e //= 2  
    return r
my_power_mod(2,5,6)
```

```python
import time
b = 2
e = 10^8  # Μεγάλος εκθέτης
m = 987654321
start_time = time.time()
my_result = my_power_mod(b, e, m)
end_time = time.time()
print(f"Αποτέλεσμα (my_power_mod): {my_result}")
print(f"Χρόνος εκτέλεσης (my_power_mod): {end_time - start_time:.6f} δευτερόλεπτα")

# Χρονομέτρηση για την ενσωματωμένη power_mod
start_time = time.time()
sage_result = power_mod(b, e, m)
end_time = time.time()
print(f"Αποτέλεσμα (power_mod): {sage_result}")
print(f"Χρόνος εκτέλεσης (power_mod): {end_time - start_time:.6f} δευτερόλεπτα")

start_time = time.time()
sage_result = b^e%m
end_time = time.time()
print(f"Αποτέλεσμα (power_mod): {sage_result}")
print(f"Χρόνος εκτέλεσης (power_mod): {end_time - start_time:.6f} δευτερόλεπτα")
```

## bit ισοτιμίας

Το bit ισοτιμίας, parity bit, αποτελεί τον απλούστερο μηχανισμό ανίχνευσης σφαλμάτων. Βασίζεται στο πλήθος των 1s σε ένα δυαδικό μήνυμα που μπορεί να είναι περιττός ή άρτιος αριθμός.

**Παράδειγμα**

Σε ένα δυαδικό μήνυμα, έστω _1011_ θα μετρήσουμε τον αριθμο των 1 και ανάλογα θα προσθέσουμε ένα bit στο τέλος έτσι ώστε ο συνολικό αριθμός 1 να είναι άρτιος. Στην συνέχεια θα ελέγξουμε αν υσχύει η άρτια ισοτιμία. Τέλος, θα "πειράξουμε" 1 bit και θα γίνει και πάλι έλεγχος ισοτιμίας.

```python
# Παραδείγματα μηνυμάτων
msg1 = vector(GF(2), [1, 0, 1, 1])
msg2 = vector(GF(2), [1, 1, 0, 0])
# Υπολογισμός bit ισοτιμίας
parityBit1 = sum(msg1) % 2
parityBit2 = sum(msg2) % 2
msg1,parityBit1,msg2,parityBit2
```

```python
# Προσθήκη του bit ισοτιμίας στο μήνυμα
msg1WithParity = msg1.list() + [parityBit1]
msg2WithParity = msg2.list() + [parityBit2]
msg1WithParity,msg2WithParity
```

```python
# έλεγχος αν τα μηνύματα έχουν σωστή ισοτιμία
sum(msg1WithParity) % 2 == 0, sum(msg1WithParity) % 2 == 0
```

## Κώδικας Hamming

Ο κώδικας Hamming μπορεί να διορθώσει σφάλματα ενός bit και να ανιχνεύσει την παρουσία σφαλμάτων δύο bit σε ένα μπλοκ δεδομένων. Χρησιμοποιεί και αυτός έναν μηχανισμό ισοτιμίας μπλοκ με τα δεδομένα να χωρίζονται σε μπλοκ. 

Η ποσότητα των δεδομένων ισοτιμίας που προστίθεται στον κώδικα Hamming δίνεται από τον τύπο $2^p \geq d + p + 1$, όπου p είναι ο αριθμός των bit ισοτιμίας και d είναι ο αριθμός των bit δεδομένων. Για παράδειγμα, εάν θέλετε να μεταδώσετε 7 bit δεδομένων, ο τύπος θα ήταν $2^4 \geq 7 + 4 + 1$, επομένως απαιτούνται 4 bit ισοτιμίας.

## RSA

Ο αλγόριθμος RSA βασίζεται στην επιλογή δύο μεγάλων πρώτων αριθμών για τη δημιουργία των κλειδιών και την εφαρμογή της θεωρίας αριθμών για την κρυπτογράφηση. Βήματα υλοποίησης:

1. **Δημιουργία κλειδιών**:Επιλέγουμε δύο (μεγάλους) πρώτους αριθμούς

2. Υπολογίζουμε το γινόμενό τους $n=p×q$, το οποίο θα χρησιμοποιηθεί ως μέρος του δημόσιου κλειδιού.

3. Υπολογίζουμε την $φ(n) = (p-1)(q-1)$

4. Επιλέγουμε έναν αριθμό e που είναι αμοιβαία πρώτος προς ϕ(n).

5. Υπολογίζουμε τον αντίστροφο του e ως προς ϕ(n), δηλαδή το d, το οποίο θα χρησιμοποιηθεί ως το ιδιωτικό κλειδί.

6. **Κρυπτογράφηση**: Μετατρέπουμε το μήνυμα σε αριθμητική μορφή και υπολογίζουμε $cipher=message^e mod n$.

7. **Αποκρυπτογράφηση**: Χρησιμοποιούμε το ιδιωτικό κλειδί για να ανακτήσουμε το αρχικό μήνυμα με τον τύπο $message=cipher^d mod n$.

```python
# Βήμα 1: Δημιουργία Κλειδιών
p = random_prime(10^5, lbound=10^4)
q = random_prime(10^5, lbound=10^4)
n = p * q
phi_n = (p - 1) * (q - 1)

# Επιλογή e ώστε gcd(e, phi_n) = 1
e = randint(2, phi_n - 1)
while gcd(e, phi_n) != 1:
    e = randint(2, phi_n - 1)

# Υπολογισμός του d, ιδιωτικό κλειδί
d = inverse_mod(e, phi_n)

# Δημόσιο και ιδιωτικό κλειδί
public_key = (e, n)
private_key = (d, n)

# Βήμα 2: Κρυπτογράφηση
message = 12345  # Το μήνυμα που θέλουμε να κρυπτογραφήσουμε
cipher = power_mod(message, e, n)  # message^e mod n

# Βήμα 3: Αποκρυπτογράφηση
decrypted_message = power_mod(cipher, d, n)  # cipher^d mod n

# Εμφάνιση Αποτελεσμάτων
print("Δημόσιο Κλειδί:", public_key)
print("Ιδιωτικό Κλειδί:", private_key)
print("Αρχικό Μήνυμα:", message)
print("Κρυπτογραφημένο Μήνυμα:", cipher)
print("Αποκρυπτογραφημένο Μήνυμα:", decrypted_message)
```
