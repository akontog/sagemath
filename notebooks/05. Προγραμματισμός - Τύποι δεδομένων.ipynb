# Συλλογές

## Συμβολοσειρές, Strings

Η συμβολοσειρά αποτελεί μια συλλογή χαρακτήρων που διατάσσονται σε μια συγκεκριμένη σειρά. Κάθε χαρακτήρας σε αυτήν τη συλλογή έχει έναν δείκτη (index). Είναι αμετάβλητες (immutable), δηλαδή δεν μπορούν να τροποποιηθούν αφού δημιουργηθούν. Δηλώνονται είτε με μονά, `'` είτε με διπλά, `"`, εισαγωγικά.

```python
s = 'SageMath'
```

Οι συμβολοσειρές, όπως και οι λίστες και οι πλειάδες που θα δουμε στη συνέχεια αλλά και τα διανύσματα και οι πίνακες στην ενότητα της γραμμικής άλγεβρας, επιτρέπουν την προσπέλαση των στοιχείων τους με τη χρήση αριθμόδεικτών (index). Ο δείκτης προσδιορίζει τη θέση του στοιχείου μέσα στην ακολουθία. Η αρίθμηση μπορεί να ξεκινά είτε από τα αριστερά και το 0 είτε από τα δεξιά με τον αριθμοδείκτη -1 να παριστάνει το τελευταίο στοιχείο.

|  0  |  1  |  2  |  3  |  4  |  5  |  6  |  7  |
| --- | --- | --- | --- | --- | --- | --- | --- |
|  S  |  a  |  g  |  e  |  M  |  a  |  t  |  h  |
| -8  | -7  | -6  | -5  | -4  | -3  | -2  | -1  |

```python
s[0], s[-1]
```

Με την δομή `for` έχουμε την δυνατότητα να προσπελάσουμε μία συμβολοσειρά.

```python
for c in s: print(c,end='|')
```

Πέρα από την χρήση δεικτών για την προσπέλαση ενός στοιχείου της συμβολοσειράς (ή λίστας, πλειάδας, διανύσματος, πίνακα) μπορούμε να χρησιμοποιήσουμε την τεχνική τεμαχισμού, slicing για να εξαγάγουμε ένα υποσύνολο στοιχείων.
Υλοποιείται με τη σύνταξη: start:stop:step.
- **start**: Ο δείκτης (index) από τον οποίο ξεκινάει το slicing (περιλαμβάνεται). Αν δεν υπάρχει, υπονοείται το 0.
- **stop**: Ο δείκτης στον οποίο σταματάει το slicing (δεν περιλαμβάνεται). Αν δεν υπάρχει, υπονοείται το τέλος.
- **step**:Το βήμα με το οποίο προχωράει το slicing. Αν δεν υπάρχει, υπονοείται το 1.

```python
s[1:4],s[:4],s[4:],s[::2],s[::-1]
```

Οι συμβολοσειρές, σε αντίθεση με τις λίστες ή τα σύνολα, δεν μπορούν να τροποποιηθούν απευθείας. Είναι όπως λέμε **immutable** δομή.

```python
s[0]='s'
```

Υποστηρίζονται μία σειρά από συναρτήσεις και μεθόδους για το string, όπως η συνάρτηση επιστροφής του μήκους του, `len` ή οι μέθοδοι εύρεσης, `find` και καταμέτρησης `count`.

```python
len(s), s.find('a'),s.find('w'),s.count('a')
```

Οι εντολές `in` και `not in` ελέγχουν αν μία συμβολοσειρά υπάρχει σε μία άλλη.

```python
show('Math' in 'SageMath')
show('math' in 'SageMath')
```

## Λίστες (lists)

Αποτελούν συλλογές που μπορούν να περιέχουν διατεταγμένα, **ordered**, διάφορους τύπους δεδομένων, δεικτοδοτούνται, **indexed**, και μεταβάλλονται, **mutable**.

```python
# Κενή λίστα
L = []
show(type(L))
L = [10, 20, 30]
show(L)
```

Μπορεί να περιέχει διάφορα είδη δεδομένων αλλά και άλλες λίστες

```python
L = [10, 20, 30, 'SageMath', 3.14, [1,2,3]];L
```

Η προσπέλαση και ο τεμαχισμός γίνεται στην ίδια λογική με τις συμβολοσειρές.

```python
L[0],L[-1],L[2:4]
```

Εφόσον η λίστα μπορεί να περιέχει και αντικείμενα που είναι δεικτοδοτούμενα και τα ίδια, μπορούμε να προσπελάσουμε και στοιχεία αυτών.

```python
L[-1][1], L[3][:4]
```

Η λίστα αποτελεί δυναμική δομή με το μέγεθός της να αυξομειώνεται. Παρέχονται διάφορες μέθοδοι για προσθήκη ή διαγραφή στοιχείων από λίστα. 
- `append` Προσθήκη στο τέλος της λίστας.
- `insert` Εισαγωγή οπουδήποτε στη λίστα.
- `pop` Αφαίρεση στοιχείου από το τέλος ή από οπουδήποτε

```python
L = [10, 20, 30, 40, 50]
show(L)
L.append(60)  # προσθήκη του 60 στο τέλος
show(L)
L.insert(2,100) # προσθήκη του 100 στη θέση 2
show(L)
L.pop() # Αφαίρεση του τελευταίου στοιχείου
show(L)
L.pop(1) # Αφαίρεση του στοιχείου 1
show(L)
```

Μπορούμε, επιπλέον, να προσθέσουμε στοιχεία σε μία λίστα χρησιμοποιώντας τον τελεστή της πρόσθεσης που όπως λέμε έχει υπερφορτωθεί, **overloaded**, για να προσθέτει λίστες αλλά ακόμη και να πολλαπλασιάσουμε μία λίστα με τον αντίστοιχο τελεστή.

```python
L1 = [1,2,3];L2=[4,5,6]
L1+L2,L1*3
```

```python
L = [[1,2,3],[4,5,6],[7,8,9]]
```

Η συνάρτηση `table` εμφανίζει την δυσδιάστατη λίστα σαν πίνακα.

```python
table(L)
```

## Σύνολα (sets)

Σε αντίθεση με τις λίστες, τα σύνολα τηρούν την πληροφορία αν περιέχουν ένα στοιχείο ή όχι. Επιπλέον, όπως και τα μαθηματικά σύνολα δεν δέχονται διπλότυπα. Δηλώνονται με `{}` αντί για `[]` ή με την `Set`.

Όλες οι γνωστές πράξεις συνόλων υλοποιούνται είτε με αντίστοιχες μεθόδους είτε με τελεστές.

| **Πράξη Συνόλου**       | **Περιγραφή**                                 | **SageMath Συνάρτηση** |
|-------------------------|-----------------------------------------------|------------------------|
| Ένωση (Union)            | Ενώνει όλα τα στοιχεία από δύο σύνολα         | `A.union(B)`           |
| Τομή (Intersection)      | Κοινά στοιχεία μεταξύ δύο συνόλων             | `A.intersection(B)`    |
| Διαφορά (Difference)     | Στοιχεία στο A που δεν ανήκουν στο B          | `A.difference(B)`      |
| Συμμετρική διαφορά (Symmetric Difference) | Στοιχεία σε ένα από τα δύο σύνολα, αλλά όχι και στα δύο | `A.symmetric_difference(B)` |
| Υποσύνολο (Subset)       | Έλεγχος αν το A είναι υποσύνολο του B         | `A.is_subset(B)`       |
| Υπερσύνολο (Superset)    | Έλεγχος αν το A είναι υπερσύνολο του B        | `A.is_superset(B)`     |
| Μέλος (Membership)       | Έλεγχος αν ένα στοιχείο ανήκει σε σύνολο      | `x in A`               |
| Μη μέλος (Non-membership)| Έλεγχος αν ένα στοιχείο δεν ανήκει σε σύνολο  | `x not in A`           |
| Ισότητα Συνόλων          | Έλεγχος αν τα δύο σύνολα είναι ίσα            | `A == B`               |
| Μη ισότητα Συνόλων       | Έλεγχος αν τα δύο σύνολα δεν είναι ίσα        | `A != B`               |

```python
A = Set([1, 2, 3, 4])
B = Set([3, 4, 5, 6])
A, B
```

```python
A.union(B);A+B
```

# Comprehension

Μία πολύ χρήσιμη λειτουργία είναι η δημιουργία των παραπάνω τύπων δεδομένων, `list`, `tuple`, `dict`, `set` με συμπυκνωμένο τρόπο χρησιμοποιώντας comprehension.

```python
squares = [x**2 for x in range(10)]
squares
```

```python
sins = tuple(n(sin(i*pi/16)) for i in range(10))
sins
```

```python
sqrts = {x:n(sqrt(x)) for x in range(10)}
sqrts
```

```python
triples = {3*x for x in range(10)}
triples
```
