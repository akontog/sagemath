Η συμβολοσειρά, `str`, αποτελεί μια συλλογή χαρακτήρων που διατάσσονται σε μια συγκεκριμένη σειρά. Κάθε χαρακτήρας έχει έναν δείκτη, **indexed**. Είναι αμετάβλητες, **immutable**, δηλαδή δεν μπορούν να τροποποιηθούν αφού δημιουργηθούν. Δηλώνονται είτε με μονά, `'` είτε με διπλά, `"`, εισαγωγικά.

Οι συμβολοσειρές, όπως και οι λίστες και οι πλειάδες που θα δουμε στη συνέχεια αλλά και τα διανύσματα και οι πίνακες που θα μελετήσουμε στην ενότητα της γραμμικής άλγεβρας, επιτρέπουν την προσπέλαση των στοιχείων τους με τη χρήση αριθμόδεικτών (index). Ο δείκτης προσδιορίζει τη θέση του στοιχείου μέσα στην ακολουθία. Η αρίθμηση μπορεί να ξεκινά είτε από τα αριστερά και το 0 είτε από τα δεξιά με τον αριθμοδείκτη -1 να υποδυκνύει το τελευταίο στοιχείο.

Πέρα από την χρήση δεικτών για την προσπέλαση ενός στοιχείου της συμβολοσειράς (ή λίστας, πλειάδας, διανύσματος, πίνακα) μπορούμε να χρησιμοποιήσουμε την τεχνική **slicing** για να εξαγάγουμε ένα υποσύνολο στοιχείων.
Υλοποιείται με τη σύνταξη: start:stop:step.
- **start**: Ο δείκτης (index) από τον οποίο ξεκινάει το slicing (περιλαμβάνεται). Αν δεν υπάρχει, υπονοείται το 0.
- **stop**: Ο δείκτης στον οποίο σταματάει το slicing (δεν περιλαμβάνεται). Αν δεν υπάρχει, υπονοείται το τέλος της συμβολοσειράς.
- **step**:Το βήμα με το οποίο προχωράει το slicing. Αν δεν υπάρχει, υπονοείται το 1.

```python
s[1:4],s[:4],s[4:],s[::2],s[::-1]
```

Με την βοήθεια της `srange` θα δημιουργήσουμε μία λίστα με _nsides_ γωνίες από 0 έως 2π. Μία αρχικά άδεια λίστα σημείων θα γεμίσει με την βοήθεια της `append` με σημεία που βρίσκονται πάνω σε έναν κύκλο με κέντρο την αρχή των αξόνων και ακτίνα _r_. Στην συνέχεια με την `list_plot` και την παράμετρο `plotjoined=True` θα δημιουργήσουμε εγγεργραμμένα πολύγωνα στον κύκλο. Μπορούμε να σχεδιάσουμε και τον κύκλο.

```python
r = 1
nsides = 10
angles = srange(0, 2*pi, 2*pi/nsides)
mypoints = []
for angle in angles:
    xpoint = r * cos(angle)
    ypoint = r * sin(angle)
    mypoints.append((xpoint, ypoint))
mypoints.append(mypoints[0])
polygon_plot = list_plot(mypoints, plotjoined=True, color='blue', thickness=1.5)
circle_plot = circle((0, 0), r, color='red', thickness=1.2)
show(polygon_plot + circle_plot)
```

**Παράδειγμα**

Θα δημιουργήσουμε ένα γραφικό εκμεταλευόμενοι την `srange`.

```python
g = Graphics()
step = 6
offset = 0.2  # μετατόπιση
paths = 16 # αριθμός κύκλων για κάθε ακτίνα
PI = math.pi
for r in srange(1, paths+1):
    for n in srange(0, 2*PI + PI/step, PI/step):
        x = (r + offset) * math.cos(n)
        y = (r + offset) * math.sin(n)
        g += circle((x, y), offset, color=(r/paths,1 - r / paths,0.5))
    rnext = (r + 1) ** 2
    offset = (rnext - r) - offset
g.show(axes=False)
```

**Παράδειγμα**

Θα δημιουργήσουμε και θα σχεδιάσουμε ένα πλέγμα διαστάσεων n×n, όπου κάθε θέση του πλέγματος έχει 50% πιθανότητα να περιέχει ένα σημείο. Το πλέγμα και τα σημεία θα σχεδιαστούν με τη βοήθεια μιας λίστας συντεταγμένων για τις θέσεις που περιέχουν σημεία.

```python
size = 10
probability = 0.5
points = []
for i in range(size+1):
    for j in range(size+1):
        if random() < probability:
            points.append((i, j))
g = Graphics()
for (x, y) in points:
    g += point((x, y), size=30, color='blue')
for i in range(size + 1):
    g += line([(i, 0), (i, size)], color='gray')
for j in range(size + 1):
    g += line([(0, j), (size, j)], color='gray')
g.show()
```

## Πλειάδες (tuples)

Οι πλειάδες μοιάζουν με τις λίστες με μια σημαντική διαφορά, είναι αμετάβλητες, **immutable**. Αυτό σημαίνει ότι τα στοιχεία τους δεν μπορούν να αλλαχτούν μετά τη δημιουργία τους, κάτι που τις κάνει χρήσιμες όταν θέλουμε να είμαστε βέβαιοι ότι τα δεδομένα δεν θα μεταβληθούν από λάθος σε μια λειτουργία ή υπολογισμό. 

Μία συχνή χρήση των tuples στο SageMath είναι η αναπαράσταση σημείων στον χώρο.

Οι πλειάδες δημιουργούνται είτε με χρήση παρνεθέσεων είτε με την `tuple`.

```python
t1 = (1,2,3)
t2 = tuple([4,5,6,7])
t1,t2
```

```python
def isFunction(relation):
    g = Graphics()
    for x, y in relation:
        g += point((0, x), size=20, color='blue')
        g += text('x='+str(x), (0, x), color='blue', fontsize=8, horizontal_alignment='right')
        g += point((1, y), size=20, color='red')
        g += text('y='+str(y), (1, y), color='red', fontsize=8, horizontal_alignment='left')
        g += line([(0, x), (1, y)], color='gray', linestyle='--')
    g.show(axes=False)
    xtoydict = {}
    for x, y in relation:
        if x in xtoydict:
            if xtoydict[x] != y:  # Αν μία τιμή x αντιστοιχεί σε πολλές τιμές y
                return False
        else:
            xtoydict[x] = y
    return True
#relation = [(1, 3), (2, 4), (3, 5), (1, 6)]
relation = [(1, 3), (2, 4), (3, 5), (1, 6), (4, 7), (5, 8), (2, 9), (3, 10), (6, 11), (7, 12)]
isFunction(relation)
```

**Παράδειγμα**

Το "Κόσκινο του Ερατοσθένη" αποτελεί έναν αλγόριθμο για την εύρεση όλων των πρώτων αριθμών μέχρι έναν συγκεκριμένο ακέραιο. Θα δημιουρήσουμε συνάρτηση η οποία θα δέχεται έναν αριθμό και θα επιστρέφει μία λίστα με όλους τους πρώτους μέχρι τον αριθμό αυτό.

1. Δημιουργούμε μια λίστα από διαδοχικούς ακέραιους από το 2 μέχρι το m: (2, 3, 4, ..., m).

2. Αρχικά, το p είναι ίσο με 2, τον 1ο πρώτο αριθμό,

3. Διαγράφουμε από τη λίστα όλα τα πολλαπλάσια του p στη λίστα (2p, 3p, 4p, κτλ)

4.	Βρίσκουμε τον 1ο αριθμό που απομένει στη λίστα μετά τον p (αυτός ο αριθμός είναι ο επόμενος πρώτος αριθμός) και αντικαθιστούμε το p με αυτόν τον αριθμό.

5.	Επαναλαμβάνουμε τα βήματα 3 και 4 μέχρι το $p^2$ να είναι μεγαλύτερο από n.

6.	Όλοι οι αριθμοί που απομένουν στη λίστα είναι πρώτοι αριθμοί.

```python
def eratosthenesSieve(m):
    # Δημιουργούμε τη λίστα με τους αριθμούς από 2 έως n
    sieve = list(range(2, m + 1))
    
    # Κόσκινο του Ερατοσθένη
    for p in range(2, int(m**0.5) + 1):
        if p in sieve:  
            # Διαγράφουμε όλα τα πολλαπλάσια του p (εκτός του ίδιου του p) ουσιαστικά κρατώντας τα υπόλοιπα 
            sieve = [x for x in sieve if x == p or x % p != 0]
    
    return sieve
eratosthenesSieve(30)
```

# Γεννήτορες (Generators)

Στο SageMath, οι γεννήτορες (generators) μπορούν να αξιοποιηθούν σε διάφορες εφαρμογές για αποδοτική επεξεργασία δεδομένων, κυρίως όταν χρειάζονται ακολουθίες μεγάλου μεγέθους ή υπάρχει αναγκαιότητα αποθήκευσης μόνο του απαραίτητου μέρους της σειράς στη μνήμη.

Μία συνάρτηση γεννήτορας δημιουργείται με τον ίδιο ακριβώς τρόπο με μία συνάρτηση python με την διαφορά ότι αντί να επιστρέφει κάποια τιμή με την `return`, αυτή επιστρέφεται με την `yield`. Επιπλέον, δεν επιστρέφει απλά την τιμή αλλά έναν `iterator`, όπως οι λίστες τα strings κ.ο.κ.. Οι `iterator` μπορούν να χρησιμοποιηθούν συνδυαστικά με μία `for` ή με την ειδική μέθοδο `next()` που επιστρέφει την επόμενη κάθε φορά τιμή.

Κάθε φορά που θα καλείται η συνάρτηση θα επιστρέφεται η επόμενη `yield`. Για παράδειγμα στις επόμενες διαδοχικές κλήσεις της _gen_ 

```python
def gen():
    yield 2
    yield 3
    yield 5
for i in gen():
    print(i,end=' ')
g = gen()
next(g),next(g),next(g)
```

Όταν εξαντληθούν τα στοιχεία που μπορεί να επιστρέψει ένας γεννήτορας, δημιουργείται μία εξαίρεση. Στο επόμενο παράδειγμα η μέθοδος `next` δεν έχει άλλα στοιχεία να επιστρέψει.

```python
next(g)
```

**Παράδειγμα**

Οι γεννήτορες επιτρέπουν τη δημιουργία μιας ακολουθίας Fibonacci χωρίς να αποθηκεύονται όλα τα στοιχεία της.

```python
# Γεννήτορας Fibonacci
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
gen = fibonacci()
for i in range(10):
    print(next(gen))
```

Ο γεννήτορας για πρώτους αριθμούς επιστρέφει μόνο τον επόμενο κάθε φορά, χωρίς να απαιτείται η αποθήκευση όλων των αριθμών που ελέγχει. Για να μπορέσουμε, όμως, να προσπελάσουμε επαναληπτικά στοιχεία θα πρέπει με την `iter` να μετατρέψουμε το αντικείμενο `Primes` σε γεννήτορα.

```python
gen = iter(Primes())
for i in range(10):
    print(next(gen),end=' ')
```

**Παράδειγμα** 

Ο γεννήτορας αποδίδει κάθε επόμενο όρο της σειράς Taylor για την προσέγγιση της συνάρτησης sin(x) σε σημείο x, μειώνοντας τον αριθμό των υπολογισμών.

```python
# Γεννήτορας για την ακολουθία Taylor της sin(x)
def sin_taylor(x):
    term, n = x, 1
    while True:
        yield term
        term *= -x**2 / ((2*n) * (2*n + 1))
        n += 1

# Υπολογισμός των πρώτων 10 όρων για x = π/4
x = pi/4
gen = sin_taylor(x)
approximation = sum(next(gen) for _ in range(10))
print(approximation)
```
