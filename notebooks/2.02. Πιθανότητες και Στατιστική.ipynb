# Πιθανότητες

Ο **πληθάριθμος** ενός συνόλου, ο αριθμός δηλαδή των στοιχείων του, δίνεται είτε με την μέθοδο `cardinality` του SageMath είτε με την `len`. Η τελευταία αποτελεί μέθοδο της Python και μας επιστρέφει το μήκος κάθε συλλογής, άρα και του συνόλου.

```python
Ω = Set([1, 2, 3, 4, 5])
len(Ω), Ω.cardinality()
```

Ο ορισμός της πιθανότητας ενός ενδεχομένου $A$ σε πεπερασμένο δειγματικό χώρο $\Omega$ είναι:  

$$
P(A) = \frac{|A|}{|\Omega|}
$$

όπου $|A|$ είναι το πλήθος των στοιχείων του ενδεχομένου $A$ και $|\Omega|$ το πλήθος των στοιχείων του δειγματικού χώρου.


**Παράδειγμα**

Αν υποθέσουμε ότι θέλουμε να βρούμε την πιθανότητα $P(\leq 3) = P(1 \lor 2 \lor 3) = P({1, 2, 3})$ στο παραπάνω σύνολο. Για $\Omega = \{1,2,3,4,5\}$ και $A = \{1,2,3\}$, έχουμε  

$$
P(A) = \frac{|A|}{|\Omega|} = \frac{3}{5}
$$

```python
A = Set([1, 2, 3])
P = A.cardinality() / Ω.cardinality()
P
```

## Διάγραμμα διασποράς, scatter plot

Ιδανικό τύπο διαγράμματος για την παρουσίαση της σχέσης μεταξύ δύο μεταβλητών αποτελεί το διάγραμμα διασποράς. Κάθε σημείο στο διάγραμμα αντιστοιχεί σε ένα ζεύγος τιμών, με τους άξονες x και y να αντιπροσωπεύουν τις δύο διαφορετικές μεταβλητές των οποίων την σχέση θέλουμε να μελετήσουμε.

Η μέθοδος του SageMath για την παράσταση διαγράμματος διασποράς είναι η `scatter_plot` η οποία δέχεται λίστα από tuples δύο στοιχείων. Αν τα δεδομένα μας είναι για παράδειγμα δύο ξεχωριστές λίστες θα πρέπει να τα φέρουμε στην αποδεκτή μορφή. Αυτό μπορεί να γίνει με διάφορους τρόπους όπως μια δομή επανάληψης, η τεχνική **comprehension** αλλά και η συνάρτηση Python `zip` που συνδυάζει πολλές συλλογές δημιουργώντας tuples με αντίστοιχα στοιχεία.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 4, 6, 5, 7, 8, 9]

Lzip = list(zip(x, y))
Lfor = []
for i in range(len(x)):
    Lfor.append((x[i], y[i]))
Lcomp = [(x[i], y[i]) for i in range(len(x))]
Lzip,Lfor,Lcomp
```

Αφού τα έχουμε φέρει στην αποδεκτή μορφή καλούμε την `scatter_plot`.

```python
scatter_plot(Lzip,figsize=(4,3))
```

Πέρα από την λίστα των tuples με τα προς εμφάνιση δεδομένα, που αποτελεί και υποχρεωτική παράμετρο, μια σειρά από προαιρετικές παραμέτρους μπορούν να αλλάξουν την εμφάνιση του γραφήματος. Η `alpha` θέτει την διαφάνεια, η `markersize` το μέγεθος, η `marker` το στυλ και η `facecolor` με την `edgecolor` το χρώμα γεμίσματος και περιγράμματος του κάθε σημείου. Με την παράμετρο `title` βάζουμε τίτλο στο γράφημa και με την `axes_labels` δίνουμε όνομα στους άξονες.

## Ιστόγραμμα, histogram

Το ιστογράμματα, είναι παρόμοια με τα ραβδογράμματα αλλά χρησιμοποιούνται για να παρουσιάσουν τη συχνότητα κατανομής ενός συνόλου δεδομένων. Οι τιμές του συνόλου χωρίζονται σε ομάδες ή κλάσεις, με τη συχνότητα κάθε κλάσης να απεικονίζεται με μπάρες ίσου πλάτους και διαφορετικού ύψους (το ύψος διαμορφώνεται ανάλογα με την συχνότητα).

Για να δημιουργηθεί στο Sage ένα ιστόγραμμα καλείται η `histogram` με μια λίστα τιμών. Η παράμετρος `bins` θέτει το πλήθος των τμημάτων (κάδων) στα οποία θα διαχωριστούν τα δεδομένα. Κάθε **bin** είναι της μορφής $[𝑎, 𝑏)$. Ο τελευταίος κάδος είναι της μορφής $[𝑎, 𝑏]$ και περιλαμβάνει φυσικά πάντα το μεγαλύτερο στοιχείο της λίστας. Έτσι, στο επόμενο παράδειγμα τα δεδομένα είναι από 0 έως 10 και έχουν χωριστεί σε 5 τμήματα, διότι θέσαμε `bins=5`. Τα τμήματα αυτά είναι: 

- τμήμα από τιμές 0 έως <2 (υπάρχουν 2 τέτοιες)

- τμήμα από τιμές >=2 έως <4 (εδώ ανήκουν οι 2, 2, 3, 3, 3 που είναι το πλήθος 5 δηλ. εμφανίζονται με συχνότητα ίση με 5) κοκ.

- Το τελευταίο τμήμα είναι εκείνο που αποτελείται από τις τιμές 8 ή 9 ή 10. Εδώ μέσα βρίσκονται οι 8, 9, 9, και 10 δηλαδή το τμήμα αυτό εμφανίζεται στο ιστόγραμμα με συχνότητα 4.

```python
data = [0, 1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 
        6, 7, 7, 7, 7, 8, 9, 9, 10]
histogram(data, bins=5,figsize=(4,3))
```

Η `bins` μπορεί να διαχωρίσει τα δεδομένα με όποιο τρόπο εμείς θέλουμε και να τα τοποθετήσει σε διαφορετικούς κάδους διαφορετικού πλάτους. Στο παρακάτω παράδειγμα οι δυο πρώτοι κλάδοι είναι ίδιου πλάτους, ενώ ο τελευταίος, δηλ. ο $[6,10]$, έχει μεγαλύτερο πλάτος και όπως βλέπουμε τυχαίνει να έχει συχνότητα 10.

```python
data = [0, 1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 6, 7, 7, 7, 7, 8, 9, 9, 10]
bins = [0, 3, 6, 10]  # Ορίσμός των διαστημάτων [0,3), [3,6), [6,10]
histogram(data, bins=bins,figsize=(4,3))
```

Η `histogram` δέχεται μια σειρά από παραμέτρους για την μορφοποίηση της εμφάνισης των γραφημάτων όπως `color` και `edgecolor` για το χρώμα `title` και `axes_labels` για τίτλους γραφήματος και αξόνων.

### Κανονική κατανομή, gaussian

Η κανονική, `gaussian`, κατανομή αναφέρεται σε συνεχείς μεταβλητές και χρησιμοποιείται για να περιγραφούν μεταβλητές πραγματικών τιμών. Οι τιμές συγκεντρώνονται κυρίως γύρω από μια μέση τιμή, _μ_ ανάλογα την διασπορά τους $σ^2$. Η κανονική κατανομή είναι πολύ σημαντική στη φύση καθώς την ακολουθούν τα περισσότερα συνεχή φαινόμενα.

Η `RealDistribution` με παράμετρο `gaussian` επιστρέφει τιμές που ακολουθούν την κανονική κατανομή. Μπορούμε επιπλέον να θέσουμε την τυπική απόκλιση σ. **Παράδειγμα**

Θα δημιουργήσουμε 1000 αριθμούς που ακολουθούν την κανονική κατανομή, αρχικά με απόκλιση σ=1. Επιπλέον θα θέσουμε μία μετατόπιση στους τυχαίους αριθμούς κατά m. ## Διακριτές κατανομές

Οι διακριτές κατανομές (**discrete distributions**) είναι εκείνες οι κατανομές όπου οι τυχαίες μεταβλητές παίρνουν μόνο διακριτές, πεπερασμένες ή άπειρες αλλά μετρούμενες τιμές. Αυτές οι τιμές μπορούν να είναι ακέραιοι ή άλλες διακριτές τιμές (π.χ. αριθμοί). Παραδείγματα τέτοιων κατανομών είναι η **διωνυμική** και η **Poisson**. Φυσικά υπάρχουν πολλές άλλες όπως π.χ. η Υπερβολική (Hypergeometric) η κατανομή Negative Binomial και η κατανομή Bernoulli.

### Διωνυμική κατανομή, binomial

Η διωνυμική κατανομή `binomial` περιγράφει την πιθανότητα να συμβούν ακριβώς k επιτυχίες σε n ανεξάρτητα πειράματα, όταν η πιθανότητα επιτυχίας σε κάθε πείραμα είναι _p_. Η διωνυμική κατανομή χρησιμοποιείται σε καταστάσεις όπου υπάρχει ένα σύνολο από δυαδικές επιλογές, όπως:

- η ρίψη ενός νομίσματος, κορόνα ή γράμματα

- η επιτυχία ή αποτυχία σε ένα τεστ. 

Η πιθανότητα P(X=k) για k επιτυχίες σε n πειράματα δίνεται από τον τύπο:
$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$
Ο διωνυμικός συντελεστής $\binom{n}{k}$ δίνεται από τον τύπο:
$$
\binom{n}{k} = \frac{n!}{k! (n-k)!}
$$
Έτσι για παράδειγμα:
$$
\binom{5}{2} = \frac{5!}{2! (5-2)!} = \frac{5 \cdot 4 \cdot 3!}{2 \cdot 1 \cdot 3!} = \frac{5 \cdot 4}{2 \cdot 1} = 10
$$

Υπάρχει και μέθοδος στο Sage που επιστρέφει τον διωνυμικό συντελεστή, η `binomial`.

```python
binomial(5, 2)
```

Η κατανομή αυτή δεν υποστηρίζεται άμεσα από το SageMath και είτε θα πρέπει να υλοποιηθεί προγραμματιστικά είτε να γίνει χρήση βιβλιοθήκης.

```python
import numpy as np
num = 20 
p = 0.5
size = 1000
data = np.random.binomial(num, p, size)
histogram(data, bins=range(num + 2), color="skyblue")
```

### Κατανομή Poisson

Η κατανομή Poisson χρησιμοποιείται για την περιγραφή του αριθμού συμβάντων που συμβαίνουν σε ένα σταθερό χρονικό διάστημα ή χώρο, όταν τα συμβάντα είναι ανεξάρτητα μεταξύ τους και η μέση συχνότητα είναι γνωστή. Παραδείγματα εφαρμογής της κατανομής Poisson, ο αριθμός πελατών που φτάνουν σε ένα κατάστημα ανά ώρα, των τηλεφωνικών κλήσεων που λαμβάνει ένα κέντρο εξυπηρέτησης κ.ο.κ.

Ο τύπος πιθανότητας για την κατανομή Poisson είναι:
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$


Όπου:

- $\lambda$: η μέση συχνότητα των συμβάντων

- $k$: ο αριθμός των συμβάντων $k = 0, 1, 2, \dots$

- $e$: η βάση του φυσικού λογαρίθμου $e \approx 2.718$

Και πάλι θα γίνει χρήση της βιβλιοθήκης `numpy`.

```python
import numpy as np
l = 5
size = 1000
data = np.random.poisson(l, size)
histogram(data, bins=range(min(data), max(data) + 2), color="skyblue")
```

# Περιγραφική στατιστική

Η περιγραφική στατιστική (descriptive statistics) αφορά την ανάλυση και παρουσίαση των δεδομένων με σκοπό την εξαγωγή χρήσιμων πληροφοριών με απλό και κατανοητό τρόπο. Αυτή η μορφή στατιστικής επικεντρώνεται στην περίληψη και παρουσίαση των βασικών
χαρακτηριστικών των δεδομένων μέσω διαγραμματικών παραστάσεων και αριθμητικών μετρήσεων.

## Συναρτήσεις SageMath

Στην συνέχεια δίνονται οι συνηθέστερες συναρτήσεις περιγραφικής στατιστικής δεδομένων.

- `mean(data)`: Μέση τιμή (ο αριθμητικός μέσος όρος των δεδομένων).

- `median(data)`: Διάμεση τιμή (η τιμή που βρίσκεται στη μέση όταν τα δεδομένα ταξινομηθούν).

- `variance(data)`: Διακύμανση (το μέτρο της διασποράς των δεδομένων από τη μέση τιμή τους).

- `std(data)`: Τυπική απόκλιση (η τετραγωνική ρίζα της διακύμανσης, που δείχνει πόσο διαφέρουν τα δεδομένα από τη μέση τιμή).

- `sum(data)`: Άθροισμα (το συνολικό άθροισμα όλων των τιμών των δεδομένων).

- `min(data)`: Ελάχιστη τιμή (η μικρότερη τιμή από το σύνολο των δεδομένων).

- `max(data)`: Μέγιστη τιμή (η μεγαλύτερη τιμή από το σύνολο των δεδομένων).

- `mode(data)`: Επικρατούσα τιμή ή κορυφή (η πιο συχνά εμφανιζόμενη τιμή στα δεδομένα).

Οι παραπάνω συναρτήσεις του SageMath αφορούν παλαιότερες εκδόσεις. Αν δοκιμαστούν στο cell ή σε κάποιο Notebook με την τελευταία έκδοση του Sage θα εμφανίσουν μεν το αποτέλεσμα αλλά με μια προειδοποίηση ότι οι μέθοδοι αυτοί δεν υποστηρίζονται πλέον, `DeprecationWarning`. Αντί για αυτές θα πρέπει να χρησιμοποιήσουμε το πακέτο `NumPy` ή το `SciPy` της Python.

Οι επόμενες κλήσεις αφορούν **μη** υποστηριζόμενες, πλέον, μεθόδους. Το περιβάλλον θα μας προειδοποιήσει με `DeprecationWarning`.

```python
data = [1, 2, 2, 3, 4, 4, 4, 5, 6]
mean(data),median(data),variance(data),\
    std(data),sum(data),min(data),max(data),mode(data)
```

## Η βιβλιοθήκη NumPy

Το NumPy (Numerical Python) είναι μια ισχυρή και θεμελιώδης βιβλιοθήκη της Python για αριθμητικούς υπολογισμούς και εργασία με πολυδιάστατους πίνακες (arrays). Χρησιμοποιείται ευρέως στην επιστήμη δεδομένων, στατιστική, μηχανική μάθηση, φυσική,
πληροφορική και άλλους επιστημονικούς τομείς.

Οι παραπάνω μέθοδοι έχουν αντικατασταθεί από το πακέτο `NumPy` που παραδοσιακά εισάγεται με το alias _np_ με τον τρόπο που ακολουθεί.

## Ακραίες τιμές, outliers

Σε μια σειρά παρατηρήσεων οι ακραίες τιμές, **outliers** είναι παρατηρήσεις που, ίσως και από λάθος, είναι πολύ διαφορετικές από το γενικό σύνολο. Η μέθοδος `np.percentile` μας βοηθά να απομονώσουμε τέτοιες τιμές και το ιστόγραμμα μπορεί να τις αναδείξει.

**Παράδειγμα**

Θα παραχθεί ένας μεγάλος αριθμός τυχαίων τιμών που ακολουθούν την κανονική κατανομή και στην συνέχεια θα απομονωθούν οι ακραίες τιμές. Στον επόμενο κώδικα _q1_ είναι το 1ο τεταρτημόριο, _q3_ το 3ο. Η διαφορά q3-q1 ονομάζεται ενδοτεταρτημοριακό εύρος, **interquartile range** (IQR), και δείχνει δείχνει το “εύρος” στο οποίο βρίσκεται η μεσαία μάζα των δεδομένων. 
Ένας κλασικός κανόνας, ο **Tukey’s rule**, για τον εντοπισμό ακραίων τιμών (outliers) είναι:

- Κάτω όριο (lower bound): $Q_1 - 1.5\cdot IQR$

- Άνω όριο (upper bound): $Q_3 + 1.5\cdot IQR$

Ό,τι βρίσκεται έξω από αυτό το διάστημα θεωρείται outlier. Στο επόμενο ιστόγραμμα οι κανονικές τιμές εμφανίζονται με γαλάζιο και οι ακραίες με κόκκινο. 

```python
import numpy as np
dist = RealDistribution('gaussian', 2) 
data = [dist.get_random_element() for i in range(5000)]

q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

normal_data = [x for x in data if lower_bound <= x <= upper_bound]
outliers = [x for x in data if x < lower_bound or x > upper_bound]

g = Graphics()
g += histogram(normal_data, bins=20, color='skyblue')
g += histogram(outliers, bins=20, color='red')
g.show()
```

# Στατιστικά Τεστ

Τα Στατιστικά Τεστ (ή στατιστικές δοκιμές, στα αγγλικά statistical tests) είναι μέθοδοι που χρησιμοποιούνται για να ελέγξουμε αν μια υπόθεση για έναν πληθυσμό είναι πιθανή, με βάση δεδομένα από ένα δείγμα. Σκοπός τους να απαντήσουμε σε ερωτήσεις όπως:

- «Έχει η νέα θεραπεία διαφορετική επίδραση από την παλιά;»

- «Η μέση τιμή του προϊόντος που παρήχθη είναι όντως 500 ml;»

- «Υπάρχει συσχέτιση μεταξύ ύψους και βάρους;»

Παραδείγματα Κοινών Στατιστικών Τεστ:

| Τεστ | Χρήση |
|------|-------|
| *t-test* | Ελέγχει αν δύο μέσα μεγέθη είναι διαφορετικά |
| $\chi^2$ (chi-square) | Ελέγχει συσχέτιση μεταξύ κατηγορικών μεταβλητών |

## t-test

Το t-test χρησιμοποιείται για να εξετάσει αν η μέση τιμή ενός δείγματος διαφέρει από μια γνωστή θεωρητική μέση τιμή ή οι μέσες τιμές δύο δειγμάτων έχουν στατιστικά σημαντική διαφορά. Τα δείγματα μπορεί να είναι 

- ανεξάρτητα, για παράδειγμα τα ύψη των ανθρώπων σε δύο διαφορετικές χώρες

- συσχετισμένα, για παράδειγμα αιματολογικές μετρήσεις στην ίδια ομάδα ανθρώπων πριν και μετά την λήψη κάποιου φαρμάκου.

Ο έλεγχος επιστρέφει αν η διαφορά μεταξύ των μέσων είναι στατιστικά σημαντική.

Στο SageMath χρησιμοποιούμε τρίτες βιβλιοθήκες για την διενέργεια του t-test. Στην βιβλιοθήκη `SciPy` υπάρχει η μέθοδος ttest_ind η οποία επιστρέφει:

- `t-statistic`:  (ή t-στατιστική τιμή) είναι ένας αριθμός που προκύπτει κατά τη διάρκεια ενός t-test και μας βοηθά να συγκρίνουμε τη μέση τιμή ενός δείγματος με τη μέση τιμή ενός άλλου δείγματος (independent t-test), ή μια θεωρητική τιμή (one-sample t-test). 

Ο μαθηματικός της τύπος είναι
$$t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}}$$ 
όπου:

- $\bar{x}$: Η μέση τιμή του δείγματος.

- $\mu_0$: Η υπόθεση για τη μέση τιμή του πληθυσμού (συνήθως 0, μέση τιμή δείγματος = μέση τιμή διαφοράς).

- s: Η τυπική απόκλιση του δείγματος.

- n: Το μέγεθος του δείγματος.


- `p-value`: Μας λέει την πιθανότητα να έχουμε τα παρατηρούμενα δεδομένα. Αν το p-value είναι μικρότερο από ένα καθορισμένο όριο, συνήθως 0.05, μπορούμε να υποστηρίξουμε ότι τα δεδομένα μας παρέχουν στατιστικά σημαντική απόδειξη για μια διαφορά μεταξύ υπόθεσης και πληθυσμού.

**Παράδειγμα**

Έστω ότι έχουμε δύο δείγματα υψών από διαφορετικές χώρες και θέλουμε να εξετάσουμε αν οι μέσες τιμές τους διαφέρουν σημαντικά.

```python
from scipy.stats import ttest_ind

h_c1 = [160, 165, 170, 175, 180, 185, 190, 195, 
        200, 155, 162, 168, 172, 178, 183, 188, 
        193, 198, 202, 207]
h_c2 = [155, 160, 165, 170, 175, 180, 185, 190, 
        195, 197, 198, 200, 201, 202, 204, 207, 
        212, 215, 220, 225]

t_stat, p_value = ttest_ind(h_c1, h_c2)

# Εμφάνιση του αποτελέσματος
print("T-Statistic:", t_stat)
print("P-value:", p_value)

# Απόφαση βασισμένη στο p-value
if p_value < 0.05:
    print("Διαφορά στατιστικά σημαντική.")
else:
    print("Δεν υπάρχει στατιστικά σημαντική διαφορά.")
```

## Συνδιακύμανση, covariance

Η συνδιακύμανση ορίζει την κατεύθυνση της σχέσης μεταξύ δύο μεταβλητών την αναλογικότητα, δηλαδή, μεταξύ των μεταβλητών. Αν οι τιμές των δύο μεταβλητών αυξάνονται ή μειώνονται ταυτόχρονα, τότε η συνδιακύμανση είναι θετική. Αν μία μεταβλητή αυξάνεται ενώ η άλλη μειώνεται, τότε η συνδιακύμανση είναι αρνητική. Υπολογίζεται ως το άθροισμα των γινομένων των διαφορών από τους μέσους όρους, διαιρούμενο με το μέγεθος του δείγματος:
$$
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} \left( X_i - \bar{X} \right) \left( Y_i - \bar{Y} \right)
$$
Ανάλογα με την τιμή της μπορούμε να εξάγουμε τα επόμενα συμπεράσματα:

- Θετική συνδιακύμανση: Όταν η μία μεταβλητή αυξάνεται, η άλλη τείνει επίσης να αυξάνεται.

- Αρνητική συνδιακύμανση: Όταν η μία μεταβλητή αυξάνεται, η άλλη τείνει να μειώνεται.

- Μηδενική συνδιακύμανση: Οι μεταβλητές είναι ανεξάρτητες (δεν σχετίζονται γραμμικά).

Στο SageMath ο υπολογισμός μπορεί να γίνει προγραμματιστικά, πρώτα υπολογίζοντας τις μέσες τιμές και στην συνέχεια εφαρμόζοντας το άθροισμα των διαφορών των τιμών από το μέσο όρο.

```python
heights = [160, 165, 170, 175, 180, 185, 190, 195, 200, 155, 
           162, 168, 172, 178, 183, 188, 193, 198, 202, 207]
weights = [55, 60, 75, 70, 75, 80, 85, 90, 95, 50, 60, 57, 68, 
           74, 79, 84, 80, 92, 105, 100]

mean_heights = mean(heights)
mean_weights = mean(weights)

# Υπολογισμός συνδιακύμανσης
n = len(heights)
covariance = sum((heights[i] - mean_heights) * 
                 (weights[i] - mean_weights) for i in range(n)) / n
N(covariance)
```

Εναλλακτικά, μπορούν να χρησιμοποιηθούν βιβλιοθήκες. Η `np.cov` της `NumPy` μας επιστρέφει πίνακα διακυμάνσεων-συνδιακυμάνσεων με τα διαγώνια στοιχεία να αντιπροσωπεύουν τη διακύμανση κάθε μεταβλητής, και τα άλλα δύο την συνδιακύμανση.

```python
import numpy as np

heights_arr = np.array(heights)
weights_arr = np.array(weights)
cov_matrix = np.cov(heights_arr, weights_arr, ddof=0)
covariance = cov_matrix[0, 1]

covariance
```

## Συντελεστής Γραμμικής Συσχέτισης, Pearson

Ένα μέτρο που μας δίνει το μέγεθος της γραμμικής σχέσης μεταξύ δύο μεταβλητών είναι ο συντελεστής γραμμικής συσχέτισης. Ουσιαστικά προκύπτει αν διαιρέσουμε την συνδιακύμανση με το γινόμενο των τυπικών αποκλίσεων. Δηλαδή, με βάση ένα δείγμα n ζευγών παρατηρήσεων $(x_i, y_i) i=1,2,...,n$, συμβολίζεται με $r(X,Y)$ ή απλούστερα r και δίνεται:
$$
r = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^n (y_i - \bar{y})^2}}
$$
Η τιμή του συντελεστή βρίσκεται μεταξύ [-1,1] και για τις τιμές: 

-  0 < r < +1: θετικά γραμμικά συσχετισμένες (r=1: Τέλεια θετική συσχέτιση)

- -1 < r <0: αρνητικά γραμμικά συσχετισμένες (r=−1: Τέλεια αρνητική συσχέτιση)

- r=0: δεν υπάρχει γραμμική συσχέτιση.

Στο SageMath δεν υπάρχει αντίστοιχη μέθοδος και θα πρέπει να υπολογιστεί προγραμματιστικά.

```python
heights = [160, 165, 170, 175, 180, 185, 190, 
           195, 200, 155, 162, 168, 172, 178, 
           183, 188, 193, 198, 202, 207]
weights = [55, 60, 75, 70, 75, 80, 85, 
           90, 95, 50, 60, 57, 68, 74, 
           79, 84, 80, 92, 105, 100]

mean_x = sum(heights) / len(heights)
mean_y = sum(weights) / len(weights)
numerator = sum((xi - mean_x) * (yi - mean_y) 
                for xi, yi in zip(heights, weights))
denominator = sqrt(sum((xi - mean_x)^2 for xi in heights) * 
                   sum((yi - mean_y)^2 for yi in weights))
pearson_r = numerator / denominator
N(pearson_r)
```

Έχουμε την δυνατότητα, όπως και σε διάφορες άλλες λειτουργίες, να εισάγουμε βιβλιοθήκες όπως η `NumPy` ή η `SciPy` και να καλέσουμε τις αντίστοιχες μεθόδους `corrcoef` ή `pearsonr`.

```python
import numpy as np
pearson_r = np.corrcoef(heights, weights)[0, 1];pearson_r
```

## Παλινδρόμηση, regression

Η παλινδρόμηση αποτελεί την στατιστική μέθοδο που χρησιμοποιείται για να βρούμε τη γραμμή που αντιπροσωπεύει όσο το δυνατόν καλύτερα τη σχέση μεταξύ δύο, ή περισσότερων, μεταβλητών, έστω x και y με την πρώτη να αποτελεί την εξαρτημένη μεταβλητή και την 2η την ανεξάρτητη. Βασίζεται στη μείωση του σφάλματος μέσω της μεθόδου των ελαχίστων τετραγώνων. Σημειώνεται ότι οι ανεξάρτητες μεταβλητές, από τις οποίες εξαρτάται η εξαρτημένη, μπορεί να είναι και περισσότερες από μια.

### Γραμμική παλινδρόμηση

**Παράδειγμα**

Θα χρησιμοποιήσουμε τα δεδομένα από το παράδειγμα με τα ύψη και τα βάρη.

```python
heights = [160, 165, 170, 175, 180, 185, 190, 
           195, 200, 155, 162, 168, 172, 178, 
           183, 188, 193, 198, 202, 207]
weights = [55, 60, 75, 70, 75, 80, 85, 
           90, 95, 50, 60, 57, 68, 74, 
           79, 84, 80, 92, 105, 100]
data = list(zip(heights,weights))
a, b = var('a b')
model(x) = a * x + b  
fits = find_fit(data, model)
fits
```

Θα χρησιμοποιήσουμε τους συντελεστές a και b για να απεικονίσουμε και την συνάρτηση $ax+b$ και τα σημεία. Υπενθυμίζεται ότι η `rhs()` επιστρέφει το δεξί μέλος των εξισώσεων. Τα σημεία μπορούν να απεικονιστούν με την `points` ή με `scatter_plot`.

```python
f(x) = fits[0].rhs()*x + fits[1].rhs()
g = Graphics()
g += plot(f(x),(x,150,210))
g+=scatter_plot(data)
g.show()
```

Όπως είναι φυσικό, ένα μοντέλο μπορεί να έχει παραπάνω από μια ανεξάρτητες μεταβλητές που επηρεάζουν την εξαρτημένη μεταβλητή. Το μοντέλο στην περίπτωση αυτή θα έχει για παράδειγμα μια τέτοια μορφή: $z=a⋅x+b⋅y+c$.

**Παράδειγμα**

Θα μελετήσουμε ένα κλασσικό παράδειγμα παλινδρόμησης τιμολόγησης σπιτιών. Θα υποθέσουμε ότι οι τιμή των σπιτιών σε χιλιάδες ευρώ, λίστα _prices_ εξαρτάται από:

- Απόσταση από το κέντρο της πόλης σε χιλιόμετρα, λίστα _distance_

- Αριθμό τετραγωνικών μέτρων του σπιτιού λίστα _size_

Για ευκολότερη διαχείριση του μοντέλου η `find_fit` μπορεί να επιστρέψει λεξικό αν θέσουμε `solution_dict=True`. 
Στην παραπάνω εξίσωση το z θα είναι η τιμή, το x η απόσταση και το y τα τετραγωνικά.

```python
distances = [1, 3, 2, 6, 4, 8, 5, 7, 
             10, 9, 3.5, 4.5, 2.5, 6.5, 5.5]
sizes = [60, 80, 55, 120, 90, 110, 100, 105, 
         140, 130, 85, 95, 70, 115, 108]
prices = [280, 260, 270, 180, 230, 150, 200, 160, 
          120, 135, 250, 240, 275, 190, 210]

data = list(zip(distances, sizes, prices))

x, y, a, b, c = var('x y a b c')
model(x, y) = a * x + b * y + c

fits = find_fit(data, model, solution_dict=True)
f(x,y) = fits[a]*x + fits[b]*y+fits[c]
f
```

### Πρόβλεψη

Τα μοντέλα που βρέθηκαν με την βοήθεια της `find_fit` μπορούν να χρησιμοποιηθούν για την πρόβλεψη νέων δεδομένων. 

**Παράδειγμα**

Αν μας δώσουν την θέση και την απόσταση νέων σπιτιών θα προβλέψουμε μια τιμή που να ακολουθεί το μοντέλο μας. Τα δύο σπίτια που μας ενδιαφέρουν βρίσκονται 4 και 6 χιλιόμετρα από το κέντρο και είναι 100 και 115 τετραγωνικά αντίστοιχα. Οι τιμές τους θα προβλεφθούν σύμφωνα με το μοντέλο μας.

```python
interpolation_points = [(4, 100), (6, 115)]
interpolation_results = [(p[0], p[1], f(p[0], p[1])) for p in interpolation_points]
f"Παρεμβολή: {interpolation_results}"
```
