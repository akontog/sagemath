# Άσκηση, Πεπλεγμένες συναρτήσεις

Υπενθύμιση:

Εξίσωση έλλειψης:
$$
\frac{x^2}{\alpha ^{2}}+{\frac {y^{2}}{\beta ^{2}}}=1
$$

Σχεδιάστε την ίδια έλλειψη με την `polar_plot`. Εξίσωση έλλειψης σε πολικές συντεταγμένες:
$$
\frac{a b}{\sqrt{(b \cos(\theta))^2 + (a \sin(\theta))^2}}
$$

Στον επόμενο κώδικα χρησιμοποιείται η συνάρτηση `parametric_plot` που χρησιμοποιείται όταν η καμπύλη έχει παρασταθεί παραμετρικά. Έτσι για παράδειγμα ο τύπος ενός κύκλου θα είναι:
$$ 
x(t) = cos(t), y(t) = sin(t) 
$$
Αν θέσουμε διαφορετικές παραμέτρους a και b δημιουργούμε ελλειψοειδή:
$$ 
x(t) = α\cos{(t)}, y(t) = b\sin{(t)}
$$
Πειραματιστείτε με τις τιμές a και b στο επόμενο παράδειγμα.

```python
var('t')
a=1
b=1
parametric_plot((a*cos(t), b*sin(t)), (t, 0, 2*pi))
```

Η _rose curve_ που σχεδιάστηκε στο μάθημα με polar_plot μπορεί να σχεδιαστεί και με `parametric_plot` βασισμένοι στους τύπους:
$$
x(t) = \cos(kt)\cos(t), \quad y(t) = \cos(kt)\sin(t)
$$
Πειραματιστείτε με την τιμή k.

```python
k = 4
parametric_plot((cos(k*t)*cos(t), cos(k*t)*sin(t)), (t, 0, 2*pi), color='purple', title='Rose Curve')
```
