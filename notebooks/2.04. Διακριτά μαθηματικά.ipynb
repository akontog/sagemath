Τα Διακριτά Μαθηματικά (Discrete Mathematics) είναι ο κλάδος των μαθηματικών που ασχολείται με δομές που είναι διακριτές, δηλαδή μη συνεχείς — σε αντίθεση με τον λογισμό (ανάλυση), που ασχολείται με συνεχείς μεταβλητές και συναρτήσεις.

Κύρια χαρακτηριστικά τους είναι ότι:

1. ασχολούνται με αντικείμενα που μπορούμε να απαριθμήσουμε ένα-ένα (όπως ακέραιοι αριθμοί, γράφοι, σύνολα).

2. δεν περιλαμβάνουν συνεχείς συναρτήσεις ή παράγωγους όπως στα μαθηματικά της ανάλυσης.

Τα κύρια πεδία των Διακριτών μαθηματικών είναι:

- Λογική – συμβολική λογική, προτασιακός και κατηγορηματικός λογισμός.

- Θεωρία Συνόλων – μελέτη συλλογών αντικειμένων.

- Συνδυαστική – μελέτη μεθόδων απαρίθμησης και καταμέτρησης.

- Θεωρία Γραφημάτων (γράφοι) – κορυφές και ακμές, δίκτυα, διαδρομές.

Άλλα πεδία που περιλαμβάνει ο κλάδος των διακριτών μαθηματικών:

- Θεωρία Αριθμών - ιδιότητες των ακεραίων

- Άλγεβρα Boole – λογικές πράξεις και κυκλώματα.

- Αλγόριθμοι και Πολυπλοκότητα – βασικά μαθηματικά για την επιστήμη της πληροφορικής

- Σχέσεις και Συναρτήσεις – μαθηματικές δομές και ιδιότητες.

# Προτασιακός Λογισμός

## Λογικές πράξεις

Οι λογικές πράξεις είναι βασικές λειτουργίες στη μαθηματική λογική και τη θεωρία συνόλων. Χρησιμοποιούνται για να συνδυάζονται λογικές προτάσεις ή συνθήκες με στόχο τη λήψη συμπερασμάτων.

Ακολουθεί πίνακας με τις βασικές λογικές πράξεις.

| Λογική Πράξη                    | Σύμβολο     | Περιγραφή                                                      |
|----------------------------------|-------------|----------------------------------------------------------------|
| Άρνηση                          | `~`        | Το αντίθετο της πρότασης                                  |
| Σύζευξη (ΚΑΙ)                   | `&`         | Όλες οι προτάσεις είναι αληθείς             |
| Διάζευξη (Ή)                    | `\|`        | Τουλάχιστον μια από τις προτάσεις είναι αληθής      |
| Αποκλειστικό Ή (XOR)            | `^`         | Περιττός αριθμός προτάσεων είναι αληθής |
| Συνεπαγωγή (Αν ... τότε)         | `->`        | Αν ισχύει η πρώτη πρόταση, πρέπει να ισχύει η δεύτερη                                               |
| Ισοδυναμία (Αν και Μόνο Αν)     | `<->`       | Οι προτάσεις είναι ταυτόσημες |


Για να εφαρμόσουμε τις παραπάνω λογικές πράξεις πάνω σε κάποιους τύπους (που συμβολίζονται συνήθως με ένα γράμμα π.χ. P, Q, a, b, c κοκ. και λέγονται λογικές μεταβλητές) κάνουμε χρήση της μεθόδου `formula` του `propcalc` με χρήση του συνδυασμού `procalc.formula.` Στη συνέχεια αποτιμάμε τους νέους τύπους που κατασκευάσαμε θέτοντας κάποιες αληθοτιμές (`True` αληθής, ή `False` ψευδής) στις λογικές μεταβλητές με την μέθοδο `evaluate`. Η αποτίμηση δίνεται με την μορφή ενός λεξικού (τύπου δηλαδή `dict`).

Ας δούμε μια μια της λογικές πράξεις.

### Άρνηση

Αν Ρ είναι ένας ισχυρισμός, ο ισχυρισμός **όχι Ρ** ονομάζεται άρνηση του Ρ, συμβολίζεται $\neg P$ και στο SageMath `~P`. Το αποτέλεσμα θα είναι:

- Αληθής αν Ρ είναι False (Ψευδής)

- Ψευδής αν Ρ True (Αληθής)

```python
expr = propcalc.formula("~P")
expr.evaluate({'P': True}),expr.evaluate({'P': False})
```

### Σύζευξη

Aν $P$ και $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ και Q**, που συμβολίζεται με $P \land Q$ αληθεύει μόνο στην περίπτωση που και οι δύο ισχυρισμοί αληθεύουν. Στο SageMath η έκφραση δημιουργείται με το `&` και το αποτέλεσμα είναι:

- Αληθής αν και το $P$ και το $Q$ είναι Αληθή

- Σε όλες τις υπόλοιπες περιπτώσεις Ψευδής.

Αν ο λογικός τελεστής της σύζευξης εφαρμοστεί σε παραπάνω από δύο τελεστέους, για να προκύψει Αληθές αποτέλεσμα, θα πρέπει όλοι οι τελεστέοι να είναι Αληθείς.

```python
expr = propcalc.formula("a & b")
expr.evaluate({'a': True, 'b': False}),expr.evaluate({'a': True, 'b': True})
```

Στο επόμενο παράδειγμα με περισσότερες λογικές μεταβλητές η μοναδική `False` μεταβλητή οδηγεί σε `False` αποτέλεσμα.

```python
expr = propcalc.formula("a & b & c & d")
expr.evaluate({'a': True, 'b': True, 'c': True, 'd': False})
```

### Διάζευξη

Aν $P$ και $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ ή Q**, που συμβολίζεται με $P \lor Q$ και στο SageMath με `|` θα είναι:

- Ψευδής αν και το $P$ και το $Q$ είναι Ψευδή

- Σε όλες τις υπόλοιπες περιπτώσεις Αληθής.

Αν ο λογικός τελεστής της διάζευξης, $\lor$, εφαρμοστεί σε παραπάνω από δύο τελεστέους, για να προκύψει Ψευδές αποτέλεσμα, θα πρέπει να είναι όλοι Ψευδείς.

```python
expr = propcalc.formula("a | b")
expr.evaluate({'a': True, 'b': False}),expr.evaluate({'a': True, 'b': True}),\
    expr.evaluate({'a': False, 'b': False})
```

Αντίστροφα με την λογική της σύζευξης η μια `True` μεταβλητή οδηγεί σε Αληθές αποτέλεσμα.

```python
expr = propcalc.formula("a | b | c | d")
expr.evaluate({'a': False, 'b': False, 'c': False, 'd': True })
```

### Αποκλειστικό ή

Aν $P$, $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ αποκλειστικό-ή Q**, που συμβολίζεται με $P \oplus Q$ και στο SageMath με τον τελεστή `^` είναι αληθής όταν μόνο ο ένας από τους δύο ισχυρισμούς είναι αληθής.

- Αληθής αν το $P$ είναι Αληθές και το $Q$ Ψευδές και αντιστρόφως.

- Ψευδής αν και τα δύο είναι Αληθή ή και τα δύο ψευδή.

Αν ο λογικός τελεστής του αποκλειστικού ή εφαρμοστεί σε παραπάνω από δύο μεταβλητές, για να προκύψει Αληθές αποτέλεσμα, ο αριθμός μεταβλητών που έχουν αποτιμηθεί Αληθής (`True`) θα πρέπει να είναι περιττός.

```python
expr = propcalc.formula("a ^ b")
expr.evaluate({'a': True, 'b': False}),\
        expr.evaluate({'a': False, 'b': False}),\
        expr.evaluate({'a': True, 'b': True})
```

```python
expr = propcalc.formula("a ^ b ^ c")
expr.evaluate({'a': True, 'b': False, 'c': True})
```

### Συνεπαγωγή

Aν $P$, $Q$ είναι δύο ισχυρισμοί, τότε ο ισχυρισμός **Ρ συνεπάγεται Q**, $P \implies Q$ που εφαρμόζεται με τον τελεστή `->` είναι Αληθής όταν

- $P$ είναι αληθής και $Q$ αληθής

- $P$ Ψευδής, ανεξάρτητα της τιμής του $Q$

```python
expr = propcalc.formula("a -> b")
expr.evaluate({'a': True, 'b': False}),\
        expr.evaluate({'a': False, 'b': False}),\
        expr.evaluate({'a': True, 'b': True})
```

### Ισοδυναμία

Aν $P$, $Q$ είναι δύο ισχυρισμοί, η σχέση ισοδυναμίας τους, $Ρ \iff Q$, θα είναι αληθής αν και μόνο αν τα $P$ και $Q$ είναι ίσα.

- $P$ αληθές και $Q$ αληθές

- $P$ ψευδές και $Q$ ψευδές

```python
expr = propcalc.formula("a <-> b")
expr.evaluate({'a': True, 'b': False}),\
        expr.evaluate({'a': False, 'b': False}),\
        expr.evaluate({'a': True, 'b': True})
```

### Προτεραιότητα

Όπως και οι αριθμητικές, οι λογικές εκφράσεις μπορεί να είναι πολύπλοκες και να κατασκευάζονται με χρήση διάφορων λογικών πράξεων. Υπάρχει και για αυτές προτεραιότητα και αν θέλουμε να την παρακάμψουμε χρησιμοποιούμε παρενθέσεις. Διαφορετικά υπάρχει μια σειρά προτεραιότητας στην σειρά εκτέλεσής τους. Στον παραπάνω πίνακα οι πράξεις παρατίθενται με τη σειρά (>) προτεραιότητάς τους. Ακολουθεί και ενα παράδειγμα με και χωρίς παρενθέσεις.

Στην επόμενη έκφραση η πράξη `&` προηγείται της `|`. Έτσι θα αποτιμηθεί πρώτα η `b & c` και στη συνέχεια η διάζευξη του αποτελέσματος με την `a`.

```python
expr = propcalc.formula("a | b & c")
expr.evaluate({'a': True, 'b': False, 'c': False})
```

Αν θέλουμε να αποτιμηθεί πρώτα η διάζευξη και στην συνέχεια η σύζευξη θα χρησιμοποιήσουμε παρενθέσεις (με, όπως αναμένεται, διαφορετικό αποτέλεσμα).

```python
expr = propcalc.formula("(a | b) & c")
expr.evaluate({'a': True, 'b': False, 'c': False})
```

## Πίνακες Αλήθειας

Ο πίνακας αληθείας χρησιμοποιείται στη λογική για να καταγράφει όλες τις δυνατές τιμές μιας έκφρασης για ένα σύνολο λογικών μεταβλητών. Δείχνει το αποτέλεσμα μιας λογικής πράξης για αυτές τις μεταβλητές. Ο πίνακας οργανώνεται με τις γραμμές να αντιστοιχούν σε όλους τους δυνατούς συνδυασμούς τιμών των μεταβλητών.

Δίνουμε ως παράδειγμα τους πίνακες αλήθειας δύο και τριών λογικών μεταβλητών 𝐴,
𝐵 και 𝐶 για την λογική πράξη της σύζευξης.
$$
\begin{array}{|c|c|c|}
\hline
A & B & A \land B \\
\hline
\text{False} & \text{False} & \text{False} \\
\text{False} & \text{True} & \text{False} \\
\text{True} & \text{False} & \text{False} \\
\text{True} & \text{True} & \text{True} \\
\hline
\end{array}
$$

$$
\begin{array}{|c|c|c|c|}
\hline
A & B & C & A \land B \land C \\
\hline
\text{False} & \text{False} & \text{False} & \text{False} \\
\text{False} & \text{False} & \text{True} & \text{False} \\
\text{False} & \text{True} & \text{False} & \text{False} \\
\text{False} & \text{True} & \text{True} & \text{False} \\
\text{True} & \text{False} & \text{False} & \text{False} \\
\text{True} & \text{False} & \text{True} & \text{False} \\
\text{True} & \text{True} & \text{False} & \text{False} \\
\text{True} & \text{True} & \text{True} & \text{True} \\
\hline
\end{array}
$$

Ο απαιτούμενος αριθμός γραμμών σε έναν πίνακα n μεταβλητών είναι $2^n$ και συνηθίζεται να γράφονται οι τιμές των μεταβλητών με την σειρά που κατατάσσονται οι αριθμοί στο δυαδικό σύστημα. Παράδειγμα: αν έχουμε τρεις μεταβλητές a, b και c, τότε:

- η πρώτη γραμμή του πίνακα θα είναι η `False`, `False`, `False` διότι αντιστοιχεί στο 000 (=0),

- η δεύτερη θα είναι η `False`, `False`, `True` διότι αντιστοιχεί στην δυαδική μορφή 001 (=1),

- η τρίτη η `False`, `True`, `False` δηλαδή στον δυαδικό αριθμό 010 (=2) κ.ο.κ.

Σημειώνεται ότι όπου έχουμε 0 εννοείται το `False` και όπου 1 το `True` και ότι η παραπάνω σειρά δεν είναι υποχρεωτική όταν φτιάχνουμε τους δικούς μας πίνακες αλήθειας. Το SageMath, όμως, θα τους επιστρέψει με την παραπάνω σειρά.

Το SageMath δημιουργεί πίνακες αλήθειας μέσω της μεθόδου `truthtable` του πακέτου `propcalc`.

```python
s = propcalc.formula("a&b")
s.truthtable()
```

```python
s = propcalc.formula("~(a & b) | c")
s.truthtable()
```

Στην περίπτωση που δεν θέλουμε να εμφανίσουμε όλες τις γραμμές, όλους δηλαδή τους συνδυασμούς σε έναν πίνακα αλήθειας, οι παράμετροι `start` και `end` καθορίζουν ποιες γραμμές θα εμφανιστούν. Η γραμμή της παραμέτρου `end` δεν θα εμφανιστεί. Επιπλέον, στο επόμενο παράδειγμα η `truthtable` καλείται άμεσα με την δημιουργία της `formula` χωρίς να είναι ανάγκη να επιστραφεί πρώτα σε μεταβλητή.

```python
propcalc.formula("~(a & b) | c").truthtable(start=3,end=6)
```

## Λογική έκφραση ως αναδρομική δομή λίστας

Η μέθοδος `sage.logic.booleval.eval_formula(t,d)`  υπολογίζει την τιμή μίας λογικής πρότασης που έχει δοθεί σε μορφή λίστας t σε δεντρική μορφή. Η αποτίμηση της πρότασης γίνεται σύμφωνα με την δεύτερη παράμετρο d που είναι το λεξικό με τις αληθοτιμές True και False.

**Παράδειγμα**

Η λογική έκφραση $(a \lor b) \land (a \land c)$ μπορεί να παρουσιαστεί σε μορφή δέντρου:

         ∧
       /   \
      ∨    ∧
     / \   / \
    a   b a   c

Οπότε μπορεί να δοθεί και με μορφή λίστας t (σε δενδρική μορφή) ως εξής:

```python
t = ['&', ['|', 'a', 'b'], ['&', 'a','c']]
d = {'a' : True, 'b' : False, 'c' : True}
sage.logic.booleval.eval_formula(t, d)
```

## Ικανοποιησιμότητα, ταυτολογία και αντίφαση

Μια πρόταση είναι **ικανοποιήσιμη** εάν υπάρχει κάποια αποτίμηση, που την ικανοποιεί, που κάνει δηλαδή την πρόταση Αληθή. Η μέθοδος του SageMath που ελέγχει την ικανοποιησιμότητα της έκφρασης είναι η `is_satisfiable`.

Έτσι, η πρόταση $a \land b$ είναι ικανοποιήσιμη. Αν μάλιστα δούμε τον πίνακα αληθείας θα δούμε ότι υπάρχει a και b για τα οποία η πρόταση ισχύει.

```python
f = propcalc.formula("a&b")
f.is_satisfiable()
```

```python
f.truthtable()
```

Αντίθετα, η πρόταση $a \land \lnot a$ είναι μη ικανοποιήσιμη, κάτι που φαίνεται και από τον πίνακα αλήθειας.

```python
f = propcalc.formula("a&~a")
f.is_satisfiable()
```

```python
f.truthtable()
```

Μια πρόταση είναι **ταυτολογία** αν είναι αληθής για όλες τις δυνατές αναθέσεις τιμών στις μεταβλητές της. Η ταυτολογία ελέγχεται με τη μέθοδο `is_tautology`.

Ελέγχοντας την λογική έκφραση $a \lor b$ βλέπουμε ότι δεν πρόκειται για ταυτολογία. Ελέγχοντας και τον σχετικό πίνακα αλήθειας βλέπουμε ότι στην περίπτωση που και οι δύο μεταβλητές έχουν την τιμή False η πρόταση είναι και αυτή False.

```python
f = propcalc.formula("a|b")
f.is_tautology()
```

```python
f.truthtable()
```

Η λογική πρόταση $a \lor \lnot a$ έχει πάντα την τιμή True ανεξάρτητα της τιμής της a και ως εκ τούτου αποτελεί ταυτολογία.

```python
f = propcalc.formula("a|~a")
f.is_tautology()
```

```python
f.truthtable()
```

Μια λογική έκφραση αποτελεί **αντίφαση** αν δεν υπάρχει καμία ανάθεση τιμών που να την καθιστά αληθή. Ουσιαστικά όταν η πρόταση δεν είναι ικανοποιήσιμη, είναι αντίφαση. Με την μέθοδο `is_contradiction` ελέγχουμε για αντιφάσεις.

```python
f = propcalc.formula("a&~a")
f.is_contradiction(),f.is_satisfiable()
```

Η μέθοδος `A.is_consequence(B)`, ή εναλλακτικά `is_consequence(A, B)` εξετάζει αν το A είναι μια **λογική συνέπεια** των υποθέσεων B. Εξετάζει το σύνολο των δυνατών καταστάσεων αληθείας για τις λογικές μεταβλητές και ελέγχει αν σε κάθε κατάσταση στην οποία η B είναι αληθής, η A είναι επίσης αληθής.

```python
A = propcalc.formula("P | Q")
B = propcalc.formula("P")
A.is_consequence(B)
```

```python
A = propcalc.formula("P & Q")
B = propcalc.formula("P")
A.is_consequence(B)
```

Οι υποθέσεις, μπορεί να είναι και περισσότερες από μια. Τις διαχωρίζουμε με κόμματα μέσα στο `is_consequence`.

```python
propcalc.formula("P").is_consequence(propcalc.formula("Q"), \
                                propcalc.formula("Q -> P"))
```

```python
propcalc.formula("P").is_consequence(propcalc.formula("Q"), \
                                propcalc.formula("P -> Q"))
```

## Λογική ισοδυναμία

Δύο εκφράσεις είναι **ισοδύναμες** αν έχουν ακριβώς τα ίδια αποτελέσματα, `True` ή `False`, για όλες τις πιθανές τιμές των μεταβλητών που περιέχουν. Το σύμβολο $\equiv$ δείχνει ότι δύο προτάσεις είναι ισοδύναμες και η μέθοδος `equivalent` ελέγχει την ισοδυναμία στο SageMath.

**Παράδειγμα**

Θα δείξουμε ότι οι προτάσεις $(p \land q) \lor (\neg p \land q)$ και $q$ είναι ισοδύναμες με χρήση της `equivalent` αλλά και πίνακα αληθείας.
$$(p \land q) \lor (\neg p \land q) \equiv q$$.

```python
lhsp = propcalc.formula("(p & q) | (~p & q)")
rhsp = propcalc.formula("q")

lhsp.equivalent(rhsp)
```

Παρατηρούμε στον επόμενο πίνακα αληθείας ότι οι στήλες αποτίμησης της $(p \land q) \lor (\neg p \land q)$ και η στήλη $q$ ταυτίζονται.

```python
lhsp.truthtable()
```

## Ιδιότητες των λογικών πράξεων

1. Προσεταιριστική ιδιότητα:
    - $(A \lor B) \lor C \equiv A \lor (B \lor C)$

    - $(A \land B) \land C \equiv A \land (B \land C)$

```python
lhs = propcalc.formula("(A | (B | C))")
rhs = propcalc.formula("((A | B) | C)")
lhs.equivalent(rhs)
```

```python
lhs = propcalc.formula("(A & (B & C))")
rhs = propcalc.formula("((A & B) & C)")
lhs.equivalent(rhs)
```

2. Μεταθετική ιδιότητα:

    - $A \lor B \equiv B \lor A$
    
    - $A \land B \equiv B \land A$

5. Ταυτοτική ιδιότητα:

    - $A \lor \text{False} \equiv A$

    - $A \land \text{True} \equiv A$

Σημείωση: Τα False, True θεωρούνται λογικές μεταβλητές και δεν αναγνωρίζονται ως λογικές σταθερές. Έτσι η `equivalent` δεν λειτουργεί όπως αναμενόταν επιστρέφοντας False αντί για True. Με άλλα λόγια με True η `procalc` δεν θεωρεί μια (οποιαδήποτε) ταυτολογία, αλλά ως μια ατομική πρόταση με το όνομα True! Όμοια συμβαίνουν για το False.

```python
lhs = propcalc.formula("(A | False)")
rhs = propcalc.formula("A")
lhs.equivalent(rhs)
```

```python
lhs = propcalc.formula("(A & True)")
rhs = propcalc.formula("A")
lhs.equivalent(rhs)
```

Ας δούμε και τους αντίστοιχους αληθοπίνακες για να δούμε πώς “σκέφτεται” το `procalc`. Όπως βλέπουμε πιστεύει ότι η πρόταση True μπορεί να γίνει σε κάποιες περιπτώσεις False!

```python
f1 = propcalc.formula("(A & True)")
f2 = propcalc.formula("A")
print(f1.truthtable())
print(f2.truthtable())
```

Μπορούμε όμως να διoρθώσουμε αυτό το σφάλμα με δυο τρόπους. Ο πρώτος τρόπος είναι να αντικαταστήσουμε το True με μια οποιαδήποτε γνωστή ταυτολογία και το False με μια οποιαδήποτε γνωστή αντιλογία.

```python
lhs = propcalc.formula("A & (B | ~B)")
rhs = propcalc.formula("A")
lhs.equivalent(rhs)
```

Το \b σημαίνει όριο λέξης (word boundary), δηλαδή: δεν υπάρχει άλλο γράμμα ή αριθμός δίπλα αλλά σημείο στίξης ή κενό!

## Σχέσεις μεταξύ συνόλων

| **Σχέση Συνόλων**       | **Περιγραφή**                                 | **SageMath Συνάρτηση** |
|-------------------------|-----------------------------------------------|------------------------|
| Υποσύνολο, Subset       | Έλεγχος αν το A είναι υποσύνολο του B         | `A.issubset(B)`, `A<=B`       |
| Γνήσιο υποσύνολο, Proper subset       | Έλεγχος αν το A είναι γνήσιο υποσύνολο του B         | `A<B`       |

## Πράξεις με σύνολα

Όλες οι γνωστές πράξεις συνόλων υλοποιούνται είτε με αντίστοιχες μεθόδους είτε με υπερφορτωμένους τελεστές.

| **Πράξη Συνόλου**       | **Περιγραφή**                                 | **SageMath Συνάρτηση** |
|-------------------------|-----------------------------------------------|------------------------|

Αν επιθυμούμε να μοιράσουμε φύλλα και σε επόμενο παίκτη, δεν θα πρέπει να καλέσουμε άμεσα την `random_element` από την _Hands_. Θα πρέπει να αφαιρέσουμε πρώτα τα φύλλα που πήρε ο πρώτος παίκτης ώστε αυτά να μην επαναλαμβάνονται. Η `difference`, ή ο υπερφορτωμένος τελεστής `-`, βρίσκουν την διαφορά συνόλων.

```python
remCards = Set(Cards) - h1
remHands   = Subsets(remCards, 5)
h2 = remHands.random_element();h2
```

## Κατασκευή ενός υποσυνόλου με κάποια συνθήκη

Με κατάλληλη χρήση comprehension μπορούμε να φιλτράρουμε εκείνα τα στοιχεία ενός συνόλου $𝐷$ που ικανοποιούν μια συγκεκριμένη συνθήκη. Για παράδειγμα από το σύνολο $bb$, που δημιουργήσαμε παραπάνω, μπορούμε να διαλέξουμε όλες τις τριάδες με άρτιο πλήθος από 1.

```python
bit = Set([0,1])
bb = cartesian_product([bit,bit,bit])
Set([x for x in bb if sum(x) % 2 == 0])
```

Μπορούμε ενναλακτικά να χρησιμοποιήσουμε κατάλληλα την `filter`.

```python
bit = Set([0, 1])
bb = cartesian_product([bit] * 3)
list(filter(lambda x: sum(x) % 2 == 0, bb))
```

Όπως βλέπουμε η `filter` δεν δημιουργεί αμέσως ολόκληρη λίστα στη μνήμη. Η `filter` είναι φτιαγμένη για αποδοτικότητα και οικονομία! Αντί γι’ αυτό, επιστρέφει έναν τεμπέλικο (lazy) iterator, ο οποίος αποδίδει τα στοιχεία μόνο όταν ζητηθούν, π.χ. μέσα σε κάποιο for-loop ή με `list`.

## Χαρακτηριστική συνάρτηση ενός συνόλου

Η χαρακτηριστική συνάρτηση ενός συνόλου A με 𝐴 ⊆ 𝑋 είναι μια συνάρτηση που δείχνει αν ένα στοιχείο του, x του X ανήκει στο A ή όχι. Αν το x ανήκει στο A παίρνει τιμή 1, διαφορετικά 0.

## Διωνυμικός συντελεστής

Ο **διωνυμικός συντελεστής**, $\binom{n}{k}$, δίνει τον αριθμό των τρόπων επιλογής _k_ αντικειμένων από ένα σύνολο _n_ αντικειμένων, χωρίς να λαμβάνεται υπόψη η σειρά):

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

Για παράδειγμα, για να επιλέξουμε 3 αντικείμενα από 5;
$$
\binom{5}{3} = \frac{5!}{3!(5-3)!} = \frac{5 \cdot 4 \cdot 3!}{3! \cdot 2!} = \frac{5 \cdot 4}{2 \cdot 1} = 10
$$

Ο υπολογισμός του $\binom{n}{k}$ στο SageMath γίνεται με την `binomial`.

```python
binomial(5, 2)
```

Το παραπάνω μπορεί να βρεθεί και με πράξεις συνόλων.

```python
S = Subsets([1,2,3,4,5], 2); S.cardinality()
```

### Eulerian

Στην πόλη Königsberg (σημερινό Kaliningrad, Ρωσία), ο ποταμός Pregel χώριζε την πόλη σε 4 περιοχές, δύο νησιά στον ποταμό και τις δύο όχθες [Seven Bridges of Königsberg](https://en.wikipedia.org/wiki/Seven_Bridges_of_K%C3%B6nigsberg). Οι περιοχές συνδέονταν με 7 γέφυρες. Κάτοικοι της πόλης θέσανε το ερώτημα αν είναι δυνατόν κάποιος κάνοντας περίπατο να περάσει και από τις 7 γέφυρες μια και μόνο μια φορά. Το πρόβλημα γνωστοποιήθηκε στον Ελβετό μαθηματικό Leonhard Euler που έκανε στάση σε κάποιο από τα ταξίδια του στο Königsberg. Ο Euler απέδειξε ότι το πρόβλημα δεν έχει λύση θέτοντας, παράλληλα, και τις βάσεις για έναν ολόκληρο κλάδο των μαθηματικών γνωστό ως θεωρία γράφων.

Στο επόμενο γράφο 'A' μπορεί να θεωρηθεί το νησί που έχει 2 γέφυρες με την βόρεια όχθη ('A','B'), ('A','B'), 2 γέφυρες με την νότια όχθη ('A','C'), ('A','C') και μια με το 2ο νησί ('A','D'). Υπάρχει και μια γέφυρα που ενώνει το 2ο νησί με την βόρεια όχθη ('B','D') καθώς και άλλη μια με την νότια ('C','D'). Επειδή υπάρχουν κορυφές που συνδέονται μεταξύ τους με παραπάνω από δύο ακμές, γέφυρες θα πρέπει να τεθεί η παράμετρος `multiedges=True`.

```python
edges = [('A','B'), ('A','B'),
             ('A','C'), ('A','C'),
             ('A','D'),
             ('B','D'), 
             ('C','D')]
K = Graph(edges, multiedges=True)
K.show(figsize=(3,2))
```

Ο βαθμός της κάθε κορυφής είναι περιττός αριθμός.

```python
K.degree()
```

Σύμφωνα και με την σκέψη του Euler, ένας περιπατητής θα πρέπει, με εξαίρεση την αρχική και την τελική κορυφή να εισέλθει και να εξέλθει των κορυφών και άρα απαιτείται άρτιος αριθμός ακμών. Οι γράφοι των οποίων μπορούμε να διασχίσουμε όλες τις ακμές από μια φορά ονομάζονται **eulerian** ή **semi-eulerian** ανάλογα με το αν με την διάσχιση μπορούμε να επιστρέψουμε στην αρχική θέση ή όχι. Το μονοπάτι που περνάει μια και μόνο μια φορά από κάθε ακμή του γράφου ονομάζεται **eulerian path**. Αν το μονοπάτι καταλήγει εκεί που ξεκίνησε πρόκειται για **eulerian cycle**.

Η μέθοδος `is_eulerian` επιστρέφει αν ο γράφος είναι eulerian. Προφανώς, στην περίπτωση του Königsberg δεν είναι.

```python
K.is_eulerian()
```

Κατά τον 2ο παγκόσμιο πόλεμο το Königsberg ισοπεδώθηκε και στην θέση του σήμερα βρίσκεται το Kaliningrad. Εκεί, στις αντίστοιχες τοποθεσίες, όπως βλέπουμε και στο αντίστοιχο στιγμιότυπο από το [open street map](https://www.openstreetmap.org/) υπάρχουν δύο γέφυρες λιγότερες τις οποίες και θα αφαιρέσουμε από τον αρχικό γράφο.
![Kaliningrad](kaliningrad.jpg)

```python
K.delete_edge('A', 'B')
K.delete_edge('A', 'C')
```

Παρατηρούμε και στο στιγμιότυπο αλλά και από τους βαθμούς των κορυφών ότι, πλέον, υπάρχουν μόνο δύο κορυφές με περιττό βαθμό. Σύμφωνα με τον Euler ο γράφος είναι **semi-eulerian**.

```python
K.degree(),K.show(figsize=(3,2))
```

Ο γράφος δεν είναι eulerian αλλά υπάρχει μονοπάτι που περνάει από όλες τις ακμές μια και μόνο μια φορά που όμως δεν επιστρέφει στην ίδια κορυφή από την οποία ξεκίνησε. Η `is_eulerian` με παράμετρο `path=True` επιστρέφει, αν υπάρχει, αρχή και τέλος μονοπατιού Euler.

```python
K.is_eulerian(),K.is_eulerian(path=True)
```

### Hamiltonian

Αρκετά χρόνια μετά τον Euler ο Ιρλανδός μαθηματικός William Rowan Hamilton δημιούργησε το παιχνίδι [icosian game](https://en.wikipedia.org/wiki/Icosian_game). ΤΟ παιχνίδι αφορούσε την εύρεση **Hamiltonian κύκλου** σε ένα δωδεκάεδρο (dodecahedron) στο οποίο οι κορυφές αντιπροσώπευαν μεγάλες πόλεις του κόσμου και ο στόχος ήταν να επισκεφτείς κάθε πόλη ακριβώς μία φορά και να επιστρέψεις στην αρχική. Ενώ ο Euler απέδειξε την ύπαρξη ή μη euler μονοπατιού ή κύκλου, για το πρόβλημα εύρεσης hamiltonian μονοπατιού ή κύκλου δεν υπάρχει αποδοτικός τρόπος επίλυσης.

```python
H = graphs.DodecahedralGraph()
H.plot(figsize=(4,3))
```

Μελετώντας αριθμό κορυφών, `order`, ακμών, `size`, και βαθμούς, `degree`, βλέπουμε ότι o γράφος του δωδεκαέδρου είναι ένας γράφος με:

- 20 κορυφές

- 30 ακμές

- βαθμό 3 σε όλες τις κορυφές

```python
H.order(), H.size(), H.degree() 
```

Η μέθοδος `is_hamiltonian` που ακολουθεί ελέγχει αν υπάρχει κύκλος που περνά από όλες τις κορυφές.

```python
H.is_hamiltonian()
```

Όπως είναι προφανές, το γράφημα μονοπατιού, `PathGraph`, δεν έχει κύκλους ενώ ο κυκλικός γράφος, `CycleGraph`, έχει.

```python
PG=graphs.PathGraph(5)
CG = graphs.CycleGraph(5)
PG.is_hamiltonian(),CG.is_hamiltonian()
```

Αντίστοιχα με τα μονοπάτια και τους κύκλους Euler, υπάρχουν και μονοπάτια και κύκλοι Hamilton. Η `hamiltonian_cycle` θα επιστρέψει, αν υπάρχει, έναν τέτοιο κύκλο.

## Αποστάσεις

Οι έννοιες της **απόστασης**, της **εκκεντρότητας**, της **ακτίνας**, της **διαμέτρου**, του **κέντρου** και της **περιφέρειας** ενός γράφου με τις οποίες θα ασχοληθούμε στην παρούσα ενότητα αποτελούν βασικά εργαλεία για την κατανόηση της δομής και της «γεωμετρίας» των δικτύων. Μέσω αυτών ποσοτικοποιείται το πόσο εκτεταμένος είναι ένας γράφος, πόσο απομακρυσμένες μπορεί να είναι μεταξύ τους οι κορυφές του, καθώς και ποιες κορυφές κατέχουν κεντρικό ή περιφερειακό ρόλο ως προς τις αποστάσεις. ## Σταθμισμένοι Γράφοι

Οι σταθμισμένοι γράφοι **weighted graphs** είναι γράφοι όπου σε κάθε ακμή αντιστοιχεί μια αριθμητική τιμή, γνωστή ως βάρος **weight**. Αυτό το βάρος μπορεί να αντιπροσωπεύει, για παράδειγμα, απόσταση, κόστος, ή χωρητικότητα.

### Δημιουργία

Στο SageMath, ένα σταθμισμένο γράφημα είτε κατευθυνόμενο, `DiGraph`, είτε μη κατευθυνόμενο, `Graph`, μπορεί να δημιουργηθεί με μια λίστα τριάδων, (u, v, w) με τις δύο πρώτες τιμές, u και v, να αντιπροσωπεύουν τις κορυφές που ενώνει η ακμή και την τρίτη, w, το βάρος της. Για να φαίνεται το βάρος της κάθε ακμής θα πρέπει να χρησιμοποιηθεί η `plot` με παράμετρο `edge_labels=True`.

```python
G = Graph([(1, 2, 6), (1, 3, 2), (2, 3, 1), (3, 4, 3)])
G.plot(edge_labels=True,figsize=(3,2))
```

Η μέθοδος `edges` θα μας επιστρέψει λίστα από tuples τριών στοιχείων.

```python
edges = G.edges();edges
```

Εφόσον γνωρίζουμε ότι το βάρος κάθε ακμής είναι το τρίτο στοιχείο σε κάθε `tuple`, εύκολα μπορούμε να βρούμε το συνολικό βάρος του γράφου.

```python
sum(w for u, v, w in edges)
```

Η μετατροπή μιας μήτρας γειτνίασης σε σταθμισμένο γράφημα είναι μια πολύ απλή διαδικασία, καθώς ο κατασκευαστής `Graph()` αναγνωρίζει αυτόματα τις τιμές του πίνακα ως βάρη.

```python
A = matrix([
    [0,6,8,0,2,0],
    [6,0,5,9,0,8],
    [8,5,0,0,3,0],
    [0,9,0,0,0,7],
    [2,0,3,0,0,4],
    [0,8,0,7,4,0]
])
G = Graph(A, weighted=True)
G.plot(edge_labels=True,figsize=(3,2))
```

### Συντομότερο Μονοπάτι 

Η μέθοδος `G.distance(u, v, by_weight=True)` μας δίνει τη διαδρομή με το μικρότερο συνολικό άθροισμα βαρών μεταξύ δύο κορυφών _u_ και _v_. Όπως έχουμε αναφέρει, η `shortest_path` επιστρέφει τη συντομότερη διαδρομή. αν ο γράφος είναι σταθμισμένος, θα πρέπει να συμπεριλάβουμε την παράμετρο `by_weight=True`. Αν δεν την συμπεριλάβουμε θα επιστραφεί η συντομότερη διαδρομή σε ότι αφορά αριθμό ακμών ανεξαρτήτως βάρους.

```python
G.distance(2,5, by_weight=True),\
        G.shortest_path(2, 5, by_weight=True), \
        G.shortest_path(2, 5)
```

Ο κλασικότερος αλγόριθμος που βρίσκει το συντομότερο μονοπάτι από μια κορυφή-πηγή σε όλες τις άλλες κορυφές σε έναν σταθμισμένο γράφο είναι ο αλγόριθμος του **Dijkstra**.

### Ελάχιστο Συνδετικό Δέντρο

Το **Minimum Spanning Tree**, ελάχιστο συνδετικό δέντρο, αποτελεί ένα υπογράφημα που συνδέει όλες τις κορυφές με το ελάχιστο δυνατό συνολικό βάρος ακμών. Η μέθοδος `min_spanning_tree` επιστρέφει λίστα ακμών.

```python
MST = G.min_spanning_tree()
MSTG = Graph(MST)
MSTG.plot(edge_labels=True,figsize=(3,2))
```

Θα σχεδιάσουμε τον αρχικό γράφο G με διαφορετικά χρώματα στις ακμές του δέντρου.

```python
#Ακμές του MST χωρίς τα βάρη
mst_edges = [(u, v) for u, v, w in MST]
# οι υπόλοιπες ακμές
all_edges = [(u, v) for u, v, w in G.edges()]
non_mst_edges = [e for e in all_edges if e not in mst_edges]
edge_colors = {
    "red": mst_edges,
    "gray": non_mst_edges
}
G.plot(edge_colors=edge_colors, 
       edge_labels=True,
          figsize=(3,2))
```
