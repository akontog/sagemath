## Monty Hall

Το Monty Hall problem, https://en.wikipedia.org/wiki/Monty_Hall_problem, αποτελεί διάσημο πρόβλημα που προέκυψε από το τηλεπαιχνίδι Let's make a Deal που ξεκίνησε την δεκαετία του 60 στην Αμερική.

Στο παιχνίδι υπάρχουν 3 πόρτες όπου:

- πίσω από 1 πόρτα βρίσκεται ένα αυτοκίνητο (που κερδίζει ο παίκτης),

- πίσω από 2 πόρτες υπάρχουν κατσίκες και ο παίκτης δεν κερδίζει τίποτα.

Ο παίκτης επιλέγει αρχικά μία πόρτα. Ο παρουσιαστής ανοίγει μία άλλη πόρτα που έχει κατσίκα. Στην συνέχεια ο παίκτης έχει δύο στρατηγικές:

1. Μένει στην αρχική επιλογή.

2. Αλλάζει επιλογή στην άλλη κλειστή πόρτα.

Με χρήση `slider` ο χρήστης επιλέγει τον αριθμό προσομοιώσεων (100–5000).

Να υπολογιστούν:

- Η πειραματική πιθανότητα επιτυχίας όταν μένει στην επιλογή του.

- Η πειραματική πιθανότητα επιτυχίας όταν αλλάζει επιλογή.
