# Πιθανότητες

Ο πληθάριθμος ενός συνόλου, ο αριθμός δηλαδή των στοιχείων του, δίνεται είτε με την μέθοδο `cardinality` του SageMath είτε με την `len`. Η τελευταία μας επιστρέφει το μήκος κάθε συλλογής, άρα και του συνόλου.

```python
Ω = Set([1, 2, 3, 4, 5])
len(Ω), Ω.cardinality()
```

Αν υποθέσουμε ότι θέλουμε να βρούμε την πιθανότητα $P(\leq 3) = P(1 \lor 2 \lor 3)$ στο παραπάνω σύνολο.

```python
A = Set([1, 2, 3])
P = A.cardinality() / Ω.cardinality()
P
```

**Παράδειγμα**

Έστω δίκαιο ζάρι το οποίο και ρίχνουμε δύο φορές. Μπορούμε να δημιουργήσουμε το σύνολο, δηλαδή τον δειγματικό χώρο $\Omega = \{(i, j) \,|\, i, j \in \{1, 2, 3, 4, 5, 6\}\}$, με comprehension. Στην συνέχεια θα υπολογίσουμε τις επόμενες πιθανότητες:

- Πιθανότητα να το άθροισμα των δύο ζαριών να είναι ίσο με 7

- Πιθανότητα να το άθροισμα των δύο ζαριών να είναι μεγαλύτερο ή ίσο με 9

- Πιθανότητα το πρώτο ζάρι να είναι 6

- Πιθανότητα το δεύτερο ζάρι να είναι μεγαλύτερο από 4

```python
Ω = Set([(x, y) for x in range(1, 7) for y in range(1, 7)])
show(Ω)
```

```python
sum7 = Set([(x, y) for (x, y) in Ω if x + y == 7])
sum7.cardinality() / Ω.cardinality()
```

```python
sumgr9 = Set([(x, y) for (x, y) in Ω if x + y >= 9])
sumgr9.cardinality() / Ω.cardinality()
```

```python
first6 = Set([(x, y) for (x, y) in Ω if x == 6])
first6.cardinality() / Ω.cardinality()
```

```python
secondgr4 = Set([(x, y) for (x, y) in Ω if y > 4])
secondgr4.cardinality() / Ω.cardinality()
```

## Διάγραμμα διασποράς, scatter plot

Ιδανικός τύπος διαγράμματος για την παρουσίαση της σχέσης μεταξύ δύο μεταβλητών αποτελεί το διάγραμμα διασποράς. Κάθε σημείο στο διάγραμμα αντιστοιχεί σε ένα ζεύγος τιμών, με τους άξονες x και y να αντιπροσωπεύουν τις δύο διαφορετικές μεταβλητές των οποίων την σχέση θέλουμε να μελετήσουμε.

Η μέθοδος του SageMath για την παράσταση διαγράμματος διασποράς είναι η `scatter_plot` η οποία δέχεται λίστα από tuples δύο στοιχείων. Αν τα δεδομένα μας είναι για παράδειγμα δύο ξεχωριστές λίστες θα πρέπει να τα φέρουμε στην αποδεκτή μορφή. Αυτό μπορεί να γίνει με διάφορους τρόπους όπως μία δομή επανάληψης, η τεχνική list comprehension αλλά και η συνάρτηση Python `zip`.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 4, 6, 5, 7, 8, 9]

Lzip = list(zip(x, y))
Lfor = []
for i in range(len(x)):
    Lfor.append((x[i], y[i]))
Lcomp = [(x[i], y[i]) for i in range(len(x))]
scatter_plot(Lzip)
Lzip,Lfor,Lcomp
```

Αφού τα έχουμε φέρει στην αποδεκτή μορφή καλούμε την `scatter_plot`.

```python
scatter_plot(Lzip)
```

## Ιστόγραμμα, histogram

Το ιστογράμματα, είναι παρόμοια με τα ραβδογράμματα αλλά χρησιμοποιούνται για να παρουσιάσουν τη συχνότητα κατανομής ενός συνόλου δεδομένων. Οι τιμές του συνόλου χωρίζονται σε ομάδες ή κλάσεις, με τη συχνότητα κάθε κλάσης να απεικονίζεται με μπάρες.

Για να δημιουργηθεί στο Sage ένα ιστόγραμμα καλείται η `histogram` με μία λίστα τιμών. Η `bins` Θέτει τα τμήματα στα οποία θα διαχωριστούν τα δεδομένα. Έτσι, το επόμενο παράδειγμα που έχει χωριστεί σε 5 τμήματα θα χωρίσει τα δεδομένα σε {0,1}, {2,3} κ.ο.κ.

```python
data = [0, 1, 2, 2, 3, 3, 3, 4, 4, 5, 6, 6, 7, 7, 7, 7, 8, 9, 9, 10]
histogram(data, bins=5)
```

Η `bins` μπορεί να διαχωρίσει τα δεδομένα με μία σειρά από σημεία τα οποία θα τεθούν από τον χρήστη με την χρήση μίας λίστας.

Η `histogram` δέχεται μία σειρά από παραμέτρους για την μορφοποίηση της εμφάνισης των γραφημάτων όπως `color` και `edgecolor` για το χρώμα `title` και `axes_labels` για τίτλους γραφήματος και αξόνων.

## Συνεχείς κατανομές 

Οι (ψευδο)τυχαίοι αριθμοί που μπορούν να παραχθούν από το SageMath μπορούν να ακολουθούν συνεχείς ή διακριτές κατανομές. Υποστηρίζονται οι συνηθέστερες συνεχείς και για επιπλέον κατανομές μπορεί να χρησιμοποιηθούν και οι βιβλιοθήκες `NumPy` ή `SciPy` οι οποίες θα πρέπει να εισαχθούν πριν χρησιμοποιηθούν.


### Κανονική κατανομή, gaussian

Η κανονική, `gaussian`, κατανομή αναφέρεται σε συνεχείς μεταβλητές και χρησιμοποιείται για να περιγραφούν μεταβλητές πραγματικών τιμών. Οι τιμές συγκεντρώνονται κυρίως γύρω από μια μέση τιμή, _μ_ ανάλογα την διακύμανσή τους $σ^2$. Η κανονική κατανομή είναι πολύ σημαντική στη φύση καθώς την ακολουθούν τα περισσότερα συνεχή φαινόμενα.

Η `RealDistribution` με παράμετρο `gaussian` επιστρέφει τιμές που ακολουθούν την κατανομή. Μπορούμε επιπλέον να θέσουμε την διασπορά.

**Παράδειγμα**

Θα δημιουργήσουμε 1000 αριθμούς που ακολουθούν την κανονική κατανομή, αρχικά με διασπορά 1. Επιπλέον θα θέσουμε μία μετατόπιση στους τυχαίους αριθμούς _m_. Μπορούμε να πειραματιστούμε με τους παραπάνω αριθμούς και να παρατηρήσουμε τις τιμές που προκύπτουν στο ιστόγραμμα.

```python
m=0
X = RealDistribution('gaussian', 1)
data = [X.get_random_element()+m for i in range(1000)]
h = histogram(data, bins=30, density=True, color='skyblue', title="Κανονική Κατανομή")
h.show()
```

Εάν επιθυμούμε να παράγονται συνεχώς οι ίδιοι ψευδοτυχαίοι αριθμοί, έχουμε την δυνατότητα να θέσουμε τον αρχικό σπόρο από τον οποίο θα ξεκινά η ακολουθία των αριθμών με την `set_seed`. Στο παράδειγμα, όσες φορές και να το εκτελέσουμε, θα παρατηρήσουμε το ίδιο ακριβώς ιστόγραμμα.

```python
m=0
X = RealDistribution('gaussian', 1)
X.set_seed(42)
data = [X.get_random_element()+m for i in range(1000)]
h = histogram(data, bins=30, density=True, color='skyblue', title="Κανονική Κατανομή")
h.show()
```

### Διωνυμική κατανομή, binomial

Η διωνυμική κατανομή `binomial` περιγράφει την πιθανότητα να συμβούν ακριβώς k επιτυχίες σε n ανεξάρτητα πειράματα, όταν η πιθανότητα επιτυχίας σε κάθε πείραμα είναι _p_. Η διωνυμική κατανομή χρησιμοποιείται σε καταστάσεις όπου υπάρχει ένα σύνολο από δυαδικές επιλογές, όπως η ρίψη ενός νομίσματος, κορόνα ή γράμματα, ή η επιτυχία σε ένα τεστ. 
Η πιθανότητα P(X=k) για k επιτυχίες σε n πειράματα δίνεται από τον τύπο:
$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$
Ο διωνυμικός συντελεστής $\binom{n}{k}$ δίνεται από τον τύπο:
$$
\binom{n}{k} = \frac{n!}{k! (n-k)!}
$$
Έτσι για παράδειγμα:
$$
\binom{5}{2} = \frac{5!}{2! (5-2)!} = \frac{5 \cdot 4 \cdot 3!}{2 \cdot 1 \cdot 3!} = \frac{5 \cdot 4}{2 \cdot 1} = 10
$$

Υπάρχει και μέθοδος στο Sage που επιστρέφει τον διωνυμικό συντελεστή, η `binomial`.

```python
binomial(5, 2)
```

Η κατανομή poisson χρησιμοποιείται για την περιγραφή του αριθμού συμβάντων που συμβαίνουν σε ένα σταθερό χρονικό διάστημα ή χώρο, όταν τα συμβάντα είναι ανεξάρτητα μεταξύ τους και η μέση συχνότητα είναι γνωστή. Παραδείγματα εφαρμογής της κατανομής poisson, ο αριθμός πελατών που φτάνουν σε ένα κατάστημα ανά ώρα, των τηλεφωνικών κλήσεων που λαμβάνει ένα κέντρο εξυπηρέτησης κ.ο.κ.

Ο τύπος πιθανότητας για την κατανομή Poisson είναι:
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

- $\lambda$: η μέση συχνότητα των συμβάντων

- $k$: ο αριθμός των συμβάντων $k = 0, 1, 2, \dots$

- $e$: η βάση του φυσικού λογαρίθμου $e \approx 2.718$

## t-test

Το t-test χρησιμοποιείται για να εξετάσει αν οι μέσες τιμές δύο δειγμάτων έχουν στατιστικά σημαντική διαφορά. Τα δείγματα μπορεί να είναι ανεξάρτητα, για παράδειγμα τα ύψη των ανθρώπων σε δύο διαφορετικές χώρες ή συσχετισμένα, για παράδειγμα αιματολογικές μετρήσεις στην ίδια ομάδα ανθρώπων πριν και μετά την λήψη κάποιου φαρμάκου. Ο έλεγχος επιστρέφει αν η διαφορά μεταξύ των μέσων είναι στατιστικά σημαντική.

Στο SageMath χρησιμοποιούμε τρίτες βιβλιοθήκες για την διενέργεια του t-test. Στην βιβλιοθήκη `SciPy` υπάρχει η μέθοδος ttest_ind η οποία επιστρέφει:

- `t-statistic`: Μας λέει πόσο μεγάλη είναι η διαφορά ανάμεσα στα δείγματα σε σχέση με την τυπική απόκλιση. $t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}}$

    - $\bar{x}$: Η μέση τιμή του δείγματος.

    - $\mu_0$: Η υπόθεση για τη μέση τιμή του πληθυσμού (συνήθως 0, μέση τιμή δείγματος = μέση τιμή διαφοράς).

    - s: Η τυπική απόκλιση του δείγματος.

    - n: Το μέγεθος του δείγματος.


- `p-value`: Μας λέει την πιθανότητα να έχουμε τα παρατηρούμενα δεδομένα. Αν το p-value είναι μικρότερο από ένα καθορισμένο όριο, συνήθως 0.05, μπορούμε να υποστηρίξουμε ότι τα δεδομένα μας παρέχουν στατιστικά σημαντική απόδειξη για μια διαφορά μεταξύ υπόθεσης και πληθυσμού.

**Παράδειγμα**

Έστω ότι έχουμε δύο δείγματα υψών από διαφορετικές χώρες και θέλουμε να εξετάσουμε αν οι μέσες τιμές τους διαφέρουν σημαντικά.

```python
from scipy.stats import ttest_ind

heights_country1 = [160, 165, 170, 175, 180, 185, 190, 195, 200, 155, 162, 168, 172, 178, 183, 188, 193, 198, 202, 207]
heights_country2 = [155, 160, 165, 170, 175, 180, 185, 190, 195, 200, 210, 215, 220, 225, 230, 235, 240, 245, 250, 255]

t_stat, p_value = ttest_ind(heights_country1, heights_country2)

# Εμφάνιση του αποτελέσματος
print("T-Statistic:", t_stat)
print("P-value:", p_value)

# Απόφαση βασισμένη στο p-value
if p_value < 0.05:
    print("Η διαφορά μεταξύ των δύο χωρών είναι στατιστικά σημαντική.")
else:
    print("Δεν υπάρχει στατιστικά σημαντική διαφορά μεταξύ των δύο χωρών.")
```

## Συνδιακύμανση, covariance

Η συνδιακύμανση ορίζει την κατεύθυνση της σχέσης μεταξύ δύο μεταβλητών την αναλογικότητα, δηλαδή, μεταξύ των μεταβλητών. Αν οι τιμές των δύο μεταβλητών αυξάνονται ή μειώνονται ταυτόχρονα, τότε η συνδιακύμανση είναι θετική. Αν μία μεταβλητή αυξάνεται ενώ η άλλη μειώνεται, τότε η συνδιακύμανση είναι αρνητική. Υπολογίζεται ως το άθροισμα των γινομένων των διαφορών από τους μέσους όρους, διαιρούμενο με το μέγεθος του δείγματος:
$$
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} \left( X_i - \bar{X} \right) \left( Y_i - \bar{Y} \right)
$$
Ανάλογα με την τιμή της μπορούμε να εξάγουμε τα επόμενα συμπεράσματα:

- Θετική συνδιακύμανση: Όταν η μία μεταβλητή αυξάνεται, η άλλη τείνει επίσης να αυξάνεται.

- Αρνητική συνδιακύμανση: Όταν η μία μεταβλητή αυξάνεται, η άλλη τείνει να μειώνεται.

- Μηδενική συνδιακύμανση: Οι μεταβλητές είναι ανεξάρτητες (δεν σχετίζονται γραμμικά).

Στο SageMath ο υπολογισμός μπορεί να γίνει προγραμματιστικά, πρώτα υπολογίζοντας τις μέσες τιμές και στην συνέχεια εφαρμόζοντας το άθροισμα των διαφορών των τιμών από το μέσο όρο.

```python
heights = [160, 165, 170, 175, 180, 185, 190, 195, 200, 155, 162, 168, 172, 178, 183, 188, 193, 198, 202, 207]
weights = [55, 60, 75, 70, 75, 80, 85, 90, 95, 50, 60, 57, 68, 74, 79, 84, 80, 92, 105, 100]

mean_heights = mean(heights)
mean_weights = mean(weights)

# Υπολογισμός συνδιακύμανσης
n = len(heights)
covariance = sum((heights[i] - mean_heights) * (weights[i] - mean_weights) for i in range(n)) / n
N(covariance)
```

Εναλλακτικά, μπορούν να χρησιμοποιηθούν βιβλιοθήκες. Η `np.cov` της `NumPy` μας επιστρέφει πίνακα διακυμάνσεων-συνδιακυμάνσεων με τα διαγώνια στοιχεία να αντιπροσωπεύουν τη διακύμανση κάθε μεταβλητής, και τα άλλα δύο την συνδιακύμανση.

```python
import numpy as np

heights_arr = np.array(heights)
weights_arr = np.array(weights)
cov_matrix = np.cov(heights_arr, weights_arr, ddof=0)
covariance = cov_matrix[0, 1]

covariance
```

## Συντελεστής Γραμμικής Συσχέτισης, Pearson

Ένα μέτρο που μας δίνει το μέγεθος της γραμμικής σχέσης μεταξύ δύο μεταβλητών είναι ο συντελεστής γραμμικής συσχέτισης. Ουσιαστικά προκύπτει αν διαιρέσουμε την συνδιακύμανση με το γινόμενο των τυπικών αποκλίσεων. Δηλαδή, με βάση ένα δείγμα n ζευγών παρατηρήσεων $(x_i, y_i) i=1,2,...,n$, συμβολίζεται με $r(X,Y)$ ή απλούστερα r και δίνεται:
$$
r = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^n (y_i - \bar{y})^2}}
$$
Η τιμή του συντελεστή βρίσκεται μεταξύ [-1,1] και για τις τιμές: 

-  0 < r < +1: θετικά γραμμικά συσχετισμένες (r=1: Τέλεια θετική συσχέτιση)

- -1 < r <0: αρνητικά γραμμικά συσχετισμένες (r=−1: Τέλεια αρνητική συσχέτιση)

- r=0: δεν υπάρχει γραμμική συσχέτιση.

Στο SageMath δεν υπάρχει αντίστοιχη μέθοδος και θα πρέπει να υπολογιστεί προγραμματιστικά.

```python
heights = [160, 165, 170, 175, 180, 185, 190, 195, 200, 155, 162, 168, 172, 178, 183, 188, 193, 198, 202, 207]
weights = [55, 60, 75, 70, 75, 80, 85, 90, 95, 50, 60, 57, 68, 74, 79, 84, 80, 92, 105, 100]

mean_x = sum(heights) / len(heights)
mean_y = sum(weights) / len(weights)
numerator = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(heights, weights))
denominator = sqrt(sum((xi - mean_x)^2 for xi in heights) * sum((yi - mean_y)^2 for yi in weights))
pearson_r = numerator / denominator
N(pearson_r)
```

Έχουμε την δυνατότητα, όπως και σε διάφορες άλλες λειτουργίες, να εισάγουμε βιβλιοθήκες όπως η `NumPy` ή η `SciPy` και να καλέσουμε τις αντίστοιχες μεθόδους `corrcoef` ή `pearsonr`.

```python
import numpy as np
pearson_r = np.corrcoef(heights, weights)[0, 1];pearson_r
```

```python
from scipy.stats import pearsonr
pearsonr(heights, weights)
```

## Παλινδρόμηση, regression

Η παλινδρόμηση αποτελεί την στατιστική μέθοδο που χρησιμοποιείται για να βρούμε τη γραμμή που αντιπροσωπεύει όσο το δυνατόν καλύτερα τη σχέση μεταξύ δύο, ή περισσότερων, μεταβλητών, έστω x και y με την πρώτη να αποτελεί την εξαρτημένη μεταβλητή και την 2η την ανεξάρτητη. Βασίζεται στη μείωση του σφάλματος μέσω της μεθόδου των ελαχίστων τετραγώνων. Σημειώνεται ότι οι ανεξάρτητες μεταβλητές, από τις οποίες εξαρτάται η εξαρτημένη, μπορεί να είναι και περισσότερες από μία.

### Γραμμική παλινδρόμηση

**Παράδειγμα**

Θα χρησιμοποιήσουμε τα δεδομένα από το παράδειγμα με τα ύψη και τα βάρη.

```python
heights = [160, 165, 170, 175, 180, 185, 190, 195, 200, 155, 162, 168, 172, 178, 183, 188, 193, 198, 202, 207]
weights = [55, 60, 75, 70, 75, 80, 85, 90, 95, 50, 60, 57, 68, 74, 79, 84, 80, 92, 105, 100]
data = list(zip(heights,weights))
a, b = var('a b')
model(x) = a * x + b  
fits = find_fit(data, model)
fits
```

Θα χρησιμοποιήσουμε τους συντελεστές a και b για να απεικονίσουμε και την συνάρτηση $ax+b$ και τα σημεία. Υπενθυμίζεται ότι η `rhs()` επιστρέφει το δεξί μέλος των εξισώσεων. Τα σημεία μπορούν να απεικονιστούν με την `points` ή με `scatter_plot`.

```python
f(x) = fits[0].rhs()*x + fits[1].rhs()
g = Graphics()
g += plot(f(x),(x,150,210))
g+=scatter_plot(data)
g.show()
```

Όπως είναι φυσικό, ένα μοντέλο μπορεί να έχει παραπάνω από μία ανεξάρτητες μεταβλητές που επηρεάζουν την εξαρτημένη μεταβλητή. Το μοντέλο στην περίπτωση αυτή θα έχει για παράδειγμα μία τέτοια μορφή: $z=a⋅x+b⋅y+c$.

**Παράδειγμα**
Θα μελετήσουμε ένα κλασσικό παράδειγμα παλινδρόμησης τιμολόγησης σπιτιών. Θα υποθέσουμε ότι οι τιμή των σπιτιών σε χιλιάδες ευρώ, λίστα _prices_ εξαρτάται από:

- Απόσταση από το κέντρο της πόλης σε χιλιόμετρα, λίστα _distance_

- Αριθμό τετραγωνικών μέτρων του σπιτιού λίστα _size_

Για ευκολότερη διαχείριση του μοντέλου η `find_fit` μπορεί να επιστρέψει λεξικό αν θέσουμε `solution_dict=False`. 
Στην παραπάνω εξίσωση το z θα είναι η τιμή, το x η απόσταση και το y τα τετραγωνικά.

```python
distances = [1, 3, 2, 6, 4, 8, 5, 7, 10, 9, 3.5, 4.5, 2.5, 6.5, 5.5]
sizes = [60, 80, 55, 120, 90, 110, 100, 105, 140, 130, 85, 95, 70, 115, 108]
prices = [280, 260, 270, 180, 230, 150, 200, 160, 120, 135, 250, 240, 275, 190, 210]

data = list(zip(distances, sizes, prices))

x, y, a, b, c = var('x y a b c')
model(x, y) = a * x + b * y + c

fits = find_fit(data, model, solution_dict=True)
f(x,y) = fits[a]*x + fits[b]*y+fits[c]
show(f)
```

### Πρόβλεψη

Τα μοντέλα που βρέθηκαν με την βοήθεια της `find_fit` μπορούν να χρησιμοποιηθούν για την πρόβλεψη νέων δεδομένων. 

**Παράδειγμα**

Αν μας δώσουν την θέση και την απόσταση νέων σπιτιών θα προβλέψουμε μία τιμή που να ακολουθεί το μοντέλο μας. Τα δύο σπίτια που μας ενδιαφέρουν βρίσκονται 4 και 6 χιλιόμετρα από το κέντρο και είναι 100 και 115 τετραγωνικά αντίστοιχα. Οι τιμές τους θα προβλεφθούν σύμφωνα με το μοντέλο μας.

```python
interpolation_points = [(4, 100), (6, 115)]
interpolation_results = [(p[0], p[1], f(p[0], p[1])) for p in interpolation_points]
show(f"Παρεμβολή: {interpolation_results}")
```
