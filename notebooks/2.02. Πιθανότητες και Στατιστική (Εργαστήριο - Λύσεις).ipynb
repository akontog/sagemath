# Πιθανότητες

## Ρίψεις νομίσματος

Έστω ότι ρίχνουμε ένα δίκαιο νόμισμα k=5 φορές.
Ο δειγματικός χώρος του πειράματος, για k ρίψεις είναι
$$
\Omega=\{(x_1,\dots,x_k)\mid x_i\in\{H,T\}\}.
$$

1. Χρησιμοποιώντας comprehension κατασκευάστε τον δειγματικό χώρο $\Omega$.
```
{('T', 'T', 'H', 'T', 'H'), ('H', 'H', 'H', 'T', 'H'),... }
```
Στη συνέχεια βρείτε τις πιθανότητες και εμφανίστε τα υποσύνολα:

2. Όπου θα έρθουν ακριβώς l=3 κορώνες (heads).

3. Θα έρθουν τουλάχιστον l=3 γράμματα (tails).

4. Η πρώτη και η τελευταία ρίψη θα είναι 'H'.

5. Θα υπάρχουν 3 συνεχόμενες ρίψεις 'T'

Υποθέτουμε ότι το νόμισμα είναι δίκαιο, οπότε κάθε στοιχείο του $\Omega$ έχει πιθανότητα: $\tfrac{1}{2^k}$.

````python
# 1
sequences = ['']  # ξεκινάμε με κενή ακολουθία
k=5
# Για κάθε μία από τις 5 θέσεις
for i in range(k):
    # Για κάθε υπάρχουσα ακολουθία, προσθέτω και 'H' και 'T'
    sequences = [seq + h for seq in sequences for h in ('H', 'T')]
# Μετατροπή string σε tuple
Ω = Set([tuple(seq) for seq in sequences])
Ω.cardinality(),Ω
````

````python
# 2
l = 3
exl = Set([seq for seq in Ω if seq.count('H') == l])
exl,exl.cardinality()/Ω.cardinality()
````

````python
# 3
l = 3
grl = Set([seq for seq in Ω if seq.count('T') >= l])
grl,grl.cardinality()/Ω.cardinality()
````

````python
# 4
firstlast = Set([seq for seq in Ω 
                 if seq[0] == 'H' and seq[-1] == 'H'])
firstlast,firstlast.cardinality()/Ω.cardinality()
````

````python
# 5
three_c_T = []
for seq in Ω:
    seq_str = ''.join(seq)
    if 'TTT' in seq_str:
        three_c_T.append(seq)
tcT = Set(three_c_T)
tcT,tcT.cardinality()/Ω.cardinality()
````

## Τυχαίος περίπατος, ιστόγραμμα

Έστω τυχαίος περίπατος πραγματοποιείται σε μια διάσταση με κάθε βήμα του περιπάτου να αποτελεί μια τυχαία μετακίνηση είτε προς τα αριστερά είτε προς τα δεξιά.

Δημιουργήστε interact στο οποίο με αντίστοιχα sliders ο χρήστης θα θέτει:

- Των αριθμό προσομοιώσεων, _simulations_, του περιπάτου από 100 έως 5000.

- Τον αριθμό βημάτων, _steps_ από 10 έως 500.

1. Θα εκτελούνται _simulations_, προσομοιώσεις τυχαίων περιπάτων, όπου κάθε περίπατος αποτελείται από _steps_ βήματα. Στο τέλος κάθε περιπάτου, θα καταγράφεται η θέση του περιπατητή δηλαδή η συνολική μετατόπιση από την αρχική θέση. Θα δημιουργείται ιστόγραμμα που θα δείχνει τη συχνότητα των τελικών θέσεων των περιπάτων. Για κάθε βήμα, χρησιμοποιήστε τη συνάρτηση `choice([-1, 1])` για την τυχαία μετακίνηση.

2 Επιπλέον, για μια προσομοίωση και με χρήση της `list_plot` θα σχεδιάζεται γράφημα στο οποίο να φαίνεται η τρέχουσα κάθε φορά θέση για κάθε βήμα.

````python
@interact
def random_walk(
    steps = slider(10, 500, 10, default=100, 
                   label="Αριθμός βημάτων"),
    simulations = slider(100, 5000, 100, default=1000, 
                    label="Αριθμός προσομοιώσεων")
):
    # Λίστα συνολικής μετατόπισης
    displacements = []
    for i in range(simulations):
        dist = 0
        # Προσομοίωση τυχαίου περιπάτου steps βημάτων
        for j in range(steps):
            # Τυχαία μετακίνηση αριστερά ή δεξιά
            dx = choice([-1, 1])  
            dist+=dx
        
        displacements.append(dist)
    show(histogram(displacements, bins=30))
    randomWalk = []
    dist = 0
    for i in range(steps):
        randomWalk.append((i,dist))
        dist+=choice([-1, 1])
    show(list_plot(randomWalk, plotjoined=True))
````

````python
# 1
shapes = ["Hearts", "Diamonds", "Spades", "Clubs"]
values = [2, 3, 4, 5, 6, 7, 8, 9, 10, 
          "Jack", "Queen", "King", "Ace"]

cards = [(s, v) for s in shapes for v in values]
#cards
# 2
@interact
def deal_cards(k = slider(2, 7, 1, default=4, 
                          label="Αριθμός παικτών (k)")):
    # Αντιγραφή τράπουλας και ανακάτεμα
    deck = cards[:]
    shuffle(deck)
    # Μοίρασμα 7 φύλλων σε κάθε παίκτη
    players = [deck[i*7:(i+1)*7] for i in range(k)]
    # Τα υπόλοιπα φύλλα
    remaining = deck[k*7:]
    print(f"Μοίρασμα σε {k} παίκτες (7 φύλλα ο καθένας):\n")
    for i in range(k):
        print(f"Παίκτης {i+1}:")
        print(players[i])
        print()
    print("Υπόλοιπη τράπουλα:")
    print(remaining)
````

## Monty Hall

Το Monty Hall problem, https://en.wikipedia.org/wiki/Monty_Hall_problem, αποτελεί διάσημο πρόβλημα που προέκυψε από το τηλεπαιχνίδι Let's make a Deal που ξεκίνησε την δεκαετία του 60 στην Αμερική.

Στο παιχνίδι υπάρχουν 3 πόρτες όπου:

- πίσω από 1 πόρτα βρίσκεται ένα αυτοκίνητο (που κερδίζει ο παίκτης),

- πίσω από 2 πόρτες υπάρχουν κατσίκες και ο παίκτης δεν κερδίζει τίποτα.

Ο παίκτης επιλέγει αρχικά μία πόρτα. Ο παρουσιαστής ανοίγει μία άλλη πόρτα που έχει κατσίκα. Στην συνέχεια ο παίκτης έχει δύο στρατηγικές:

1. Μένει στην αρχική επιλογή.

2. Αλλάζει επιλογή στην άλλη κλειστή πόρτα.

Με χρήση `slider` ο χρήστης επιλέγει τον αριθμό προσομοιώσεων (100–5000).

Να υπολογιστούν:

- Η πειραματική πιθανότητα επιτυχίας όταν μένει στην επιλογή του.

- Η πειραματική πιθανότητα επιτυχίας όταν αλλάζει επιλογή.

````python
@interact
def monty(simulations = slider(100, 5000, 100, 
                        default=1000, label="Πλήθος προσομοιώσεων")):
    # Αρχικές νίκες, ανάλογα την στρατηγική
    stay_wins = 0
    switch_wins = 0
    
    for i in range(simulations):
        # Τυχαία επιλογή αυτοκινήτου 
        car = randint(1,3)
        # και αρχικής επιλογής παίκτη  
        init_choice = randint(1,3)
        # Όλες οι πόρτες
        doors = {1,2,3}
        # Οι πόρτες με κατσίκες
        goat_doors = doors - {car}
        
        # ο παρουσιαστής επιλέγει πόρτα που δεν επιλέχθηκε
        # από παίκτη (και δεν έχει αυτοκίνητο)
        possible_open = goat_doors - {init_choice}
        opened = choice(list(possible_open))
        
        # η πόρτα που μένει για "switch"
        remaining = doors - {init_choice,opened}
        # στρατηγική stay
        if init_choice == car:
            stay_wins += 1
        # στρατηγική switch
        if remaining == {car}:
            switch_wins += 1
    
    # Πιθανότητες
    p_stay = stay_wins / simulations
    p_switch = switch_wins / simulations
    
    print('Πιθανότητα επιτυχίας αν μείνεις:', p_stay)
    print('Πιθανότητα επιτυχίας αν αλλάξεις:', p_switch)
````

## Κινητός μέσος όρος

Καταγράφουμε τις καθημερινές θερμοκρασίες μιας πόλης κατά τον μήνα Ιούνιο και υπολογίζουμε τον κινητό μέσο όρο 5 ημερών. Οι θερμοκρασίες να δημιουργούνται χρησιμοποιώντας μια τυχαία ομοιόμορφη κατανομή με τις θερμοκρασίες να κυμαίνονται από 20 μέχρι 35. Ο κινούμενος μέσος όρος υπολογίζεται τόσο ως απλός, με ίσα βάρη, όσο και ως σταθμισμένος, με μεγαλύτερα βάρη για τις πιο κοντινές στην ημέρα τιμές. Σχεδιάστε γραφικές παραστάσεις που να δείχνουν τις θερμοκρασίες και τους κινούμενους μέσους όρους στο ίδιο γράφημα αλλά με άλλα χρώματα και διαφορετική εμφάνιση σημείων.

````python
import numpy as np
uniform_t = RealDistribution('uniform', [20, 35])
temps = [uniform_t.get_random_element() for _ in range(30)]
days = list(range(1, len(temps) + 1))
window = 5
moving_avg = np.convolve(temps, 
                         np.ones(window)/window, mode='valid')
weights = np.array([0.1,0.2,0.4,0.2,0.1])
weighted_moving_avg = np.convolve(temps, 
                      weights/weights.sum(), mode='valid')
g = Graphics()
g += list_plot(list(zip(days, temps)), 
               color='blue', marker='o',plotjoined=True)
g += list_plot(list(zip(days[window - 1:], moving_avg)), 
               color='red', marker='x',plotjoined=True)
g += list_plot(list(zip(days[window - 2:], weighted_moving_avg)), 
               color='green', marker='s',plotjoined=True)
g.show(figsize=(4,3))
````
